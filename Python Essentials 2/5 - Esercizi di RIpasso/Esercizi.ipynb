````python
while True:
    a = input("Inserisci un numero: ")
    try:
        a = int(a)
        break
    except ValueError:
        print("Devi inserire un intero")
        continue
    finally:
        print(f"Hai inserito {a}")
````

### Esercizio 1.1: Il generatore di numeri dispari
Scrivi una funzione generatore chiamata `dispari_fino_a(n)` che utilizzi l'istruzione `yield` per produrre tutti i numeri dispari da `1` fino a `n`. Testala con un ciclo for.

````python
def dispari_fino_a(n):
    for i in range(n+1):
        if i % 2 != 0:
            yield i
````

### Esercizio 1.5: Generatore infinito limitato
Crea un generatore che produca la sequenza infinita delle potenze di 2 ($2^1, 2^2, 2^3, \dots$). Nel codice principale, usa un ciclo e un contatore per fermare l'estrazione e stampare solo le prime 10 potenze, evitando un ciclo infinito.

````python
def pwr():
    c = 0
    while True:
        yield 2**c
        c += 1


for c, i in enumerate(pwr()):
    if c > 10:
        break
    print(i)

````

````python
gen = pwr()
contatore = 0
for _ in range(10+1):
    valore = next(gen)
    print(f"2^{contatore} = {valore}")
    contatore += 1
````

## L'Esercizio Flash: Il Generatore di Scadenze Notturne

Scrivi un piccolo script che faccia questo:Il Generatore di Giorni: 
1. Crea una funzione generatore chiamata `prossimi_giorni(n)`. Utilizzando un ciclo e `yield`, deve produrre gli oggetti `date` (o `datetime`) dei prossimi $n$ giorni a partire da oggi. (Usa `timedelta` per sommare i giorni di volta in volta).
2. Il Controllo del Weekend (Operatore Ternario): Nel ciclo principale, estrai i giorni dal generatore (imposta $n=5$). Per ogni data ricevuta, usa il metodo `.weekday()` e un operatore ternario per definire una stringa: "Weekend" se il giorno è sabato (5) o domenica (6), altrimenti "Giorno Lavorativo".
3. La Verifica Bisestile: Prima di chiudere il ciclo, usa la funzione dedicata del modulo calendar per controllare al volo se l'anno di quella data è bisestile, e stampa tutto in una riga pulita.

````python
from datetime import date, timedelta

def prossimi_giorni(n):
    oggi = date.today()
    for i in range(n):
        yield oggi + timedelta(days=i)


for d in prossimi_giorni(10):
    print(f"{d}: {'Weekend' if d.weekday() >= 5 else 'Giorno Lavorativo'}, {'Bisestile' if calendar.isleap(d.year) else 'Non Bisestile'}")
````
