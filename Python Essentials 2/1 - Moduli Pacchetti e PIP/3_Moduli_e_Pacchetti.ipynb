# Che cos'è un pacchetto (*package*)?

Scrivere i propri moduli non differisce molto dallo scrivere normali script.

Ci sono alcuni aspetti specifici di cui devi essere consapevole, ma non si tratta certamente di qualcosa di trascendentale. Lo vedrai molto presto.

Riassumiamo alcune questioni importanti:

* Un modulo è una sorta di **contenitore pieno di funzioni**: puoi inserire tutte le funzioni che desideri in un singolo modulo e distribuirlo in tutto il mondo.
* Naturalmente, è un'ottima idea evitare di mescolare all'interno dello stesso modulo funzioni destinate ad ambiti applicativi diversi (proprio come in una biblioteca, nessuno si aspetta di trovare saggi scientifici in mezzo ai fumetti). Pertanto, raggruppa le tue funzioni con attenzione e assegna al modulo che le contiene un nome chiaro e intuitivo (ad esempio, non dare il nome `giochi_arcade` a un modulo che contiene funzioni destinate a partizionare e formattare i dischi rigidi).


Creare molti moduli, alla lunga, può generare un po' di disordine. Prima o poi sentirai l'esigenza di raggruppare i tuoi moduli esattamente nello stesso modo in cui in precedenza hai raggruppato le funzioni. Esiste un contenitore più generale rispetto a un modulo?

Sì, esiste: si chiama **pacchetto** (*package*). Nel mondo dei moduli, un pacchetto svolge un ruolo del tutto simile a quello di una cartella o directory nel mondo dei file.

Abbiamo chiamato il file `module.py`. Non è molto creativo, ma è semplice e chiaro.

## Passo 2

Il secondo file contiene il codice che utilizza il nuovo modulo. Il suo nome è `main.py` e per ora il suo contenuto è molto breve:

```python
import module

```

> Entrambi i file devono trovarsi nella stessa cartella. Ti consigliamo caldamente di creare una nuova cartella vuota per ospitarli entrambi: renderà le cose più semplici.

Avvia IDLE (o qualsiasi altro IDE che preferisci) ed esponi in esecuzione il file `main.py`. Cosa vedi?
Non dovresti vedere nulla. Questo significa che Python ha importato con successo il contenuto del file `module.py`.

Non importa se il modulo per ora è vuoto. Il primissimo passo è stato fatto, ma prima di passare al successivo, dai un'occhiata alla cartella in cui si trovano i due file.
Noti qualcosa di interessante?

È apparsa una nuova sottocartella: riesci a vederla? Il suo nome è `__pycache__`. Guarda cosa c'è dentro.
Troverai un file chiamato (più o meno) `module.cpython-xy.pyc`, dove `x` e `y` sono cifre che derivano dalla tua versione di Python (ad esempio, saranno `3` e `8` se usi Python 3.8).

Il nome del file coincide con quello del tuo modulo (in questo caso `module`). La parte dopo il primo punto indica quale implementazione di Python ha creato il file (`CPython`) e il relativo numero di versione. L'ultima parte (`pyc`) deriva dalle parole *Python* e *compiled* (compilato).

Se provi a guardare dentro questo file, vedrai che il contenuto è completamente illeggibile per un essere umano. È normale che sia così, poiché si tratta di un file destinato all'uso esclusivo di Python.

Quando Python importa un modulo per la prima volta, ne traduce il contenuto in una forma parzialmente compilata. Non si tratta di codice macchina, bensì di un codice semi-compilato interno a Python (**bytecode**), pronto per essere eseguito dall'interprete. In questo modo il file non richiede molti dei controlli necessari per un file sorgente puro: l'esecuzione si avvia più rapidamente e il programma gira anche più velocemente.

Grazie a questo meccanismo, ogni importazione successiva sarà molto più rapida rispetto a dover reinterpretare il testo del codice sorgente da zero.
Python è in grado di verificare in automatico se il file sorgente del modulo è stato modificato: in caso positivo, il file `.pyc` verrà rigenerato; in caso contrario, verrà eseguito immediatamente il file `.pyc` già pronto. Poiché questo processo è completamente automatico e trasparente, non devi preoccupartene.

Che cosa significa concretamente?
Quando un modulo viene importato, **il suo contenuto viene implicitamente eseguito da Python**. Questo dà al modulo la possibilità di inizializzare alcuni dei suoi aspetti interni (ad esempio, può assegnare valori utili a determinate variabili).

> L'inizializzazione avviene **una sola volta**, ovvero quando avviene il primo import. In questo modo le assegnazioni eseguite dal modulo non vengono ripetute inutilmente.

Immagina questo scenario:

* Esiste un modulo chiamato `mod1`;
* Esiste un modulo chiamato `mod2` che contiene l'istruzione `import mod1`;
* Esiste un file principale (`main`) che contiene sia l'istruzione `import mod1` sia `import mod2`.

A prima vista si potrebbe pensare che `mod1` verrà importato due volte. Fortunatamente, viene eseguita solo la prima importazione. Python tiene traccia dei moduli già importati e salta silenziosamente tutte le importazioni successive dello stesso modulo.

## Passo 5

Python fa molto di più che importare semplicemente il modulo: crea anche una variabile speciale chiamata `__name__`.
Inoltre, ogni file sorgente utilizza la propria versione separata di questa variabile, che non viene condivisa tra i moduli.

Vediamo come usarla. Modifica leggermente il modulo:

```python
print("Mi piace essere un modulo.")
print(__name__)

```


Ora esegui direttamente il file `module.py`. Dovresti vedere le seguenti righe:

**Output:**

```text
Mi piace essere un modulo.
__main__

```

Ora esegui invece il file `main.py`. Vedi lo stesso risultato?

**Output:**

```text
Mi piace essere un modulo.
module

```

Possiamo sintetizzare la regola in questo modo:

* Quando si esegue un file direttamente, la sua variabile `__name__` viene impostata su `__main__`;
* Quando un file viene importato come modulo, la sua variabile `__name__` viene impostata sul **nome del file stesso** (esclusa l'estensione `.py`).

## Passo 6

Ecco come puoi sfruttare la variabile `__name__` per rilevare il contesto in cui il tuo codice è stato attivato:

```python
if __name__ == "__main__":
    print("Preferisco essere un modulo.")
else:
    print("Mi piace essere un modulo.")

```


Esiste tuttavia un modo ancora più intelligente di utilizzare questa variabile. Se scrivi un modulo contenente una serie di funzioni complesse, puoi sfruttare questo meccanismo per inserire dei test di verifica del corretto funzionamento delle funzioni.
Ogni volta che modifichi una di queste funzioni, ti basterà eseguire direttamente il modulo per assicurarti che le modifiche non abbiano compromesso il codice. Questi test verranno poi ignorati quando il codice verrà importato come modulo all'interno di un altro file.

```python
counter = 0
   
if __name__ == "__main__":
    print("Preferisco essere un modulo.")
else:
    print("Mi piace essere un modulo.")

```


## Passo 8

L'introduzione di una variabile di questo tipo è assolutamente corretta, ma può causare importanti effetti collaterali di cui devi essere consapevole.
Dai un'occhiata al file `main.py` modificato:

```python
import module
print(module.counter)

```


Come puoi vedere, il file principale tenta di accedere alla variabile `counter` del modulo. È legale? Sì, lo è. È utile? Può esserlo molto. È sicuro?
Dipende. Se ti fidi degli utenti del tuo modulo non c'è problema; tuttavia, potresti preferire che il resto del mondo non veda o non acceda a una tua variabile personale o privata.

A differenza di molti altri linguaggi di programmazione, Python non ha strumenti rigidi per nascondere tali variabili agli occhi degli utenti del modulo. Puoi solo informare chi usa il modulo che si tratta di una variabile interna: possono leggerla, ma non dovrebbero modificarla in nessun caso.

Questo si fa facendo precedere il nome della variabile da un underscore (`_`) o da due underscore (`__`). Ricorda però che **si tratta solo di una convenzione**. Gli utenti del tuo modulo potrebbero rispettarla oppure no.

Naturalmente noi seguiremo la convenzione. Ora inseriamo due funzioni nel modulo: calcoleranno la somma e il prodotto dei numeri contenuti in una lista. Inoltre, aggiungeremo delle rifiniture e rimuoveremo i vecchi residui superflui.

## Passo 9

Scriviamo del codice completamente nuovo nel nostro file `module.py`. Ecco il modulo aggiornato e pronto:

```python
#!/usr/bin/env python3
 
""" module.py - un esempio di modulo Python """
 
__counter = 0
 
 
def suml(the_list):
    global __counter
    __counter += 1
    the_sum = 0
    for element in the_list:
        the_sum += element
    return the_sum
 
 
def prodl(the_list):
    global __counter
    __counter += 1
    prod = 1
    for element in the_list:
        prod *= element
    return prod
 
 
if __name__ == "__main__":
    print("Preferisco essere un modulo, ma posso fare dei test per te.")
    my_list = [i+1 for i in range(5)]
    print(suml(my_list) == 15)
    print(prodl(my_list) == 120)

```


Ci sono alcuni elementi che meritano una spiegazione:

* La riga che inizia con `#!` ha molti nomi: viene chiamata *shabang*, *shebang*, *hashbang*, *poundbang* o persino *hashpling*. Dal punto di vista di Python è semplicemente un commento, dato che inizia con `#`. Per i sistemi operativi Unix e Unix-like (incluso macOS), questa riga indica al sistema operativo come eseguire il contenuto del file (in altre parole, quale programma lanciare per interpretare il testo). In alcuni ambienti (specialmente quelli legati ai server web) l'assenza di questa riga può causare problemi;
* Una stringa (anche multilinea) posizionata prima di qualsiasi istruzione del modulo (comprese le importazioni) è chiamata **doc-string** (stringa di documentazione) e serve a spiegare brevemente lo scopo e il contenuto del modulo;
* Le funzioni definite all'interno del modulo (`suml()` e `prodl()`) sono disponibili per essere importate;
* Abbiamo usato la variabile `__name__` per rilevare quando il file viene eseguito in modalità stand-alone, cogliendo l'occasione per eseguire alcuni semplici test di verifica.

## Passo 10

Ora è possibile utilizzare il modulo aggiornato. Ecco un modo per farlo:

```python
from module import suml, prodl
 
zeroes = [0 for i in range(5)]
ones = [1 for i in range(5)]
print(suml(zeroes))
print(prodl(ones))

```

# Il tuo primo pacchetto (*package*)

## Passo 1

Immagina che in un futuro non troppo lontano tu e i tuoi collaboratori scriviate un gran numero di funzioni Python.
Il tuo team decide di raggruppare le funzioni in moduli separati, e questo è il risultato finale del riordino:

```python
#! /usr/bin/env python3
 
""" modulo: alpha """
 
def funA():
    return "Alpha"
 
if __name__ == "__main__":
    print("Preferisco essere un modulo.")

```

> Abbiamo mostrato l'intero contenuto solo per il modulo `alpha.py`. Immagina che tutti gli altri moduli siano simili (contengono una sola funzione chiamata `funX`, dove `X` è la prima lettera del nome del modulo).

## Passo 2

All'improvviso, qualcuno nota che questi moduli formano una vera e propria gerarchia, quindi metterli tutti insieme in una struttura piatta (ovvero nella stessa cartella, senza sottocartelle) non è una buona idea.

Dopo una breve discussione, il team giunge alla conclusione che i moduli devono essere raggruppati. Tutti i partecipanti concordano sul fatto che la seguente struttura ad albero rifletta perfettamente le relazioni reciproche tra i moduli:

* **extra** (il gruppo principale)
* **iota** (modulo)
* **good** (sottogruppo)
* **alpha** (modulo)
* **beta** (modulo)
* **best** (sottogruppo)
* **sigma** (modulo)
* **tau** (modulo)


* **bad** (sottogruppo)
* **ugly** (sottogruppo)
* **psi** (modulo)
* **omega** (modulo)


Esaminiamola dal basso verso l'alto:

* Il gruppo `ugly` contiene due moduli: `psi` e `omega`;
* Il gruppo `best` contiene due moduli: `sigma` e `tau`;
* Il gruppo `good` contiene due moduli (`alpha` e `beta`) e un sottogruppo (`best`);
* Il gruppo `extra` contiene due sottogruppi (`good` e `bad`) e un modulo (`iota`).

Analizza attentamente la struttura. Ricorda qualcosa, vero? Assomiglia esattamente alla struttura delle cartelle di un computer. Costruiamo quindi un albero che rifletta le dipendenze previste tra i moduli.

## Passo 3

Una struttura di questo tipo è quasi un pacchetto (nel senso inteso da Python). Le manca solo un piccolo dettaglio per essere sia funzionale che operativa. La completeremo tra un momento.

Se assumiamo che `extra` sia il nome di un pacchetto appena creato (pensa ad esso come alla radice del pacchetto), questo imporrà una regola di denominazione che ti permetterà di identificare chiaramente ogni entità dell'albero.

Ad esempio:

* La posizione di una funzione chiamata `funT()` all'interno del modulo `tau` può essere descritta come: `extra.good.best.tau.funT()`
* Una funzione indicata come: `extra.bad.ugly.psi.funP()` proviene dal modulo `psi`, che è memorizzato nel sottopacchetto `ugly` del pacchetto `extra`.

## Passo 4

Ora ci sono due domande a cui rispondere:

1. Come si trasforma un albero di questo tipo (in realtà, un sottoalbero) in un vero e proprio pacchetto Python? In altre parole, come si convince Python che questo albero non è solo un mucchio di file sparsi, ma un insieme organizzato di moduli?
2. Dove si inserisce questo sottoalbero per renderlo accessibile a Python?

La prima domanda ha una risposta sorprendente: i pacchetti, proprio come i moduli, possono richiedere un'inizializzazione.

L'inizializzazione di un modulo viene eseguita dal codice slegato da funzioni (ovvero non inserito in alcuna funzione) che si trova all'interno del file del modulo stesso. Poiché un pacchetto non è un singolo file ma una cartella, questa tecnica non è applicabile.

È necessario utilizzare un trucco diverso: Python si aspetta che all'interno della cartella del pacchetto sia presente un file con un nome molto specifico: **`__init__.py`**.

Il contenuto di questo file viene eseguito quando viene importato uno qualsiasi dei moduli del pacchetto. Se non hai bisogno di inizializzazioni particolari, puoi lasciare il file completamente vuoto, ma **non puoi ometterlo**.

## Passo 5

> È proprio la presenza del file `__init__.py` a definire e costituire ufficialmente il pacchetto.

Nota: non è solo la cartella radice a dover contenere il file `__init__.py`: puoi inserirlo anche all'interno di una qualsiasi delle sue sottocartelle (sottopacchetti). Questo è utile se alcuni sottopacchetti richiedono un trattamento individuale e tipi speciali di inizializzazione.

Ora è il momento di rispondere alla seconda domanda: dove si inserisce l'albero per renderlo accessibile a Python? La risposta è semplice: ovunque. Devi solo assicurarti che Python sia a conoscenza della posizione del pacchetto, e sai già come fare. Sei pronto a utilizzare il tuo primo pacchetto.

## Passo 7

Vogliamo accedere alla funzione `funI()` dal modulo `iota` partendo dall'alto del pacchetto `extra`. Questo ci obbliga a usare i nomi di pacchetto qualificati (pensa a come si nominano le cartelle e le sottocartelle: le convenzioni sono molto simili).

Ecco come fare:

```python
from sys import path
path.append('..\\packages')
 
import extra.iota
print(extra.iota.funI())

```

Nota bene:

* Abbiamo modificato la variabile `path` per rendere la cartella `packages` accessibile a Python;
* L'istruzione `import` non punta direttamente al modulo, ma specifica il percorso completo e qualificato a partire dalla cima del pacchetto;
* Sostituire `import extra.iota` con `import iota` causerà un errore di importazione.

Anche la seguente variante è valida:

```python
from sys import path
path.append('..\\packages')
 
from extra.iota import funI
print(funI())

```

Nota il nome qualificato del modulo `iota`.

## Passo 8

Ora spingiamoci fino in fondo all'albero: ecco come ottenere l'accesso ai moduli `sigma` e `tau`:

```python
from sys import path
path.append('..\\packages')
 
import extra.good.best.sigma
from extra.good.best.tau import funT
 
print(extra.good.best.sigma.funS())
print(funT())

```

Puoi semplificarti la vita utilizzando gli alias (`as`):

```python
from sys import path
path.append('..\\packages')
 
import extra.good.best.sigma as sig
import extra.good.alpha as alp
 
print(sig.funS())
print(alp.funA())

```

## Passo 9

Supponiamo ora di aver compresso l'intera sottocartella in un archivio zip, partendo dalla cartella `extra` (inclusa), ottenendo un file chiamato `extrapack.zip`. Successivamente, inseriamo questo file zip all'interno della cartella `packages`.

Siamo in grado di utilizzare il file zip direttamente come se fosse una cartella di pacchetti:

```python
from sys import path
path.append('..\\packages\\extrapack.zip')
 
import extra.good.best.sigma as sig
import extra.good.alpha as alp
from extra.iota import funI
from extra.good.beta import funB
 
print(sig.funS())
print(alp.funA())
print(funI())
print(funB())

```

# Riepilogo della sezione

1. Mentre un modulo è progettato per unire entità correlate come funzioni, variabili o costanti, un **pacchetto** (*package*) è un contenitore che permette di raggruppare più moduli correlati sotto un unico nome comune. Tale contenitore può essere distribuito così com'è (come un insieme di file organizzati in una struttura di sottocartelle) oppure può essere compresso all'interno di un file `.zip`.
2. Durante la primissima importazione di un modulo, Python ne traduce il codice sorgente in un formato semi-compilato (**bytecode**) memorizzato all'interno di file `.pyc`, e inserisce questi ultimi nella cartella `__pycache__` situata nella directory principale del modulo stesso.
3. Se desideri segnalare all'utente del tuo modulo che una determinata entità deve essere trattata come **privata** (ovvero non destinata a essere utilizzata esplicitamente al di fuori del modulo), puoi far precedere il suo nome da un singolo (`_`) o da un doppio (`__`) carattere di underscore. Ricorda sempre che questa è solo una raccomandazione e non un divieto categorico.
4. I termini *shabang*, *shebang*, *hashbang*, *poundbang* e *hashpling* descrivono il digramma scritto come `#!`, utilizzato per indicare ai sistemi operativi di tipo Unix (come Linux e macOS) in che modo debba essere avviato il file sorgente Python. Questa convenzione non ha alcun effetto sui sistemi operativi MS Windows.
5. Se vuoi fare in modo che Python prenda in considerazione la cartella di un pacchetto non standard, il percorso di quest'ultima deve essere inserito o aggiunto alla lista dei percorsi di importazione, memorizzata nella variabile **`path`** all'interno del modulo **`sys`**.
6. Un file Python denominato **`__init__.py`** viene eseguito implicitamente quando il pacchetto che lo contiene viene importato. Viene utilizzato per inizializzare il pacchetto e/o i suoi eventuali sottopacchetti. Il file può anche essere completamente vuoto, ma la sua presenza è obbligatoria.
