# L'ecosistema dei pacchetti di Python e come usarlo

Python è uno strumento estremamente potente — e speriamo che tu l'abbia già sperimentato di persona. Moltissime persone in tutto il mondo la pensano allo stesso modo e usano Python regolarmente per sviluppare e potenziare le proprie attività in campi completamente diversi. Ciò significa che Python è diventato uno strumento interdisciplinare, impiegato in innumerevoli applicazioni. Non possiamo passare in rassegna tutti i settori in cui Python mostra brillantemente le sue capacità, quindi ti parleremo solo di quelli più impressionanti.

In primo luogo, Python è diventato il leader indiscusso nella ricerca sull'**intelligenza artificiale**. Anche il *data mining*, una delle discipline scientifiche moderne più promettenti, si serve di Python. Matematici, psicologi, genetisti, meteorologi, linguisti: tutte queste persone usano già Python o, se non lo fanno ancora, siamo certi che inizieranno molto presto. È un trend da cui non si può sfuggire.

Naturalmente, non avrebbe alcun senso costringere tutti gli utenti di Python a scrivere il proprio codice da zero, tenendoli perfettamente isolati dal mondo esterno e dai traguardi raggiunti dagli altri programmatori. Sarebbe sia innaturale che controproducente.

La cosa più auspicabile ed efficiente è **consentire a tutti i membri della community di Python di scambiarsi liberamente codice ed esperienze**. In questo modello, nessuno è costretto a iniziare il lavoro da zero, poiché vi è un'alta probabilità che qualcun altro stia già lavorando (o abbia lavorato) sullo stesso problema o su uno molto simile.

Come sai, Python è nato come software open-source, e questo rappresenta anche un invito per tutti i programmatori a mantenere l'intero ecosistema di Python come un ambiente aperto, amichevole e gratuito. Per far sì che questo modello funzioni ed evolva, è necessario fornire alcuni strumenti aggiuntivi: tool che aiutino i creatori a pubblicare, gestire e prendersi cura del proprio codice.

Questi stessi strumenti devono aiutare gli utenti a sfruttare il codice, sia quello già esistente, sia quello nuovo che compare ogni giorno. Grazie a ciò, scrivere nuovo codice per affrontare nuove sfide non è come costruire una casa partendo ogni volta dalle fondamenta.

Inoltre, il programmatore è libero di modificare il codice di qualcun altro per adattarlo alle proprie esigenze e, di fatto, costruire un prodotto completamente nuovo che potrà poi essere utilizzato da un altro sviluppatore ancora. Questo processo sembra non avere fine. Fortunatamente.

Per far girare questo mondo, devono essere stabilite e mantenute in movimento due entità fondamentali:

1. Un **repository centralizzato** di tutti i pacchetti software disponibili;
2. Uno **strumento** (tool) che consenta agli utenti di accedere a tale repository.

Entrambe queste entità esistono già e possono essere utilizzate in qualsiasi momento.

Il repository (o *repo* in breve) che abbiamo menzionato prima si chiama **PyPI** (acronimo di *Python Package Index*) ed è gestito da un gruppo di lavoro chiamato *Packaging Working Group*, una sezione della *Python Software Foundation*, il cui compito principale è supportare gli sviluppatori Python nella diffusione efficiente del codice.

Puoi trovare il loro sito web qui: [https://wiki.python.org/psf/PackagingWG]().

Il sito web di PyPI (amministrato dal PWG) si trova invece all'indirizzo: [https://pypi.org/]().

Quando abbiamo dato un'occhiata veloce all'inizio di luglio 2021, abbiamo scoperto che PyPI ospitava ben 315.000 progetti, composti da oltre 4.500.000 file gestiti da 520.000 utenti.

Questi tre numeri da soli mostrano chiaramente la potenza della community di Python e l'importanza della cooperazione tra gli sviluppatori.

Dobbiamo precisare che PyPI non è l'unico repository Python esistente. Al contrario, ce ne sono moltissimi, creati per progetti specifici e gestiti da comunità Python più o meno grandi. È probabile che un giorno tu e i tuoi colleghi desideriate creare il vostro repository personale.

In ogni caso, PyPI rimane il repository Python più importante al mondo. Modificando leggermente il classico proverbio, possiamo affermare che "tutte le strade di Python portano a PyPI", e questa non è affatto un'esagerazione.

PyPI è completamente gratuito e ti basta scegliere il codice che desideri per usarlo; non incontrerai né cassieri né guardie giurate. Naturalmente questo non ti esime dall'essere educato e onesto: devi comunque rispettare tutti i termini delle licenze d'uso, quindi non dimenticare di leggerle.

"D'accordo", potresti dire, "il concetto di negozio adesso è chiaro, ma cosa c'entra il formaggio con Python?"

*The Cheese Shop* è uno dei più famosi sketch dei **Monty Python** (il gruppo comico britannico da cui il linguaggio prende il nome). Rappresenta la surreale avventura di un uomo inglese che tenta di comprare del formaggio. ### Lo strumento per gli acquisti: `pip`

PyPI è un negozio molto specifico, non solo perché offre tutti i suoi prodotti gratuitamente, ma anche perché richiede uno strumento speciale per poter essere utilizzato.

Fortunatamente anche questo strumento è gratuito. Quindi, se vuoi comporre il tuo "cheeseburger digitale" utilizzando i prodotti offerti dal negozio PyPI, avrai bisogno di un tool gratuito chiamato **`pip`**.

No, non hai capito male. Proprio `pip`. Si tratta ovviamente di un altro acronimo, ma la sua natura è più complessa rispetto al già citato PyPI: è infatti un esempio di **acronimo ricorsivo**. Ciò significa che l'acronimo fa riferimento a se stesso, rendendo la sua spiegazione un processo infinito.

Perché? Perché `pip` significa *"pip installs packages"* (pip installa i pacchetti), e il `pip` contenuto all'interno di *"pip installs packages"* significa a sua volta *"pip installs packages"*, e così via...

Fermiamoci qui, grazie per la collaborazione.

> Esistono altri acronimi ricorsivi molto famosi nel mondo dell'informatica. Uno dei più noti è **Linux**, che può essere interpretato come *"Linux is Not Unix"* (Linux non è Unix), oppure **GNU**, che sta per *"GNU's Not Unix"*.

# Le dipendenze (*Dependencies*)

Ora che siamo certi che `pip` sia pronto ai nostri comandi, per il momento limiteremo la nostra attenzione al solo sistema operativo MS Windows, dato che il suo comportamento è (o dovrebbe essere) identico in tutti i sistemi operativi. Prima di iniziare, però, dobbiamo spiegare un concetto fondamentale e parlarti delle **dipendenze**.

Immagina di aver creato una geniale applicazione Python chiamata `redsuspenders`, in grado di prevedere l'andamento dei mercati azionari con un'accuratezza del 99% (a proposito, se ci riesci davvero, contattaci immediatamente!).

Ovviamente, per raggiungere questo obiettivo hai utilizzato del codice già esistente: ad esempio, la tua app importa un pacchetto chiamato `nyse` che contiene alcune funzioni e classi cruciali. A sua volta, il pacchetto `nyse` importa un altro pacchetto chiamato `wallstreet`, il quale importa altri due pacchetti essenziali chiamati `bull` e `bear`.

Come avrai già intuito, i collegamenti tra questi pacchetti sono vitali. Se qualcuno decidesse di utilizzare il tuo codice (ma ricordati che abbiamo la precedenza noi!), dovrà assicurarsi che tutti i pacchetti richiesti siano installati e presenti nel sistema.

In parole povere, una **dipendenza** è un fenomeno che si manifesta ogni volta che utilizzi un software che si affida a un altro software per funzionare. Nota che una dipendenza può includere (e generalmente include) più livelli di sviluppo software concatenati.

Significa forse che un potenziale utente del pacchetto `nyse` sia obbligato a tracciare tutte le dipendenze e a installare manualmente ogni singolo pacchetto necessario? Sarebbe terribile, vero?

Sì, è decisamente orribile, e non dovrebbe sorprenderti che il processo di soddisfare faticosamente tutti i requisiti successivi abbia un nome preciso: nel mondo dell'informatica viene chiamato **inferno delle dipendenze** (*dependency hell*).

Come possiamo risolvere questo problema? Ogni utente è condannato a visitare l'inferno solo per avviare un codice per la prima volta?

Fortunatamente no: **`pip` può fare tutto questo al posto tuo**. È in grado di scoprire, identificare e risolvere tutte le dipendenze. Inoltre, lo fa nel modo più intelligente possibile, evitando download e reinstallazioni inutili.

## Come usare pip

Ora siamo pronti a chiedere a `pip` cosa può fare per noi. Facciamolo, immettendo il seguente comando nel terminale (o prompt dei comandi):

```bash
pip help

```

### Capire i comandi di `pip`

L'esecuzione di questo comando mostrerà l'elenco completo delle opzioni e dei comandi disponibili. Di seguito sono riportati i comandi principali che utilizzerai più spesso:

| Comando | Descrizione |
| --- | --- |
| **`pip install <nome_pacchetto>`** | Scarica e installa un pacchetto da PyPI, risolvendo automaticamente tutte le sue dipendenze. |
| **`pip uninstall <nome_pacchetto>`** | Rimuove un pacchetto precedentemente installato dal sistema. |
| **`pip list`** | Mostra l'elenco di tutti i pacchetti Python attualmente installati nell'ambiente corrente. |
| **`pip show <nome_pacchetto>`** | Fornisce informazioni dettagliate su un pacchetto specifico (versione, autore, dipendenze, ecc.). |
| **`pip search <termine>`** | Consente di cercare tra i pacchetti presenti su PyPI (nota: questa funzione potrebbe essere disabilitata o limitata sul server PyPI ufficiale per motivi di sovraccarico). |


> Se desideri ricevere aiuto o maggiori dettagli su un comando specifico, puoi combinare `pip help` con il nome del comando stesso. Ad esempio, digitando `pip help install` visualizzerai la guida dettagliata dedicata solo al processo di installazione.

Se non ami leggere i testi all'interno della console, puoi utilizzare il metodo alternativo per esplorare i contenuti di PyPI offerto dal motore di ricerca disponibile all'indirizzo: [https://pypi.org/search](https://pypi.org/search).

Supponendo che la tua ricerca vada a buon fine (o che tu sia determinato a installare un pacchetto specifico di cui conosci già il nome), puoi usare `pip` per installarlo sul tuo computer.

A questo punto si possono configurare due possibili scenari:

1. **Vuoi installare un nuovo pacchetto solo per te:** non sarà disponibile per nessun altro utente (account) presente sul computer. Questa procedura è l'unica praticabile se non disponi di privilegi elevati e non puoi agire come amministratore di sistema.
2. **Hai deciso di installare il pacchetto a livello di sistema (*system-wide*):** disponi dei diritti di amministratore e non hai paura di usarli.

Per distinguere queste due azioni, `pip` utilizza un'opzione dedicata chiamata **`--user`** (nota il doppio trattino). La presenza di questa opzione indica a `pip` di agire localmente per conto del tuo utente (non amministratore). Se non la inserisci, `pip` darà per scontato che tu sia un amministratore di sistema e non farà nulla per correggerti qualora tu non lo sia.

Nel nostro caso, installeremo un pacchetto chiamato **`pygame`**: si tratta di una libreria estesa e complessa che permette ai programmatori di sviluppare videogiochi utilizzando Python. Se desideri saperne di più sul progetto e sulla community che lo guida, puoi visitare il sito [https://www.pygame.org](https://www.pygame.org).

Se sei un amministratore di sistema, puoi installare `pygame` usando il seguente comando:

```bash
pip install pygame

```

Se invece non sei un amministratore, o semplicemente non vuoi appesantire il tuo sistema operativo installando `pygame` a livello globale, puoi installarlo solo per il tuo utente:

```bash
pip install --user pygame

```

La scelta di quale delle due procedure adottare spetta interamente a te.

`pip` ha l'abitudine di mostrare una simpatica animazione testuale che indica l'avanzamento dell'installazione, quindi tieni d'occhio lo schermo per non perderti lo spettacolo! Se il processo va a buon fine, vedrai qualcosa di simile a questo:

```text
Collecting pygame
  Downloading pygame-2.1.2-cp39-cp39-win_amd64.whl (8.4 MB)
Installing collected packages: pygame
Successfully installed pygame-2.1.2

```

Ti invitiamo a utilizzare i seguenti comandi:

```bash
pip show pygame

```

e

```bash
pip list

```

per ottenere maggiori informazioni su ciò che è effettivamente accaduto sul tuo computer.

## Un semplice programma di test
Ora che pygame è finalmente accessibile, possiamo provare a utilizzarlo in un semplicissimo programma di test:

```python
import pygame

run = True
width = 400
height = 100

# Inizializza i moduli di pygame
pygame.init()

# Crea la finestra dell'applicazione
screen = pygame.display.set_mode((width, height))

# Imposta il titolo della finestra
pygame.display.set_caption("Il mio primo programma con Pygame")

# Scegli il font e la dimensione del testo
font = pygame.font.SysFont("arial", 48)

# Esegui il rendering del testo in una superficie (surface)
text = font.render("Funziona!", True, (255, 255, 255))

# Ciclo principale del programma
while run:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
            
    # Riempie lo schermo di nero
    screen.fill((0, 0, 0))
    
    # Disegna il testo sullo schermo
    screen.blit(text, (width // 2 - text.get_width() // 2, height // 2 - text.get_height() // 2))
    
    # Aggiorna il display
    pygame.display.flip()
```
