# Cos'è un modulo?

Il codice informatico ha la tendenza a crescere. Potremmo dire che il codice che non cresce è probabilmente del tutto inutilizzabile o abbandonato. Un codice reale, desiderato e ampiamente utilizzato si sviluppa continuamente, poiché sia le richieste che le aspettative degli utenti evolvono secondo i propri ritmi.

Un codice che non è in grado di rispondere alle esigenze degli utenti verrà dimenticato rapidamente e istantaneamente sostituito da uno nuovo, migliore e più flessibile. Sii pronto a questo, e non pensare mai che uno dei tuoi programmi sia finalmente completato. La "completezza" è uno stato di transizione e solitamente passa in fretta, subito dopo la prima segnalazione di bug. Python stesso è un ottimo esempio di come agisce questa regola.

Il codice che cresce è, di fatto, un problema che cresce. Un codice più ampio significa sempre una manutenzione più difficile. La ricerca dei bug è sempre più semplice dove il codice è più piccolo (proprio come trovare un guasto meccanico è più semplice quando il macchinario è più semplice e piccolo).

Inoltre, quando si prevede che il codice in fase di creazione diventi molto grande (puoi usare il numero totale di righe di codice sorgente come misura utile, sebbene non molto accurata, della dimensione di un codice), potresti voler (o meglio, sarai costretto a) dividerlo in molte parti, implementate in parallelo da pochi, una dozzina, diverse dozzine o persino diverse centinaia di singoli sviluppatori.

Naturalmente, questo non può essere fatto utilizzando un unico grande file sorgente modificato da tutti i programmatori contemporaneamente. Ciò porterebbe sicuramente a un disastro spettacolare.

Se vuoi che un progetto software di questo tipo venga completato con successo, devi disporre dei mezzi che ti consentano di:

* **Dividere** tutti i compiti tra gli sviluppatori;
* **Unire** tutte le parti create in un unico insieme funzionante.

Per esempio, un determinato progetto può essere diviso in due parti principali:

1. **L'interfaccia utente** (la parte che comunica con l'utente tramite widget e uno schermo grafico);
2. **La logica** (la parte che elabora i dati e produce i risultati).

Ciascuna di queste parti può essere (molto probabilmente) divisa in parti più piccole, e così via. Tale processo è spesso chiamato **decomposizione**.

Per fare un esempio, se ti venisse chiesto di organizzare un matrimonio, non faresti tutto da solo: cercheresti un certo numero di professionisti e divideresti il compito tra tutti loro.

Come si divide un software in parti separate ma cooperanti? Questa è la domanda. **I moduli sono la risposta.**

# Come utilizzare un modulo?

Quindi, cos'è un modulo? Il tutorial di Python lo definisce come un file contenente definizioni e istruzioni Python, che può essere successivamente importato e utilizzato quando necessario.
La gestione dei moduli si divide in due diverse questioni:

* **La prima** (probabilmente la più comune) si verifica quando vuoi utilizzare un modulo già esistente, scritto da qualcun altro o creato da te stesso durante il lavoro su un progetto complesso: in questo caso, tu sei l'**utente** del modulo;
* **La seconda** si verifica quando vuoi creare un modulo ex novo, sia per uso personale che per facilitare la vita ad altri programmatori: in questo caso, tu sei il **fornitore** del modulo.

Discutiamone separatamente.
Innanzitutto, un modulo è identificato dal suo **nome**. Se vuoi usare un modulo, devi conoscerne il nome. Un numero (piuttosto elevato) di moduli viene fornito insieme a Python stesso. Puoi pensarli come una sorta di "equipaggiamento extra di Python".


Tutti questi moduli, insieme alle funzioni integrate (*built-in*), formano la **libreria standard di Python**: una sorta di biblioteca speciale dove i moduli giocano il ruolo dei libri (possiamo persino dire che le cartelle giocano il ruolo degli scaffali). Se vuoi dare un'occhiata all'elenco completo di tutti i "volumi" raccolti in questa biblioteca, puoi trovarlo qui: [https://docs.python.org/3/library/index.html](https://docs.python.org/3/library/index.html).

Ogni modulo è composto da **entità** (proprio come un libro è composto da capitoli). Queste entità possono essere:

* Funzioni
* Variabili
* Costanti
* Classi
* Oggetti

Se sai come accedere a un particolare modulo, puoi utilizzare qualsiasi entità in esso contenuta.

Iniziamo la discussione con uno dei moduli più utilizzati, chiamato **math**. Il suo nome parla da sé: il modulo contiene una ricca collezione di entità (non solo funzioni) che permettono a un programmatore di implementare efficacemente calcoli che richiedono l'uso di funzioni matematiche, come $\sin()$ o $\log()$.

# Importare un modulo

Per rendere utilizzabile un modulo, devi importarlo (pensa a questo gesto come al prendere un libro dallo scaffale). L'importazione di un modulo avviene tramite un'istruzione chiamata `import`. Nota: `import` è anche una **parola chiave** (con tutte le conseguenze che ne derivano).

Supponiamo che tu voglia utilizzare due entità fornite dal modulo `math`:

* Un simbolo (**costante**) che rappresenta il valore di $\pi$ (preciso quanto lo consente l'aritmetica in virgola mobile a doppia precisione). Sebbene in Python sia tecnicamente possibile usare una lettera greca per nominare una variabile, il simbolo si chiama `pi`: una soluzione più pratica, specialmente per chi non possiede una tastiera greca.
* Una funzione chiamata `sin()` (l'equivalente informatico della funzione trigonometrica seno).

Entrambe queste entità sono disponibili tramite il modulo `math`, ma il modo in cui potrai usarle dipende strettamente da come è stata eseguita l'importazione.


Il modo più semplice per importare un modulo specifico è usare l'istruzione `import` in questo modo:

```python
import math

```

La clausola contiene:

1. La parola chiave **`import`**;
2. Il **nome del modulo** che si intende importare.

L'istruzione può essere collocata in qualsiasi punto del codice, ma deve trovarsi obbligatoriamente **prima del primo utilizzo** di una qualsiasi entità del modulo.

Se desideri (o devi) importare più di un modulo, puoi farlo ripetendo la clausola `import` (metodo preferibile):

```python
import math
import sys

```

Oppure elencando i moduli dopo la parola chiave `import`, come in questo esempio:

```python
import math, sys

```

Questa istruzione importa due moduli: prima quello denominato `math` e successivamente quello denominato `sys`. L'elenco dei moduli può essere arbitrariamente lungo.

# Namespace (Spazio dei nomi)

Per proseguire, è necessario familiarizzare con un termine importante: **namespace**. Non preoccuparti, non entreremo troppo nel dettaglio: questa spiegazione sarà la più breve possibile.

Un namespace è uno spazio (inteso in un contesto non fisico) in cui esistono dei nomi che **non sono in conflitto tra loro** (ovvero, non esistono due oggetti diversi con lo stesso nome). Possiamo dire che ogni gruppo sociale è un namespace: il gruppo tende a nominare ciascuno dei suoi membri in modo univoco (ad esempio, i genitori non danno ai propri figli lo stesso nome di battesimo).

Questa univocità può essere ottenuta in molti modi, ad esempio usando dei soprannomi insieme ai nomi (funziona all'interno di un piccolo gruppo come una classe scolastica) o assegnando identificatori speciali a tutti i membri del gruppo (il codice fiscale è un buon esempio di questa pratica).

## Unicità e conflitti

All'interno di un determinato namespace, **ogni nome deve rimanere unico**. Ciò può significare che alcuni nomi potrebbero "scomparire" quando un'altra entità con un nome già noto entra nel namespace. Ti mostreremo come funziona e come controllarlo, ma prima torniamo alle importazioni.

Se il modulo con il nome specificato esiste ed è accessibile (un modulo è di fatto un file sorgente Python), Python ne importa il contenuto: cioè, tutti i nomi definiti nel modulo diventano noti, ma **non entrano nel namespace del tuo codice**.

Questo significa che puoi avere le tue entità personali chiamate `sin` o `pi` e queste non verranno influenzate in alcun modo dall'importazione.

## Qualificazione dei nomi

A questo punto, ti starai chiedendo come fare per accedere al `pi` che proviene dal modulo `math`.
Per farlo, devi **qualificare** il nome `pi` con il nome del suo modulo di origine.

In termini pratici, se hai usato `import math`, per richiamare la costante non userai semplicemente `pi`, ma dovrai scrivere `math.pi`.

## Esempio

Questo primo esempio non sarà molto avanzato: vogliamo solo stampare il valore di $\sin(\frac{1}{2}\pi)$. Guarda il codice nell'editor. Ecco come lo testiamo:

````python
import math
print(math.sin(math.pi/2))
````

Il codice produce il valore atteso: `1.0`.

Osserva il frammento qui sotto: questo è il modo in cui **qualifichi** i nomi di `pi` e `sin` con il nome del modulo di origine:

```python
math.pi
math.sin
```

È semplice, inserisci:

1. Il **nome del modulo** (es. `math`)
2. Un **punto** (cioè `.`)
3. Il **nome dell'entità** (es. `pi`)

Tale forma indica chiaramente il namespace in cui esiste il nome.
**Nota:** l'uso di questa qualificazione è obbligatorio se un modulo è stato importato con l'istruzione `import nome_modulo`. Non importa se i nomi del tuo codice e quelli del namespace del modulo siano in conflitto o meno.

**Nota:** la rimozione di una qualsiasi delle due qualificazioni renderà il codice errato. Non c'è altro modo per accedere al namespace di `math` se hai utilizzato `import math`.

Ora ti mostreremo come i due namespace (il tuo e quello del modulo) possono coesistere. Guarda l'esempio nell'editor: qui abbiamo definito i nostri `pi` e `sin`.

````python
import math


def sin(x):
    if 2 * x == pi:
        return 0.99999999
    else:
        return None


pi = 3.14

print(sin(pi/2))
print(math.sin(math.pi/2))
````

Eseguendo il programma, il codice dovrebbe produrre il seguente output:

```text
0.99999999
1.0

```

Come puoi vedere, le entità non si influenzano a vicenda.

Nel secondo metodo, la sintassi dell'importazione indica con precisione quale entità (o entità) del modulo sono accettate nel codice:

```python
from math import pi
```

L'istruzione è composta dai seguenti elementi:

* La parola chiave **`from`**;
* Il **nome del modulo** da importare (selettivamente);
* La parola chiave **`import`**;
* Il **nome o l'elenco di nomi** delle entità che vengono importate nel namespace.

Questa istruzione ha i seguenti effetti:

1. Le entità elencate (e solo quelle) vengono importate dal modulo indicato;
2. I nomi delle entità importate sono **accessibili senza qualificazione**.

**Nota:** non vengono importate altre entità. Inoltre, non puoi importare entità aggiuntive usando la qualificazione: una riga come `print(math.e)` causerà un errore (`e` è il numero di Eulero: 2.71828...).

Riscriviamo lo script precedente usando la nuova tecnica.

````python
from math import sin, pi

print(sin(pi/2))
````

L'output sarà lo stesso di prima (`1.0`), poiché abbiamo usato le stesse entità. Ma questa forma di importazione ha effetti importanti sul namespace.

Analizza attentamente questo scenario:

1. **Riga 1:** esegui l'importazione selettiva;
2. **Riga 3:** usi le entità importate e ottieni il risultato atteso (`1.0`);
3. **Righe 5-12:** ridefinisci il significato di `pi` e `sin`; di fatto, questi **sostituiscono** le definizioni originali (importate) all'interno del namespace del codice;
4. **Riga 15:** ottieni `0.99999999`, confermando che le tue definizioni hanno preso il sopravvento.

````python
from math import sin, pi

print(sin(pi / 2))

pi = 3.14


def sin(x):
    if 2 * x == pi:
        return 0.99999999
    else:
        return None


print(sin(pi / 2))
````


Facciamo un altro test invertendo la sequenza:

````python
pi = 3.14


def sin(x):
    if 2 * x == pi:
        return 0.99999999
    else:
        return None


print(sin(pi / 2))

from math import sin, pi

print(sin(pi / 2))
````

1. **Righe 1-8:** definiamo i nostri `pi` e `sin`;
2. **Riga 11:** li usiamo (appare `0.99999999`);
3. **Riga 13:** eseguiamo l'importazione: i simboli importati **sovras scrivono** le loro definizioni precedenti nel namespace;
4. **Riga 15:** otteniamo `1.0` come risultato.

# Importare un modulo: `*`

Nel terzo metodo, la sintassi dell'importazione è una forma più "aggressiva" di quella presentata in precedenza:

```python
from module import *

```

Come puoi vedere, il nome di un'entità (o l'elenco dei nomi delle entità) è sostituito da un singolo **asterisco** (`*`).
Tale istruzione importa **tutte le entità** dal modulo indicato.


* **È comodo?** Sì, lo è, poiché ti libera dall'obbligo di elencare tutti i nomi di cui hai bisogno.
* **È pericoloso?** Sì, lo è: a meno che tu non conosca tutti i nomi forniti dal modulo, potresti non essere in grado di evitare **conflitti tra nomi**.

> Considera questa come una soluzione temporanea e cerca di non usarla nel codice definitivo o professionale.

L'uso indiscriminato dell'asterisco "inquina" il tuo namespace locale con una moltitudine di nomi che potrebbero sovrascrivere involontariamente le tue variabili o funzioni preesistenti.

# La parola chiave `as`

Se utilizzi la variante `import modulo` e il nome di un modulo specifico non ti soddisfa (ad esempio perché è identico a una delle tue entità già definite, rendendo la qualificazione problematica), puoi assegnargli un nome a tua scelta: questo processo è chiamato **aliasing** (creazione di un alias).

L'aliasing fa sì che il modulo venga identificato con un nome diverso da quello originale. Questo può servire anche a scorciare i nomi qualificati.

La creazione di un alias avviene contestualmente all'importazione del modulo e richiede la seguente forma dell'istruzione `import`:

```python
import modulo as alias

```

"Modulo" identifica il nome originale del modulo, mentre "alias" è il nome che desideri utilizzare al suo posto.
**Nota:** `as` è una parola chiave.


# Aliasing del modulo

Se hai la necessità di cambiare la parola `math`, puoi introdurre un nome scelto da te, proprio come in questo esempio:

```python
import math as m
    
print(m.sin(m.pi/2))

```

> Dopo l'esecuzione corretta di un'importazione con alias, il nome originale del modulo diventa **inaccessibile** e non deve più essere utilizzato.


# Aliasing delle entità

Se invece utilizzi la variante `from modulo import nome` e hai bisogno di cambiare il nome di una specifica entità, puoi creare un alias per l'entità stessa. In questo modo, il nome originale verrà sostituito dall'alias scelto.

Ecco come si fa:

```python
from modulo import nome as alias

```

Come nel caso precedente, il nome originale (senza alias) diventa inaccessibile.

La combinazione `nome as alias` può essere ripetuta: usa le virgole per separare le varie espressioni, in questo modo:

```python
from modulo import n as a, m as b, o as c

```

L'esempio seguente potrebbe sembrare un po' insolito, ma funziona perfettamente:

```python
from math import pi as PI, sin as sine
  
print(sine(PI/2))

```

# RIEPILOGO DELLA SEZIONE

1. Se desideri importare un modulo nella sua interezza, puoi farlo utilizzando l'istruzione `import nome_modulo`. È consentito importare più di un modulo contemporaneamente utilizzando un elenco separato da virgole. Per esempio:

```python
import mod1
import mod2, mod3, mod4

```

Tuttavia, quest'ultima forma non è raccomandata per ragioni stilistiche; è preferibile e più elegante esprimere la stessa intenzione in una forma più estesa ed esplicita, come questa:

```python
import mod2
import mod3
import mod4

```

2. Se un modulo viene importato nel modo sopra descritto e desideri accedere a una qualsiasi delle sue entità, devi anteporre al nome dell'entità il nome del modulo usando la **notazione col punto** (*dot notation*). Per esempio:

```python
import my_module
   
result = my_module.my_function(my_module.my_data)

```

Questo frammento di codice utilizza due entità provenienti dal modulo `my_module`: una funzione chiamata `my_function()` e una variabile chiamata `my_data`. Entrambi i nomi devono essere preceduti da `my_module.`. Nessuno dei nomi delle entità importate entra in conflitto con eventuali nomi identici già esistenti nel namespace del tuo codice.

3. È consentito non solo importare un modulo nella sua interezza, ma anche importare **solo singole entità** da esso. In questo caso, le entità importate non devono essere precedute dal nome del modulo quando vengono utilizzate. Per esempio:

```python
from module import my_function, my_data
  
result = my_function(my_data)

```

Questo metodo, nonostante la sua comodità, non è raccomandato a causa del pericolo di generare conflitti con i nomi già presenti nel namespace del codice.

4. La forma più generale dell'istruzione precedente consente di importare **tutte le entità** offerte da un modulo:

```python
from my_module import *
  
result = my_function(my_data)

```

**Nota:** questa variante di importazione non è raccomandata per le stesse ragioni menzionate in precedenza (il rischio di un conflitto di nomi è qui ancora più pericoloso).

5. Puoi cambiare il nome dell'entità importata "al volo" (creare un alias) utilizzando la clausola `as` dell'importazione. Per esempio:

```python
from module import my_function as fun, my_data as dat
  
result = fun(dat)

```
