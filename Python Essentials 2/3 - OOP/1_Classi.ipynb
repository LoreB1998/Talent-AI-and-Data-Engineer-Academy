# Concetti fondamentali OOP

## I concetti base dell'approccio orientato agli oggetti

Facciamo un passo al di fuori della programmazione informatica e dei computer in generale, e discutiamo delle questioni legate alla programmazione a oggetti.

Quasi tutti i programmi e le tecniche che hai utilizzato fino ad ora rientrano nello **stile di programmazione procedurale**. Certo, hai fatto uso di alcuni oggetti integrati, ma quando ci siamo riferiti a loro abbiamo menzionato solo lo stretto indispensabile.

Lo stile procedurale è stato l'approccio dominante nello sviluppo del software per decenni nell'informatica, ed è tuttora in uso. Inoltre, non è destinato a scomparire in futuro, poiché funziona molto bene per specifici tipi di progetti (in genere non molto complessi e non di grandi dimensioni, anche se ci sono molte eccezioni a questa regola).

L'**approccio a oggetti** è piuttosto giovane (molto più giovane di quello procedurale) ed è particolarmente utile quando viene applicato a progetti grandi e complessi, portati avanti da team numerosi composti da molti sviluppatori.

Questo modo di concepire la struttura di un progetto semplifica molti compiti importanti, come ad esempio la divisione del progetto in piccole parti indipendenti e lo sviluppo autonomo dei diversi elementi che lo compongono.

**Python è uno strumento universale** sia per la programmazione a oggetti che per quella procedurale, e può essere utilizzato con successo in entrambi gli ambiti.

Inoltre, puoi creare un'infinità di applicazioni utili anche se non sai nulla di classi e oggetti; tuttavia, devi tenere a mente che alcuni problemi (ad esempio la gestione dell'interfaccia grafica utente, o GUI) potrebbero richiedere un approccio rigorosamente orientato agli oggetti.

Fortunatamente, la programmazione a oggetti è relativamente semplice.
## Approccio procedurale vs. orientato agli oggetti

Nell'approccio procedurale è possibile distinguere due mondi diversi e completamente separati: il mondo dei dati e il mondo del codice. Il mondo dei dati è popolato da variabili di diverso tipo, mentre il mondo del codice è abitato da codice raggruppato in moduli e funzioni.

Le funzioni sono in grado di usare i dati, ma non viceversa. Inoltre, le funzioni possono abusare dei dati, vale a dire utilizzarne il valore in modo non autorizzato (ad esempio, quando la funzione seno riceve come parametro il saldo di un conto bancario).

In passato abbiamo detto che i dati non possono usare le funzioni. Ma è del tutto vero? Esistono tipi speciali di dati in grado di utilizzare le funzioni?

Sì, esistono e si chiamano metodi. Si tratta di funzioni che vengono invocate dall'interno dei dati, non al loro fianco. Se riesci a cogliere questa distinzione, hai fatto il primo passo nella programmazione a oggetti.

L'approccio a oggetti suggerisce un modo di pensare completamente diverso. I dati e il codice sono racchiusi insieme nello stesso mondo, suddiviso in classi.

Ogni classe è come una ricetta che può essere utilizzata quando si desidera creare un oggetto utile (è da qui che deriva il nome dell'approccio). Puoi produrre tutti gli oggetti di cui hai bisogno per risolvere il tuo problema.

Ogni oggetto possiede un insieme di caratteristiche (chiamate proprietà o attributi – useremo entrambi i termini come sinonimi) ed è in grado di eseguire un insieme di attività (chiamate metodi).

Le ricette possono essere modificate se si rivelano inadeguate per scopi specifici e, di conseguenza, possono essere create nuove classi. Queste nuove classi ereditano proprietà e metodi da quelle originali e solitamente ne aggiungono di nuovi, creando strumenti nuovi e più specifici.

Gli oggetti sono incarnazioni di idee espresse nelle classi, proprio come una cheesecake sul tuo piatto è l'incarnazione dell'idea espressa in una ricetta stampata su un vecchio libro di cucina.

Gli oggetti interagiscono tra loro scambiandosi dati o attivando i propri metodi. Una classe opportunamente costruita (e quindi i suoi oggetti) è in grado di proteggere i dati sensibili e di nasconderli da modifiche non autorizzate.

Non esiste un confine netto tra dati e codice: vivono come un'unica entità all'interno degli oggetti.

Tutti questi concetti non sono così astratti come potresti inizialmente sospettare. Al contrario, sono tutti tratti da esperienze della vita reale e per questo sono estremamente utili nella programmazione informatica: non creano una vita artificiale, ma riflettono fatti, relazioni e circostanze reali.

## Gerarchie di classi

La parola *classe* ha molti significati, ma non tutti sono compatibili con le idee che vogliamo discutere qui. La classe che interessa a noi è simile a una **categoria**, derivante da somiglianze definite in modo preciso. Cercheremo di indicare alcune classi che costituiscono ottimi esempi di questo concetto.

Diamo un'occhiata per un momento ai veicoli. Tutti i veicoli esistenti (e quelli che ancora non esistono) sono accomunati da una sola, importante caratteristica: la capacità di muoversi. Si potrebbe obiettare che anche un cane si muove; un cane è quindi un veicolo? No, non lo è. Dobbiamo migliorare la definizione, arricchirla con altri criteri per distinguere i veicoli dagli altri esseri viventi e creare un legame più forte. Prendiamo in considerazione le seguenti circostanze: i veicoli sono entità create artificialmente, utilizzate per il trasporto, mosse da forze della natura e dirette (guidate) dall'uomo.

In base a questa definizione, un cane non è un veicolo.

La classe dei veicoli è molto ampia. Troppo ampia. Dobbiamo quindi definire alcune classi più specializzate. Le classi specializzate sono le **sottoclassi**. La classe dei veicoli sarà una **superclasse** per tutte loro.

> **Nota:** La gerarchia cresce dall'alto verso il basso, come le radici di un albero, non come i rami. La classe più generale e più ampia si trova sempre in alto (la superclasse), mentre i suoi discendenti si trovano al di sotto (le sottoclassi).

A questo punto, puoi probabilmente individuare alcune potenziali sottoclassi per la superclasse *Veicoli*. Esistono molte classificazioni possibili. Noi abbiamo scelto le sottoclassi in base all'ambiente, e possiamo dire che ci sono (almeno) quattro sottoclassi:

* veicoli terrestri;
* veicoli acquatici;
* veicoli aerei;
* veicoli spaziali.

In questo esempio discuteremo solo la prima sottoclasse: i veicoli terrestri. Se desideri, puoi continuare con le restanti classi.

I veicoli terrestri possono essere ulteriormente suddivisi a seconda del modo in cui impattano sul terreno. Quindi, possiamo elencare:

* veicoli su ruote;
* veicoli cingolati;
* hovercraft.

Nota la direzione delle frecce: puntano sempre verso la superclasse. La classe di livello superiore rappresenta un'eccezione, in quanto non ha una propria superclasse.

* mammiferi;
* rettili;

## Cos'è un oggetto?

Una classe (tra le varie definizioni) è un insieme di oggetti. Un **oggetto** è un'entità che appartiene a una classe.

Un oggetto è l'incarnazione dei requisiti, dei tratti e delle qualità assegnati a una specifica classe. Può sembrare semplice, ma occorre notare le seguenti dinamiche importanti legate alla gerarchia delle classi:

* Un oggetto che appartiene a una classe specifica **appartiene contemporaneamente a tutte le sue superclassi**.
* Al contrario, un oggetto che appartiene a una superclasse **potrebbe non appartenere** a nessuna delle sue sottoclassi.

**Esempi:**

* Qualsiasi automobile privata è un oggetto che appartiene alla classe dei *veicoli su ruote*. Ciò significa che la stessa auto appartiene anche a tutte le superclassi della sua classe di origine; di conseguenza, è anche un membro della classe *veicoli*.
* Il tuo cane (o il tuo gatto) è un oggetto incluso nella classe dei *mammiferi domestici*, il che significa esplicitamente che è incluso anche nella classe *animali*.

Ogni sottoclasse è più **specializzata** (o più concreta) rispetto alla sua superclasse. Al contrario, ogni superclasse è più **generale** (piò astratta) di qualsiasi sua sottoclasse.

*(Nota: in questa analisi abbiamo ipotizzato che una classe possa avere una sola superclasse; questo non è sempre vero, ma ne discuteremo più avanti).*

## Ereditarietà

Definiamo uno dei concetti fondamentali della programmazione a oggetti: l'**ereditarietà**.

> Qualsiasi oggetto legato a uno specifico livello di una gerarchia di classi **eredita tutti i tratti** (così come i requisiti e le qualità) definiti all'interno di una qualsiasi delle sue superclassi.

La classe di origine dell'oggetto può definire nuovi tratti che saranno a loro volta ereditati dalle sue sottoclassi. Non dovresti avere problemi a riscontrare questa regola negli esempi specifici, sia che si applichi agli animali sia ai veicoli.

## Cosa possiede un oggetto?

La convenzione della programmazione a oggetti presuppone che ogni oggetto esistente possa essere dotato di tre gruppi di attributi:

1. **Un nome** che lo identifica in modo univoco all'interno del proprio spazio dei nomi (namespace), sebbene possano esistere anche oggetti anonimi.
2. **Un insieme di proprietà individuali** che lo rendono originale, unico o differente dagli altri (anche se è possibile che alcuni oggetti non abbiano alcuna proprietà).
3. **Un insieme di capacità** di eseguire attività specifiche, in grado di modificare l'oggetto stesso o altri oggetti.

Esiste un piccolo suggerimento linguistico (anche se non funziona sempre) che può aiutarti a identificare queste tre sfere. Quando descrivi un oggetto e usi:

* Un **sostantivo**: probabilmente stai definendo il **nome** dell'oggetto.
* Un **aggettivo**: probabilmente stai definendo una **proprietà** dell'oggetto.
* Un **verbo**: probabilmente stai definendo un'**attività** (metodo) dell'oggetto.

### Esempi pratici:

* *"Una Cadillac rosa andava velocemente."*
* **Nome dell'oggetto:** Cadillac
* **Classe di origine:** Veicoli su ruote
* **Proprietà:** Colore (rosa)
* **Attività:** Andare (velocemente)


* *"Rudolph è un gatto grande che dorme tutto il giorno."*
* **Nome dell'oggetto:** Rudolph
* **Classe di origine:** Gatto
* **Proprietà:** Taglia (grande)
* **Attività:** Dormire (tutto il giorno)

## La tua prima classe

La programmazione a oggetti è l'arte di definire ed espandere le classi. Una classe è il modello di una parte molto specifica della realtà, che riflette proprietà e attività riscontrabili nel mondo reale.

Le classi definite all'inizio sono spesso troppo generali e imprecise per coprire il maggior numero possibile di casi reali. Nulla vieta però di definire nuove sottoclassi più precise: esse ereditano tutto dalla loro superclasse, quindi il lavoro svolto per crearla non va perduto.

La nuova classe può aggiungere nuove proprietà e nuove attività, diventando così più utile per applicazioni specifiche. Ovviamente, può essere utilizzata a sua volta come superclasse per un numero qualsiasi di sottoclassi appena create. Questo processo non ha una fine: puoi creare tutte le classi di cui hai bisogno.

**Attenzione:** La classe definita non ha nulla a che fare con l'oggetto in sé: l'esistenza di una classe non significa che uno degli oggetti compatibili venga creato automaticamente. La classe da sola non è in grado di creare un oggetto: devi crearlo tu stesso, e Python ti permette di farlo.

È il momento di definire la classe più semplice possibile e di creare un oggetto. Dai un'occhiata all'esempio seguente:

```python
class TheSimplestClass:
    pass

```

Abbiamo definito una classe. È una classe piuttosto povera: non ha né proprietà né attività. È vuota, in realtà, ma per ora non importa. Più la classe è semplice, meglio è per i nostri scopi.

* La definizione inizia con la parola chiave `class`.
* La parola chiave è seguita da un **identificatore** che darà il nome alla classe (nota: non confonderlo con il nome dell'oggetto, sono due cose diverse).
* Successivamente si aggiungono i **due punti** (`:`), poiché le classi, come le funzioni, formano un proprio blocco nidificato. Il contenuto all'interno del blocco definisce tutte le proprietà e le attività della classe.
* La parola chiave `pass` riempie la classe con il nulla. Non contiene metodi o proprietà.


## Il tuo primo oggetto

La classe appena definita diventa uno strumento in grado di creare nuovi oggetti. Questo strumento deve essere utilizzato in modo esplicito, su richiesta.

Immagina di voler creare un oggetto (esattamente uno) della classe `TheSimplestClass`. Per fare questo, devi assegnare una variabile che memorizzi l'oggetto appena creato di quella classe e, contemporaneamente, creare l'oggetto stesso.

Lo si fa nel seguente modo:

```python
my_first_object = TheSimplestClass()

```

**Nota:**

* Il nome della classe cerca di "fingere" di essere una funzione – riesci a vederlo? Ne discuteremo presto.
* L'oggetto appena creato è dotato di tutto ciò che la classe porta con sé; poiché questa classe è completamente vuota, anche l'oggetto è vuoto.

L'atto di creare un oggetto della classe selezionata viene anche chiamato **istanziazione** (poiché l'oggetto diventa un'**istanza** della classe).

Lasciamo da parte le classi per un breve momento, poiché ora ti diremo due parole sugli stack (pile). Sappiamo che il concetto di classi e oggetti potrebbe non essere ancora del tutto chiaro. Non preoccuparti, spiegheremo tutto molto presto.

# Un breve viaggio dall'approccio procedurale a quello a oggetti

## Cos'è uno stack?

Uno **stack** (o pila) è una struttura sviluppata per memorizzare i dati in un modo molto specifico. Immagina una pila di monete. Non sei in grado di mettere una moneta in nessun altro posto se non in cima alla pila.

Allo stesso modo, non puoi togliere una moneta dalla pila da nessun altro punto se non dalla cima. Se vuoi prendere la moneta che si trova sul fondo, devi rimuovere tutte le monete dei livelli superiori.

Il nome alternativo per uno stack (ma solo nella terminologia informatica) è **LIFO**. È l'acronimo di una descrizione molto chiara del comportamento dello stack: **Last In – First Out** (L'ultimo a entrare è il primo a uscire). La moneta che è arrivata per ultima sulla pila sarà la prima ad andarsene.

Uno stack è un oggetto dotato di due operazioni elementari, convenzionalmente chiamate:

* **push** (quando un nuovo elemento viene messo in cima);
* **pop** (quando un elemento esistente viene rimosso dalla cima).

Gli stack sono usati molto spesso in molti algoritmi classici, ed è difficile immaginare l'implementazione di molti strumenti ampiamente utilizzati senza il loro impiego.

Proviamo a implementare uno stack in Python. Sarà uno stack molto semplice e ti mostreremo come realizzarlo attraverso due approcci indipendenti: procedurale e a oggetti. Iniziamo con il primo.


## Lo stack: l'approccio procedurale

Per prima cosa, devi decidere come memorizzare i valori che arriveranno sullo stack. Ti suggeriamo di usare il metodo più semplice, impiegando una **lista** per questo compito. Assumiamo che la dimensione dello stack non sia limitata in alcun modo. Assumiamo anche che l'ultimo elemento della lista rappresenti l'elemento in cima.

Lo stack vero e proprio è già creato:

```python
stack = []

```

Ora siamo pronti a definire una funzione che inserisce un valore nello stack. Ecco i presupposti:

* il nome della funzione è `push`;
* la funzione riceve un parametro (il valore da inserire nello stack);
* la funzione non restituisce nulla;
* la funzione aggiunge il valore del parametro alla fine dello stack.

Ecco come l'abbiamo realizzata:

```python
def push(val):
    stack.append(val)

```

Ora è il turno della funzione per togliere un valore dallo stack. Ecco come puoi fare:

* il nome della funzione è `pop`;
* la funzione non riceve alcun parametro;
* la funzione restituisce il valore rimosso dallo stack;
* la funzione legge il valore in cima allo stack e lo rimuove.

Ecco la funzione:

```python
def pop():
    val = stack[-1]
    del stack[-1]
    return val

```

*Nota: la funzione non controlla se ci sono elementi nello stack.*

Mettiamo insieme tutti i pezzi per avviare lo stack. Il programma completo inserisce tre numeri nello stack, li tira fuori e stampa i loro valori sullo schermo. Puoi vederlo nell'editor.



````python
stack = []


def push(val):
    stack.append(val)


def pop():
    val = stack[-1]
    del stack[-1]
    return val


push(3)
push(2)
push(1)

print(pop())
print(pop())
print(pop())
    
````

## Lo stack: approccio procedurale vs. approccio orientato agli oggetti

Lo stack procedurale è pronto. Naturalmente presenta alcune debolezze e l'implementazione potrebbe essere migliorata in molti modi (ad esempio, sfruttando la gestione delle eccezioni), ma in generale lo stack è completamente implementato e pronto all'uso in caso di necessità.

Tuttavia, più lo userai, più andrai incontro ad alcuni svantaggi. Eccone alcuni:

* **L'estrema vulnerabilità della variabile essenziale (la lista `stack`):** chiunque può modificarla in modo incontrollato, distruggendo di fatto lo stack. Questo non significa necessariamente che venga fatto con cattive intenzioni; al contrario, può accadere per disattenzione, ad esempio confondendo i nomi delle variabili. Immagina di scrivere accidentalmente qualcosa del genere:
```python
stack[0] = 0

```


Il funzionamento dello stack verrebbe completamente compromesso.
* **La necessità di avere più di uno stack:** se un giorno avessi bisogno di due o tre stack, dovresti creare un'altra lista per memorizzare i dati di ogni nuovo stack e, molto probabilmente, anche altre funzioni `push` e `pop` dedicate.
* **Il bisogno di nuove funzionalità:** se oltre a `push` e `pop` ti servissero altre comodità, dovresti implementarle da zero per ogni singolo stack. Prova a immaginare cosa succederebbe se ti trovassi a gestire dozzine di stack implementati separatamente.

### La soluzione dell'approccio a oggetti

L'approccio orientato agli oggetti (OOP) offre una soluzione mirata per ciascuno di questi problemi attraverso tre pilastri fondamentali:

1. **Incapsulamento (Encapsulation):** È la capacità di nascondere (proteggere) determinati valori dall'accesso non autorizzato. I valori incapsulati non possono essere letti né modificati direttamente dall'esterno, garantendo l'integrità della struttura dati.
2. **Riutilizzabilità del codice:** Quando definisci una classe che implementa tutti i comportamenti necessari allo stack, puoi produrre tutti gli stack che desideri (*istanze* della classe). Non avrai bisogno di copiare o replicare alcuna parte del codice.
3. **Ereditarietà (Inheritance):** È la capacità di arricchire lo stack con nuove funzioni. Puoi creare una nuova classe (una sottoclasse) che eredita tutte le caratteristiche esistenti dalla superclasse e ne aggiunge di nuove, senza dover riscrivere ciò che è già stato fatto.

Scriviamo ora una nuova implementazione dello stack partendo da zero. Questa volta utilizzeremo l'approccio a oggetti, guidandoti passo dopo passo nel mondo della programmazione orientata agli oggetti.

## Lo stack: l'approccio a oggetti

Naturalmente l'idea di fondo rimane la stessa: useremo una lista per memorizzare i dati dello stack. Dobbiamo solo capire come inserire questa lista all'interno della classe.

Partiamo dall'inizio assoluto: ecco come comincia il nostro stack a oggetti:

```python
class Stack:
    pass

```

Ora, ci aspettiamo due cose da questa classe:

1. Vogliamo che abbia una proprietà che funga da contenitore per lo stack: dobbiamo "installare" una lista all'interno di **ogni oggetto** della classe (nota: ogni oggetto deve avere la propria lista, la quale non deve essere condivisa con altri stack).
2. Vogliamo che questa lista sia **nascosta** agli occhi degli utenti della classe (incapsulamento).

Come si fa?
A differenza di altri linguaggi di programmazione, Python non dispone di costrutti che permettono di dichiarare una simile proprietà in modo diretto nel corpo della classe. Le proprietà devono essere aggiunte all'oggetto manualmente tramite un'apposita istruzione.

Come si può garantire che questa azione avvenga **ogni volta** che viene creato un nuovo stack?
Esiste un modo semplice: bisogna dotare la classe di una funzione speciale, la cui particolarità è duplice:

* Deve avere un nome ben preciso e rigido.
* Viene invocata implicitamente (in automatico) quando viene creato un nuovo oggetto.

Una funzione di questo tipo si chiama **costruttore**, poiché il suo scopo generale è proprio quello di costruire un nuovo oggetto. Il costruttore deve conoscere l'intera struttura dell'oggetto e deve eseguire tutte le inizializzazioni necessarie.

Aggiungiamo un costruttore molto semplice alla nostra nuova classe. Dai un'occhiata a questo frammento di codice:

```python
class Stack:
    def __init__(self):
        print("Hi!")

stack_object = Stack()

```

Analizziamo i dettagli importanti:

* Il nome del costruttore è sempre **`__init__`** (con due trattini bassi prima e dopo).
* Deve avere **almeno un parametro** (ne discuteremo meglio più avanti). Questo parametro serve a rappresentare l'oggetto appena creato; puoi usarlo per manipolare l'oggetto stesso e arricchirlo con le proprietà necessarie. Lo userai molto presto.
* **Nota:** per convenzione, questo parametro obbligatorio viene chiamato **`self`**. È solo una regola comunemente accettata, ma ti conviene seguirla scrupolosamente: semplifica enormi vantaggi nella lettura e nella comprensione del codice da parte di altri sviluppatori.

````python
class Stack:  # Defining the Stack class.
    def __init__(self):  # Defining the constructor function.
        print("Hi!")


stack_object = Stack()  # Instantiating the object.
    
````

Nota come nel codice non ci sia alcuna traccia di un'invocazione esplicita del costruttore. È stato invocato in modo implicito e automatico. Sfruttiamo subito questa caratteristica.

Qualsiasi modifica apportata all'interno del costruttore che alteri lo stato del parametro `self` si rifletterà sull'oggetto appena creato. Questo significa che puoi aggiungere qualsiasi proprietà all'oggetto, e tale proprietà rimarrà lì finché l'oggetto non esaurirà il suo ciclo di vita o finché non verrà rimossa esplicitamente.

Ora aggiungiamo una sola proprietà al nuovo oggetto: una lista che funga da contenitore per lo stack. La chiameremo `stack_list`.

Ecco come si fa:

```python
class Stack:
    def __init__(self):
        self.stack_list = []

stack_object = Stack()

```

Analizziamo quello che abbiamo appena fatto:

* Abbiamo usato la **notazione con il punto** (`.`), proprio come quando si invocano i metodi. Questo è il modo standard in Python per accedere alle proprietà o ai metodi di un oggetto.
* Scrivendo `self.stack_list`, stiamo dicendo a Python: *"Prendi l'oggetto corrente (`self`) e installa al suo interno una proprietà chiamata `stack_list`"*.
* Abbiamo inizializzato questa proprietà assegnandole una lista vuota `[]`.

Ora, ogni volta che creerai un nuovo oggetto della classe `Stack` tramite l'istruzione `Stack()`, Python eseguirà automaticamente il costruttore, creando uno stack indipendente dotato della propria lista privata e vuota.

````python
class Stack:
    def __init__(self):
        self.stack_list = []


stack_object = Stack()
print(len(stack_object.stack_list))
    
````

Nota:

* Abbiamo usato la **notazione con il punto**, proprio come quando si invocano i metodi. Questa è la convenzione generale per accedere alle proprietà di un oggetto: devi nominare l'oggetto, mettere un punto (`.`) dopo di esso e specificare il nome della proprietà desiderata. Non usare le parentesi! Non stai invocando un metodo, vuoi solo accedere a una proprietà.
* Se imposti il valore di una proprietà per la primissima volta (come all'interno del costruttore), la stai creando; da quel momento in poi, l'oggetto possiede quella proprietà ed è pronto a usarne il valore.
* Nel codice abbiamo fatto qualcosa in più: abbiamo cercato di accedere alla proprietà `stack_list` dall'esterno della classe subito dopo la creazione dell'oggetto; volevamo controllare la lunghezza corrente dello stack. Ci siamo riusciti?

Sì, ci siamo riusciti, ma **questo non è ciò che vogliamo** dal nostro stack. Preferiamo che `stack_list` sia nascosta al mondo esterno. È possibile?

Sì, ed è semplice, anche se non molto intuitivo.

Dai un'occhiata: abbiamo aggiunto due trattini bassi (underscore) prima del nome `stack_list`, niente di più:

```python
class Stack:
    def __init__(self):
        self.__stack_list = []

stack_object = Stack()

```

Quando Python vede due trattini bassi all'inizio del nome di una proprietà o di un metodo (ad esempio, `__stack_list`), attiva un meccanismo speciale chiamato **name mangling** (storpiatura del nome).

In questo modo, la proprietà diventa privata e non è più accessibile direttamente dall'esterno della classe digitando `stack_object.__stack_list`. Se provi a farlo, Python genererà un errore di tipo `AttributeError`. La proprietà è ora protetta e nascosta all'interno dell'oggetto.

## L'approccio a oggetti: uno stack da zero

Ora è il momento di implementare le due funzioni (metodi) per le operazioni di `push` e `pop`. Python presuppone che una funzione di questo tipo (un'attività della classe) debba essere immersa all'interno del corpo della classe, esattamente come il costruttore.

Vogliamo poter invocare queste funzioni per inserire ed estrarre i valori. Ciò significa che devono essere entrambe **accessibili a chiunque utilizzi la classe** (al contrario della lista creata in precedenza, che è nascosta ai normali utenti della classe).

Un componente di questo tipo viene definito **pubblico**, quindi il suo nome non può iniziare con due (o più) trattini bassi. C'è anche un altro requisito: il nome non deve avere più di un trattino basso alla fine. Poiché non avere alcun trattino basso alla fine soddisfa pienamente il requisito, possiamo considerare accettabili i nomi standard.

Le funzioni in sé sono semplici. Dai un'occhiata:

```python
class Stack:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        val = self.__stack_list[-1]
        del self.__stack_list[-1]
        return val

```

Analizziamo alcuni dettagli importanti di questa implementazione:

* **Il parametro `self` è obbligatorio:** Entrambi i metodi contengono il parametro `self`. Questo parametro è fondamentale perché permette al metodo di sapere *quale specifico oggetto* lo sta invocando, consentendogli di accedere alla rispettiva lista privata (`self.__stack_list`).
* **Accesso ai dati privati:** Poiché i metodi `push` e `pop` si trovano *all'interno* del corpo della classe, hanno il pieno permesso di accedere e manipolare la proprietà privata `self.__stack_list`, cosa che invece è vietata al codice esterno alla classe.

````python
class Stack:
    def __init__(self):
        self.__stack_list = []


    def push(self, val):
        self.__stack_list.append(val)


    def pop(self):
        val = self.__stack_list[-1]
        del self.__stack_list[-1]
        return val


stack_object = Stack()

stack_object.push(3)
stack_object.push(2)
stack_object.push(1)

print(stack_object.pop())
print(stack_object.pop())
print(stack_object.pop())
````

Tuttavia, c'è qualcosa di veramente strano nel codice. Le funzioni sembrano familiari, ma hanno più parametri rispetto alle loro controparti procedurali.

In questo caso, entrambe le funzioni hanno un parametro chiamato `self` nella prima posizione dell'elenco dei parametri.

È davvero necessario? Sì, lo è.

**Tutti i metodi devono avere questo parametro.** Svolge lo stesso identico ruolo del primo parametro del costruttore: permette al metodo di accedere alle entità (proprietà e attività/metodi) gestite dallo specifico oggetto in uso. Non puoi ometterlo. Ogni volta che Python invoca un metodo, invia implicitamente l'oggetto corrente come primo argomento.

Ciò significa che un metodo è obbligato ad avere almeno un parametro, che viene utilizzato da Python stesso; tu non hai alcuna influenza su questo meccanismo. Se il tuo metodo non richiede alcun parametro, questo deve comunque essere specificato. Se è progettato per elaborare un solo parametro, dovrai specificarne due, e il ruolo del primo rimarrà sempre lo stesso.

C'è ancora una cosa che richiede una spiegazione: il modo in cui i metodi vengono invocati dall'interno della variabile `__stack_list`. Fortunatamente, è molto più semplice di quanto sembri:

1. Il primo passaggio fornisce l'oggetto nel suo complesso $\rightarrow$ `self`;
2. Successivamente, devi raggiungere la lista `__stack_list` $\rightarrow$ `self.__stack_list`;
3. Con la `__stack_list` pronta per essere utilizzata, puoi eseguire il terzo e ultimo passaggio $\rightarrow$ `self.__stack_list.append(val)`.

La dichiarazione della classe è completa e tutti i suoi componenti sono stati elencati. La classe è pronta per l'uso.


Disporre di una classe simile apre nuove possibilità. Ad esempio, ora puoi avere **più di uno stack** che si comporta nello stesso modo. Ogni stack avrà la propria copia di dati privati, ma utilizzerà lo stesso insieme di metodi. Questo è esattamente ciò che vogliamo per questo esempio.

Analizza il codice completo qui sotto:

````python
class Stack:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        val = self.__stack_list[-1]
        del self.__stack_list[-1]
        return val


stack_object_1 = Stack()
stack_object_2 = Stack()

stack_object_1.push(3)
stack_object_2.push(stack_object_1.pop())

print(stack_object_2.pop())
````

Ora facciamo un piccolo passo in avanti. Aggiungiamo una nuova classe per la gestione degli stack.

La nuova classe deve essere in grado di **valutare la somma di tutti gli elementi** attualmente memorizzati nello stack. Tuttavia, non vogliamo modificare lo stack definito in precedenza. Quello va già bene per le sue applicazioni e non vogliamo alterarlo in alcun modo. Vogliamo uno stack nuovo, con nuove capacità. In altre parole, vogliamo costruire una **sottoclasse** della classe `Stack` già esistente.

Il primo passo è semplice: basta definire una nuova sottoclasse indicando la classe che verrà utilizzata come superclasse.

Il codice si presenta così:

```python
class AddingStack(Stack):

```

La classe non definisce ancora nessun nuovo componente, ma questo non significa che sia vuota. **Eredita tutti i componenti** definiti dalla sua superclasse; il nome della superclasse viene scritto tra parentesi prima dei due punti, subito dopo il nome della nuova classe.

Ecco cosa vogliamo dal nuovo stack:

1. Vogliamo che il metodo `push` non si limiti a inserire il valore nello stack, ma che lo **sommi** anche a una variabile `sum`.
2. Vogliamo che la funzione `pop` non si limiti a estrarre il valore dallo stack, ma che lo **sottragga** anche dalla variabile `sum`.

Per prima cosa, aggiungiamo una nuova variabile alla classe. Sarà una variabile privata, proprio come la lista dello stack. Non vogliamo che nessuno manipoli il valore della somma dall'esterno.

Come già sai, l'aggiunta di una nuova proprietà a una classe viene eseguita dal costruttore. Sai già come fare, ma c'è un dettaglio davvero intrigante all'interno di questo costruttore. Dai un'occhiata:

```python
class AddingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__sum = 0

```

La seconda riga del corpo del costruttore crea una proprietà chiamata `__sum`, che memorizzerà il totale di tutti i valori dello stack.

La riga precedente, invece, ha un aspetto diverso. Cosa fa? È davvero necessaria? Sì, lo è.

A differenza di molti altri linguaggi, **Python ti obbliga a invocare esplicitamente il costruttore della superclasse**. Omettere questo passaggio avrà effetti dannosi: l'oggetto verrà privato della lista `__stack_list`. Di conseguenza, uno stack del genere non funzionerà correttamente.

Questa è l'unica circostanza in cui puoi invocare esplicitamente uno dei costruttori disponibili: lo si fa all'interno del costruttore della sottoclasse.

Nota la sintassi:

* Si specifica il **nome della superclasse** (la classe di cui si vuole eseguire il costruttore);
* Si inserisce un **punto** (`.`);
* Si specifica il **nome del costruttore** (`__init__`);
* Bisogna **indicare l'oggetto** (l'istanza della classe) che deve essere inizializzato dal costruttore: ecco perché è necessario specificare l'argomento e usare qui la variabile `self`.

> **Nota di sintassi:** Invocare un metodo (compresi i costruttori) dall'esterno della classe non richiede mai di inserire l'argomento `self` nella lista degli argomenti. Al contrario, invocare un metodo della superclasse dall'interno di una sottoclasse richiede l'uso esplicito dell'argomento `self`, che deve essere posizionato come primo della lista.

*Nota generale:* È una pratica raccomandata invocare il costruttore della superclasse **prima** di qualsiasi altra inizializzazione che si desidera eseguire all'interno della sottoclasse. Questa è la regola che abbiamo seguito nel frammento di codice.

````python
class Stack:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        val = self.__stack_list[-1]
        del self.__stack_list[-1]
        return val


class AddingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__sum = 0
    
````

In secondo luogo, aggiungiamo due metodi. Ma sorge spontanea una domanda: si tratta davvero di un'aggiunta? Abbiamo già questi metodi nella superclasse. È possibile fare una cosa del genere?

Sì, è possibile. Significa che andremo a **modificare la funzionalità** dei metodi, non i loro nomi. Più precisamente, possiamo dire che l'interfaccia della classe (il modo in cui gli oggetti vengono gestiti) rimane la stessa, mentre l'implementazione interna cambia.

Iniziamo con l'implementazione della funzione `push`. Ecco cosa ci aspettiamo che faccia:

1. Aggiungere il valore alla variabile `__sum`;
2. Inserire il valore nello stack.

**Nota:** La seconda attività è già stata implementata all'interno della superclasse, quindi possiamo (anzi, dobbiamo) riutilizzarla. Inoltre, siamo obbligati a farlo, poiché non abbiamo altro modo per accedere alla variabile privata `__stack_list` della superclasse.

Ecco come si presenta il metodo `push` nella sottoclasse:

```python
def push(self, val):
```

Nota il modo in cui abbiamo invocato la precedente implementazione del metodo `push` (quella disponibile nella superclasse):

* Dobbiamo **specificare il nome della superclasse**; questo è necessario per indicare chiaramente quale classe contiene il metodo ed evitare di confonderlo con qualsiasi altra funzione avente lo stesso nome.
* Dobbiamo **specificare l'oggetto di destinazione** e passarlo come primo argomento (in questo contesto non viene aggiunto implicitamente all'invocazione).

In informatica si dice che il metodo `push` è stato **sovrascritto** (*overridden*): lo stesso nome presente nella superclasse ora rappresenta una funzionalità differente.

Dobbiamo definire un nuovo metodo. Lo chiameremo `get_sum`. Il suo unico compito sarà quello di restituire il valore memorizzato in `__sum`.

# OOP: Proprietà

## Variabili di istanza

In generale, una classe può essere dotata di due diversi tipi di dati per formare le sue proprietà. Hai già visto uno di questi tipi quando abbiamo analizzato gli stack.

Questo tipo di proprietà esiste quando e solo quando viene creata ed aggiunta esplicitamente a un oggetto. Come già sai, questo può essere fatto durante l'inizializzazione dell'oggetto, eseguita dal costruttore. Inoltre, può essere fatto in qualsiasi momento della vita dell'oggetto ed è persino possibile rimuovere qualsiasi proprietà esistente in ogni istante.

Questo approccio comporta alcune conseguenze importanti:

* **Oggetti diversi della stessa classe possono possedere insiemi di proprietà diversi.**
* Deve esserci un modo per verificare in sicurezza se uno specifico oggetto possiede la proprietà che si desidera utilizzare (a meno che non si voglia provocare un'eccezione, un'opzione che vale sempre la pena considerare).
* Ogni oggetto porta con sé il proprio insieme di proprietà, le quali non interferiscono in alcun modo tra loro.

Tali variabili (o proprietà) vengono chiamate **variabili di istanza**. La parola *istanza* suggerisce che sono strettamente collegate agli oggetti (che sono istanze della classe) e non alle classi stesse. Diamo un'occhiata più da vicino.

Ecco un esempio:

```python
class ExampleClass:
    def __init__(self, val = 1):
        self.first = val
 
    def set_second(self, val):
        self.second = val
 
 
example_object_1 = ExampleClass()
example_object_2 = ExampleClass(2)
 
example_object_2.set_second(3)
 
example_object_3 = ExampleClass(4)
example_object_3.third = 5
 
print(example_object_1.__dict__)
print(example_object_2.__dict__)
print(example_object_3.__dict__)

```

Prima di entrare nei dettagli, è necessaria una spiegazione aggiuntiva riguardo alle ultime tre righe di codice.

Quando vengono creati, gli oggetti Python sono dotati di un piccolo insieme di proprietà e metodi predefiniti. Ogni oggetto li possiede, che tu lo voglia o meno. Uno di questi è una variabile chiamata **`__dict__`** (che è un dizionario). Questa variabile contiene i nomi e i valori di tutte le proprietà (variabili) che l'oggetto ospita in quel preciso momento. Usiamola per presentare in modo sicuro il contenuto di un oggetto.

Esaminiamo ora il comportamento del codice:

* La classe chiamata `ExampleClass` ha un costruttore che crea incondizionatamente una variabile di istanza chiamata `first` e le assegna il valore passato attraverso il primo argomento (dal punto di vista dell'utente della classe) o il secondo argomento (dal punto di vista del costruttore). Nota il valore di default del parametro: qualsiasi trucco utilizzabile con i parametri di una normale funzione può essere applicato anche ai metodi.
* La classe ha anche un metodo che crea un'altra variabile di istanza, chiamata `second`.
* Abbiamo creato tre oggetti della classe `ExampleClass`, ma tutte queste istanze differiscono tra loro:
* `example_object_1` ha solo la proprietà chiamata `first`.
* `example_object_2` ha due proprietà: `first` e `second`.
* `example_object_3` è stato arricchito con una proprietà chiamata `third` direttamente al volo, all'esterno del codice della classe: questo è possibile e del tutto consentito.



L'output del programma mostra chiaramente che le nostre ipotesi sono corrette:

**Output:**

```text
{'first': 1}
{'first': 2, 'second': 3}
{'first': 4, 'third': 5}

```

C'è un'ulteriore conclusione che va sottolineata: la modifica di una variabile di istanza di un oggetto non ha alcun impatto su tutti i restanti oggetti. Le variabili di istanza sono perfettamente isolate l'una dall'altra.

````python
class ExampleClass:
    def __init__(self, val = 1):
        self.__first = val

    def set_second(self, val = 2):
        self.__second = val


example_object_1 = ExampleClass()
example_object_2 = ExampleClass(2)

example_object_2.set_second(3)

example_object_3 = ExampleClass(4)
example_object_3.__third = 5


print(example_object_1.__dict__)
print(example_object_2.__dict__)
print(example_object_3.__dict__)
    
````

È quasi identico al precedente. L'unica differenza risiede nei nomi delle proprietà, in quanto abbiamo aggiunto due trattini bassi (`__`) davanti a essi.

Come sai, questa aggiunta rende privata la variabile di istanza, rendendola inaccessibile dal mondo esterno.

Il comportamento effettivo di questi nomi è però un po' più complicato, quindi eseguiamo il programma. Questo è l'output:

**Output:**

```text
{'_ExampleClass__first': 1}
{'_ExampleClass__first': 2, '_ExampleClass__second': 3}
{'_ExampleClass__first': 4, '__third': 5}

```

Riesci a vedere questi strani nomi pieni di trattini bassi? Da dove sono saltati fuori?

Quando Python nota che vuoi aggiungere una variabile di istanza a un oggetto e che lo stai facendo *all'interno* di uno dei metodi dell'oggetto stesso, "storpia" (**mangles**) l'operazione nel seguente modo:

* Inserisce il nome della classe prima del tuo nome;
* Aggiunge un ulteriore trattino basso all'inizio.

Ecco perché `__first` diventa `_ExampleClass__first`.

A questo punto, il nome diventa pienamente accessibile dall'esterno della classe. Puoi eseguire un codice come questo:

```python
print(example_object_1._ExampleClass__first)

```

E otterrai un risultato valido senza errori o eccezioni.

Come puoi notare, rendere privata una proprietà ha i suoi limiti. La storpiatura del nome (*mangling*) non funziona se aggiungi una variabile di istanza privata all'esterno del codice della classe. In questo caso specifico (come si vede per `__third` nell'output), si comporterà come una qualsiasi altra proprietà ordinaria.

## Variabili di classe

Una **variabile di classe** è una proprietà che esiste in un'unica copia e viene memorizzata al di fuori di qualsiasi oggetto.

> **Nota:** una variabile di istanza non esiste se non c'è un oggetto della classe; al contrario, una variabile di classe esiste in una singola copia **anche se non ci sono oggetti** all'interno della classe.

Le variabili di classe vengono create in modo diverso rispetto alle loro controparti di istanza. L'esempio seguente chiarirà le idee:

```python
class ExampleClass:
    counter = 0
    def __init__(self, val = 1):
        self.__first = val
        ExampleClass.counter += 1
 
 
example_object_1 = ExampleClass()
example_object_2 = ExampleClass(2)
example_object_3 = ExampleClass(4)
 
print(example_object_1.__dict__, example_object_1.counter)
print(example_object_2.__dict__, example_object_2.counter)
print(example_object_3.__dict__, example_object_3.counter)

```

Osserva bene:

* C'è un'assegnazione nella prima riga della definizione della classe: imposta a `0` la variabile chiamata `counter`. Inizializzare la variabile all'interno della classe, ma **al di fuori di qualsiasi metodo**, la rende a tutti gli effetti una variabile di classe.
* L'accesso a tale variabile può apparire identico a quello di un attributo di istanza. Puoi vederlo all'interno del corpo del costruttore: come noti, il costruttore incrementa la variabile di uno. Di fatto, questa variabile conta tutti gli oggetti che vengono via via creati.

L'esecuzione del codice genererà il seguente output:

**Output:**

```text
{'_ExampleClass__first': 1} 3
{'_ExampleClass__first': 2} 3
{'_ExampleClass__first': 4} 3

```

Da questo esempio derivano due conclusioni fondamentali:

1. Le variabili di classe **non vengono mostrate** nel dizionario `__dict__` dell'oggetto (questo è naturale, poiché le variabili di classe non fanno parte dell'oggetto), ma puoi sempre accedere alla variabile dallo stesso nome a livello di classe (vedremo presto come).
2. Una variabile di classe presenta **sempre lo stesso valore** in tutte le istanze (oggetti) della classe.

---

### Il Name Mangling applicato alle variabili di classe

La storpiatura del nome (*mangling*) su una variabile di classe produce gli stessi identici effetti che abbiamo già incontrato con le variabili di istanza.

Se davanti al nome di una variabile di classe si aggiungono due trattini bassi (es. `__counter`), Python ne modificherà il nome internamente includendo il prefisso della classe (es. `_NomeClasse__counter`), proteggendola così da accessi diretti e accidentali dall'esterno.

````python
class ExampleClass:
    __counter = 0
    def __init__(self, val = 1):
        self.__first = val
        ExampleClass.__counter += 1


example_object_1 = ExampleClass()
example_object_2 = ExampleClass(2)
example_object_3 = ExampleClass(4)

print(example_object_1.__dict__, example_object_1._ExampleClass__counter)
print(example_object_2.__dict__, example_object_2._ExampleClass__counter)
print(example_object_3.__dict__, example_object_3._ExampleClass__counter)
````

````python
class ExampleClass:
    varia = 1
    def __init__(self, val):
        ExampleClass.varia = val


print(ExampleClass.__dict__)
example_object = ExampleClass(2)

print(ExampleClass.__dict__)
print(example_object.__dict__)
    
````

Diamo un'occhiata più da vicino al comportamento delle variabili di classe e di istanza:

* Definiamo una classe chiamata `ExampleClass`;
* La classe definisce una variabile di classe chiamata `varia`;
* Il costruttore della classe imposta questa variabile con il valore del parametro.

Il modo in cui si fa riferimento alla variabile è l'aspetto più importante dell'esempio perché:

* Cambiare l'assegnazione in `self.varia = val` creerebbe una **variabile di istanza** con lo stesso nome di quella della classe;
* Cambiare l'assegnazione in `varia = val` opererebbe semplicemente su una **variabile locale** del metodo. (Ti incoraggiamo vivamente a testare entrambi i casi sopra citati: ti aiuterà a ricordare meglio la differenza).
* La prima riga di codice all'esterno della classe stampa il valore dell'attributo `ExampleClass.varia`. Nota bene: utilizziamo questo valore **prima** ancora che venga istanziato il primissimo oggetto della classe.

Esegui il codice nell'editor e controlla il suo output.

## Verificare l'esistenza di un attributo

L'approccio di Python all'istanza degli oggetti solleva una questione importante: a differenza di altri linguaggi di programmazione, non puoi dare per scontato che tutti gli oggetti della stessa classe possiedano lo stesso insieme di proprietà.

Esattamente come nell'esempio presente nell'editor. Guardalo attentamente.

```python
class ExampleClass:
    def __init__(self, val):
        if val % 2 != 0:
            self.a = 1
        else:
            self.b = 1


example_object = ExampleClass(1)

print(example_object.a)
print(example_object.b)
```

L'oggetto creato dal costruttore può avere solo uno tra due possibili attributi: `a` oppure `b`.

L'esecuzione di un codice di questo tipo produrrà il seguente output:

**Output:**

```text
1
Traceback (most recent call last):
  File ".main.py", line 11, in <module>
    print(example_object.b)
AttributeError: 'ExampleClass' object has no attribute 'b'

```

Come puoi vedere, il tentativo di accedere a un attributo dell'oggetto (o della classe) non esistente provoca un'eccezione di tipo `AttributeError`.

L'istruzione `try-except` ti offre la possibilità di evitare problemi legati a proprietà inesistenti. È molto semplice: guarda il codice nell'editor.

````python
class ExampleClass:
    def __init__(self, val):
        if val % 2 != 0:
            self.a = 1
        else:
            self.b = 1


example_object = ExampleClass(1)
print(example_object.a)

try:
    print(example_object.b)
except AttributeError:
    pass
````

Come puoi vedere, questa azione non è molto sofisticata. In sostanza, abbiamo solo nascosto il problema sotto il tappeto. Fortunatamente, esiste un altro modo per gestire la questione.

Python fornisce una funzione in grado di verificare in sicurezza se un oggetto o una classe contengono una determinata proprietà. La funzione si chiama **`hasattr`** e richiede il passaggio di due argomenti:

* La classe o l'oggetto da controllare;
* Il nome della proprietà la cui esistenza deve essere verificata (nota: deve essere una **stringa** contenente il nome dell'attributo, non il nome da solo).

La funzione restituisce `True` o `False`. Ecco come puoi utilizzarla:

```python
class ExampleClass:
    def __init__(self, val):
        if val % 2 != 0:
            self.a = 1
        else:
            self.b = 1
 
 
example_object = ExampleClass(1)
print(example_object.a)
 
if hasattr(example_object, 'b'):
    print(example_object.b)

```

Non dimenticare che la funzione `hasattr()` può operare anche sulle **classi**. Puoi usarla per scoprire se una variabile di classe è disponibile, proprio come nell'esempio presente nell'editor.

## I metodi in dettaglio

Riassumiamo tutti i fatti fondamentali riguardanti l'uso dei metodi all'interno delle classi in Python.

Come già sai, un metodo è essenzialmente una funzione incorporata (*embedded*) all'interno di una classe. Esiste però un **requisito fondamentale**: un metodo è obbligato ad avere **almeno un parametro**. Non esistono metodi dichiarati completamente senza parametri; un metodo può essere *invocato* senza passare argomenti, ma non può essere *definito* senza parametri.

Il primo (e talvolta l'unico) parametro viene tradizionalmente chiamato **`self`**. Ti consigliamo vivamente di seguire sempre questa convenzione: è uno standard universale nel mondo Python e useresti nomi diversi a tuo rischio e pericolo, creando molta confusione in chi legge il tuo codice.

Il nome stesso `self` suggerisce lo scopo di questo parametro: identifica in modo univoco **l'oggetto specifico (l'istanza)** per il quale il metodo è stato invocato. Quando chiami un metodo su un oggetto, non devi passare esplicitamente l'argomento per il parametro `self`: Python lo fa automaticamente per te, passando il riferimento all'oggetto stesso.

L'esempio nell'editor mostra chiaramente questa dinamica e la differenza tra la definizione e l'invocazione. Dai un'occhiata al codice per vedere come si comporta.

````python
class Classy:
    def method(self):
        print("method")


obj = Classy()
obj.method()
````

Nota il modo in cui abbiamo creato l'oggetto: abbiamo trattato il nome della classe come se fosse una funzione. Questa chiamata restituisce un oggetto appena istanziato di quella specifica classe.

2. Passa i relativi argomenti durante l'invocazione **senza specificare `self**` (esattamente come abbiamo visto prima).

## Il Costruttore: `__init__`

Se assegni a un metodo il nome speciale **`__init__`**, non avrai più un metodo comune, ma avrai definito il **costruttore** della classe.

Quando una classe possiede un costruttore, questo viene **invocato automaticamente e implicitamente** nel momento esatto in cui l'oggetto della classe viene istanziato.

Il costruttore:

* **È obbligato ad avere il parametro `self**` (che viene impostato automaticamente da Python, come di consueto).
* Può (ma non deve necessariamente) avere più parametri oltre a `self`. Se sono presenti parametri aggiuntivi, il modo in cui usi il nome della classe per creare l'oggetto dovrà riflettere gli argomenti richiesti dalla definizione di `__init__`.
* Viene utilizzato principalmente per **configurare l'oggetto**, ovvero per inizializzare correttamente il suo stato interno, creare le variabili di istanza, allocare risorse o istanziare altri oggetti necessari al suo funzionamento.

Dai un'occhiata al codice nell'editor: l'esempio mostra un costruttore molto semplice all'opera.

````python
class Classy:
    def __init__(self, value):
        self.var = value


obj_1 = Classy("object")

print(obj_1.var)
````

Nota bene che il costruttore:

* **Non può restituire alcun valore** tramite l'istruzione `return`, in quanto è progettato nativamente per restituire l'oggetto appena creato e nient'altro. Provare a fargli restituire qualcosa che non sia `None` causerà un errore di runtime (`TypeError`).
* **Non può essere invocato direttamente** né dall'oggetto appena creato né dall'interno della classe stessa come se fosse un metodo comune (è possibile invocare il costruttore di una classe superiore all'interno di una sua sottoclasse, ma approfondiremo questo scenario più avanti).

Poiché `__init__` è a tutti gli effetti un metodo, e un metodo è strutturalmente una funzione, puoi applicare ai costruttori e ai metodi gli stessi identici "trucchi" che usi con le funzioni ordinarie.

Ad esempio, puoi definire parametri opzionali impostando dei valori di default. L'esempio nell'editor mostra proprio come dichiarare un costruttore con un argomento predefinito. Eseguilo per vedere come reagisce quando crei l'oggetto passando o omettendo quel parametro.

````python
class Classy:
    def __init__(self, value = None):
        self.var = value


obj_1 = Classy("object")
obj_2 = Classy()

print(obj_1.var)
print(obj_2.var)
````

Tutto ciò che abbiamo visto in precedenza riguardo al name mangling (la manipolazione dei nomi) per le proprietà si applica esattamente allo stesso modo anche ai nomi dei metodi: un metodo il cui nome inizia con un doppio trattino basso (`__`) viene (parzialmente) nascosto all'esterno della classe.

````python
class Classy:
    def visible(self):
        print("visible")

    def __hidden(self):
        print("hidden")


obj = Classy()
obj.visible()

try:
    obj.__hidden()
except:
    print("failed")

obj._Classy__hidden()
````

## La vita interiore di classi e oggetti

Ogni classe e ogni oggetto in Python sono equipaggiati fin dalla nascita con una serie di attributi predefiniti utili per esaminare le loro caratteristiche e capacità in tempo reale (un processo chiamato *introspezione*).

Conosci già una di queste proprietà speciali: si tratta di **`__dict__`**.

Vediamo ora come si comporta questo dizionario nei confronti dei metodi. Dai un'occhiata al codice nell'editor ed eseguilo per analizzare la struttura interna della classe e dell'istanza.

````python
class Classy:
    varia = 1
    def __init__(self):
        self.var = 2

    def method(self):
        pass

    def __hidden(self):
        pass


obj = Classy()

print(obj.__dict__)
print(Classy.__dict__)
````

Quando esegui il codice per esaminare la proprietà `__dict__`, l'output mostra chiaramente la netta separazione tra lo spazio di memoria dell'oggetto (l'istanza) e quello della classe.

Ecco cosa scoprirai analizzando l'output:

* **All'interno dell'oggetto (`obj.__dict__`):** Troverai soltanto le variabili di istanza (le proprietà specifiche di quell'oggetto, come quelle create dentro il costruttore `__init__` o associate a `self`). I metodi **non** risiedono qui.
* **All'interno della classe (`Classy.__dict__`):** È qui che vivono tutti i metodi che hai definito (`__init__`, `method`, ecc.), memorizzati come riferimenti a funzioni, insieme alle variabili di classe e ad altri attributi di sistema.

Python fa questo per ottimizzare la memoria: non avrebbe senso duplicare il codice dei metodi per ogni singolo oggetto creato; tutti gli oggetti della stessa classe condividono gli stessi metodi memorizzati nella classe madre.

### La proprietà `__name__`

Un'altra proprietà integrata fondamentale è **`__name__`**, che contiene semplicemente una stringa con il nome della classe.

> **Nota bene:** L'attributo `__name__` **non esiste all'interno dell'oggetto**. Esiste esclusivamente all'interno delle classi.

Se provi a scrivere `obj.__name__` otterrai un errore (`AttributeError`). Se hai a disposizione solo l'oggetto e vuoi conoscere il nome della sua classe, devi prima risalire alla classe tramite la proprietà speciale `__class__`, in questo modo:

```python
print(obj.__class__.__name__)

```

### La funzione `type()`

Se vuoi scoprire al volo la classe di un particolare oggetto, puoi usare la funzione integrata **`type()`**. Tra le sue varie funzionalità, questa funzione è in grado di restituire proprio la classe che è stata utilizzata per istanziare l'oggetto in questione.

Dai un'occhiata al codice nell'editor, eseguilo e osserva come `type(obj)` ti restituisca il riferimento diretto alla classe d'origine.

````python
class Classy:
    pass


print(Classy.__name__)
obj = Classy()
print(type(obj).__name__)
    
````


**causerà un errore** di tipo `AttributeError`. Questo accade perché gli oggetti (le istanze) non possiedono l'attributo `__name__`; solo le classi lo contengono nativamente.


### La proprietà `__module__`

Anche **`__module__`** è una stringa speciale integrata. Il suo scopo è memorizzare il **nome del modulo** che contiene la definizione di quella specifica classe.

Se la classe è definita all'interno del file principale che stai eseguendo in quel momento, il valore di questa proprietà sarà la stringa speciale `__main__`. Se invece la classe fa parte di un modulo importato da un altro file (ad esempio, se importi la classe `Canvas` da un ipotetico modulo `draw`), la proprietà conterrà il nome del modulo stesso (in quel caso, `'draw'`).

Dai un'occhiata al codice nell'editor ed eseguilo per verificare cosa restituisce nel tuo ambiente di test.

````python
class Classy:
    pass


print(Classy.__module__)
obj = Classy()
print(obj.__module__)
    
````

Esatto, hai centrato perfettamente il punto: quando vedi la stringa `__main__`, significa che la classe è nata all'interno del file che ha dato il via all'esecuzione del programma.

### La proprietà `__bases__`

La proprietà speciale **`__bases__`** è una **tupla**. Questa tupla contiene le classi reali (attenzione: gli oggetti classe stessi, non le semplici stringhe con i loro nomi) che fungono da **superclassi dirette** (classi madri) per la classe in questione.

L'ordine in cui le classi appaiono all'interno della tupla rispecchia fedelmente l'ordine in cui sono state elencate nella definizione della classe (fondamentale quando si usa l'ereditarietà multipla).

> **Nota bene:** Esattamente come per `__name__`, anche l'attributo `__bases__` **esiste solo all'interno delle classi**. Gli oggetti (le istanze) non lo possiedono.

Per vedere come funziona, analizziamo un esempio classico. Abbiamo definito una funzione di supporto chiamata `printbases()` per formattare l'output in modo pulito e mostrarci il contenuto di questa tupla:

````python
class SuperOne:
    pass


class SuperTwo:
    pass


class Sub(SuperOne, SuperTwo):
    pass


def printBases(cls):
    print('( ', end='')

    for x in cls.__bases__:
        print(x.__name__, end=' ')
    print(')')


printBases(SuperOne)
printBases(SuperTwo)
printBases(Sub)
    
````

Se analizzi ed esegui questo codice, noterai due comportamenti speculari molto interessanti:

* Per la classe `Sub`, l'output sarà `['SuperOne', 'SuperTwo']`. La tupla registra con precisione le sue genitrici dirette nell'ordine esatto.
* Per la classe `SuperOne` (che non sembra ereditare da nessuno), l'output in Python 3 sarà `['object']`. Questo accade perché, anche se non lo scrivi esplicitamente, in Python ogni classe eredita automaticamente dalla classe base suprema del linguaggio, chiamata appunto `object`.

Questo attributo ci tornerà utilissimo più avanti, soprattutto per mappare ed esplorare l'albero genealogico delle eccezioni.

## Riflessione e introspezione

Tutti questi strumenti permettono al programmatore Python di eseguire due attività cruciali, tipiche di molti linguaggi orientati agli oggetti:

* **L'introspezione**, ovvero la capacità di un programma di esaminare il tipo o le proprietà di un oggetto in tempo reale, durante l'esecuzione (*runtime*).
* **La riflessione**, che si spinge un passo oltre ed è la capacità di un programma di manipolare i valori, le proprietà e/o le funzioni di un oggetto sempre a *runtime*.

In altre parole, non hai bisogno di conoscere l'intera definizione di una classe o di un oggetto per poterlo gestire o modificare: l'oggetto stesso (e la sua classe) contengono già tutti i **metadati** necessari per permetterti di riconoscerne e modificarne le caratteristiche mentre il programma è in esecuzione.

## Investigare le classi

Cosa puoi scoprire sulle classi in Python? La risposta è disarmante: **tutto**.

Sia la riflessione che l'introspezione offrono a chi programma il potere totale di analizzare e manipolare qualsiasi oggetto a runtime, indipendentemente dalla sua provenienza (che sia una classe scritta da te o un modulo di terze parti precompilato).

Attraverso strumenti come `__dict__`, `type()`, `hasattr()`, `getattr()` e le altre proprietà integrate che abbiamo appena visto, il codice è in grado di "auto-esaminarsi".

Dai un'occhiata al codice nell'editor per vedere come questi concetti si traducono in pratica: analizzalo ed eseguilo per osservare come Python metta a nudo la struttura profonda di ogni sua componente.

````python
class MyClass:
    pass


obj = MyClass()
obj.a = 1
obj.b = 2
obj.i = 3
obj.ireal = 3.5
obj.integer = 4
obj.z = 5


def incIntsI(obj):
    for name in obj.__dict__.keys():
        if name.startswith('i'):
            val = getattr(obj, name)
            if isinstance(val, int):
                setattr(obj, name, val + 1)


print(obj.__dict__)
incIntsI(obj)
print(obj.__dict__)
    
````

Ecco come Python rende possibile quella che a prima vista potrebbe sembrare una magia. Grazie all'introspezione e alla riflessione, possiamo analizzare un oggetto sconosciuto "al microscopio" e modificarne il contenuto al volo.

Analizziamo il funzionamento del codice passo dopo passo, seguendo la logica delle righe:

* **Riga 1:** Viene definita una classe molto semplice (ad esempio `class MyClass:`).
* **Righe da 5 a 11:** Si istanzia l'oggetto e lo si riempie con diversi attributi di vario tipo (stringhe, numeri interi, float), alcuni dei quali iniziano con la lettera `i`.
* **Riga 14:** Ecco la nostra funzione `incIntsI(obj)`.
* **Riga 15:** La funzione avvia un ciclo `for` che scansiona il dizionario `obj.__dict__.keys()`. In questo modo estraiamo uno alla volta i nomi di tutte le proprietà presenti in quell'oggetto specifico.
* **Riga 16:** Un controllo condizionale verifica se il nome dell'attributo (che è una stringa) inizia con la lettera ricercata: `if name.startswith('i'):`.
* **Riga 17:** Se la condizione è vera, usiamo la funzione integrata **`getattr()`** per recuperare il valore attuale di quella proprietà.
> `getattr(object, name)` accetta due argomenti: l'oggetto da esaminare e il nome della proprietà (sotto forma di stringa). Restituisce il valore memorizzato in quella proprietà.


* **Riga 18:** Verifichiamo se il valore recuperato è effettivamente un numero intero utilizzando la funzione **`isinstance()`** (es. `if isinstance(val, int):`). Questa funzione controlla se un oggetto appartiene a una determinata classe o tipo di dato.
* **Riga 19:** Se anche questo controllo va a buon fine, incrementiamo il valore di uno e lo salviamo nuovamente nell'oggetto usando la funzione **`setattr()`**.
> `setattr(object, name, value)` accetta tre argomenti: l'oggetto da modificare, il nome della proprietà (come stringa) e il nuovo valore da assegnarle.



Il risultato? Tutte le proprietà intere che iniziano con "i" sono state aggiornate, mentre le altre sono rimaste intatte, il tutto senza che la funzione sapesse a priori come fosse strutturata la classe originale!

# OOP: Fondamenti di Ereditarietà

## Ereditarietà: come e perché?

Prima di addentrarci nei meccanismi dell'ereditarietà, c'è un altro strumento estremamente comodo e nativo di Python che dobbiamo scoprire. Riguarda il modo in cui un oggetto è in grado di **presentare se stesso** (o meglio, di convertirsi in una stringa leggibile) quando viene stampato o esaminato.

Fino ad ora, se provavi a passare un intero oggetto direttamente alla funzione `print()`, l'output standard di Python non era molto amichevole. Mostrava solitamente qualcosa del genere:
`<__main__.Classy object at 0x7f81b0...>`

Questo ci dice il nome della classe e la sua posizione in memoria, ma non ci dice nulla sul suo reale contenuto. Esiste un metodo speciale per cambiare radicalmente questo comportamento.

Diamo un'occhiata all'esempio nell'editor per vedere come gli oggetti possono iniziare a "raccontarsi" da soli in modo chiaro.

````python
class Star:
    def __init__(self, name, galaxy):
        self.name = name
        self.galaxy = galaxy


sun = Star("Sun", "Milky Way")
print(sun)
````

Se esegui lo stesso codice sul tuo computer vedrai un output molto simile, anche se il numero esadecimale (la parte che inizia con `0x`) sarà diverso. Quel valore non è altro che l'indirizzo di memoria dell'oggetto o l'identificatore interno usato dall'interprete, ed è pressoché impossibile che coincida tra esecuzioni in ambienti differenti.

Come hai notato, questa stampa standard è piuttosto criptica: non descrive lo stato dell'oggetto ed è decisamente poco elegante. Sarebbe molto meglio poter mostrare qualcosa di più specifico, chiaro e personalizzato.

Fortunatamente Python mette a disposizione un meccanismo nativo per farlo.

Quando Python ha la necessità di convertire o rappresentare un oggetto sotto forma di stringa — ad esempio quando lo passi come argomento alla funzione `print()` o a `str()` — l'interprete cerca e invoca automaticamente un metodo speciale chiamato **`__str__()`** all'interno dell'oggetto, utilizzandone il testo restituito.

Il metodo `__str__()` predefinito (quello ereditato implicitamente) restituisce proprio quella stringa un po' grezza che abbiamo visto prima. Per cambiare le cose ti basta **sovrascrivere questo metodo**, definendone uno tuo con lo stesso identico nome all'interno della classe.


````python

class Star:
    def __init__(self, name, galaxy):
        self.name = name
        self.galaxy = galaxy

    def __str__(self):
        return self.name + ' in ' + self.galaxy


sun = Star("Sun", "Milky Way")
print(sun)
    
````

Questo nuovo metodo `__str__()` compone una stringa unendo il nome della stella e quello della galassia: niente di trascendentale, ma il risultato stampato a schermo ha un aspetto decisamente migliore, non trovi?

Se hai indovinato l'output prima di premere "Run", avrai visto comparire esattamente la descrizione pulita che abbiamo configurato, al posto del vecchio e criptico indirizzo di memoria.


### L'ereditarietà: cos'è e come funziona

Il termine **ereditarietà** (*inheritance*) è molto più antico della programmazione informatica e descrive la pratica comune di trasferire beni da una persona all'altra dopo la sua scomparsa. Nel contesto della programmazione, tuttavia, assume un significato completamente diverso.

Definiamo questo termine per i nostri scopi:

> L'**ereditarietà** è la pratica standard (nella programmazione orientata agli oggetti) che consiste nel trasferire attributi e metodi da una classe esistente, chiamata **superclassi** (o classe madre), a una classe appena creata, detta **sottoclasse** (o classe figlia).

In altre parole, l'ereditarietà è un modo per costruire una nuova classe non partendo da zero, ma riutilizzando un repertorio di caratteristiche già pronte e definite. La nuova classe eredita (ed è questa la chiave) tutto l'equipaggiamento esistente, ma ha la possibilità di aggiungere nuovi elementi o funzionalità se necessario.

Grazie a questo approccio, è possibile strutturare classi via via più specializzate (più concrete) partendo da un insieme di regole e comportamenti generali già predefiniti.

### Le relazioni gerarchiche

Il fattore più importante di questo processo è la relazione che si stabilisce tra la superclasse e tutte le sue sottoclassi.

Un dettaglio fondamentale: questa relazione è completamente **transitiva**. Significa che se la classe *B* è una sottoclasse di *A*, e la classe *C* è una sottoclasse di *B*, allora anche *C* è automaticamente una sottoclasse di *A*.

Ecco un esempio molto semplice di ereditarietà a due livelli:

```python
class Vehicle:
    pass
 
 
class LandVehicle(Vehicle):
    pass
 
 
class TrackedVehicle(LandVehicle):
    pass

```

Tutte le classi presentate sono momentaneamente vuote (`pass`), poiché l'obiettivo attuale è comprendere come funzionano le relazioni reciproche. Le riempiremo di contenuti molto presto.

Analizzando il codice, possiamo affermare che:

* La classe `Vehicle` è la superclasse sia di `LandVehicle` che di `TrackedVehicle`.
* La classe `LandVehicle` è contemporaneamente una sottoclasse di `Vehicle` e la superclasse di `TrackedVehicle`.
* La classe `TrackedVehicle` è una sottoclasse sia di `LandVehicle` che di `Vehicle` (per la proprietà transitiva).

Noi riusciamo a comprendere questa gerarchia semplicemente leggendo le righe di codice. Ma Python è altrettanto consapevole di questi legami? È possibile interrogarlo a riguardo? Certamente.


## La funzione `issubclass()`

Python mette a disposizione una funzione integrata in grado di identificare la relazione di parentela tra due classi. Anche se la sua diagnosi è diretta, è uno strumento potentissimo per verificare se una determinata classe è una sottoclasse di un'altra.

La sintassi si presenta così:

```python
issubclass(ClasseUno, ClasseDue)

```

La funzione restituisce **`True`** se `ClasseUno` è una sottoclasse di `ClasseDue`, e **`False`** in caso contrario.

Vediamola in azione nel codice dell'editor: analizzalo attentamente, i risultati potrebbero riservarti qualche sorpresa!

````python
class Vehicle:
    pass


class LandVehicle(Vehicle):
    pass


class TrackedVehicle(LandVehicle):
    pass


for cls1 in [Vehicle, LandVehicle, TrackedVehicle]:
    for cls2 in [Vehicle, LandVehicle, TrackedVehicle]:
        print(issubclass(cls1, cls2), end="\t")
    print()
    
````

Ecco cosa succede dietro le quinte: i due cicli annidati prendono le nostre tre classi, creano tutte le combinazioni possibili e interrogano `issubclass()`.

Se organizziamo il risultato in una struttura a matrice per renderlo immediatamentamente leggibile, l'output mostra una mappa chiarissima delle relazioni:

| Classe esaminata | ...è sottoclasse di `Vehicle`? | ...è sottoclasse di `LandVehicle`? | ...è sottoclasse di `TrackedVehicle`? |
| --- | --- | --- | --- |
| **`Vehicle`** | **`True`** | `False` | `False` |
| **`LandVehicle`** | **`True`** | **`True`** | `False` |
| **`TrackedVehicle`** | **`True`** | **`True`** | **`True`** |


### Un'osservazione fondamentale

Analizzando la diagonale della tabella (i valori in grassetto), balza subito all'occhio un dettaglio cruciale: **ogni classe è considerata sottoclasse di se stessa**.

```python
issubclass(Vehicle, Vehicle) # Restituisce True

```

In Python, la relazione verificata da `issubclass(A, B)` non è di tipo "sottoclasse stretta". Corrisponde invece al concetto matematico di sottoinsieme in senso lato ($\subseteq$): la funzione restituisce `True` se la classe `A` eredita da `B`, oppure se `A` e `B` sono **la stessa identica classe**.

Inoltre, guardando la riga di `TrackedVehicle`, puoi vedere la prova pratica della **transitività**: è `True` sia per `LandVehicle` (sua madre diretta) sia per `Vehicle` (sua nonna), confermando che la catena di ereditarietà viene risalita e riconosciuta interamente dall'interprete.

## `isinstance()`

Come sai già, un oggetto è l'incarnazione di una classe. Questo significa che l'oggetto è un po' come una torta sfornata seguendo una ricetta contenuta all'interno della classe.

Questo scenario può sollevare alcune questioni importanti.
Ipotizziamo che ti venga consegnata una torta (ad esempio come argomento passato a una tua funzione). Vuoi sapere quale ricetta sia stata usata per prepararla. Perché? Perché vuoi sapere cosa aspettarti, ad esempio se contiene noci o meno, un'informazione che per alcune persone è fondamentale.

Allo stesso modo, nella programmazione può essere cruciale sapere se un oggetto possiede (o non possiede) determinate caratteristiche. In altre parole, se è l'oggetto di una specifica classe o meno.

Un fatto del genere può essere rilevato tramite la funzione chiamata **`isinstance()`**:

```python
isinstance(nomeOggetto, NomeClasse)

```

La funzione restituisce **`True`** se l'oggetto è un'istanza della classe, oppure **`False`** in caso contrario.

Essere un'istanza di una classe significa che l'oggetto (la torta) è stato preparato usando una ricetta contenuta in quella specifica classe **oppure in una delle sue superclassi**.

Non dimenticare: se una sottoclasse contiene come minimo lo stesso equipaggiamento di una qualsiasi delle sue superclassi, significa che gli oggetti della sottoclasse possono fare le stesse identiche cose degli oggetti derivati dalla superclasse. Ergo, l'oggetto è un'istanza sia della sua classe nativa, sia di tutte le sue superclassi.


Mettiamolo alla prova. Analizza il codice nell'editor ed eseguilo per osservare come `isinstance()` risalga l'albero genealogico dell'oggetto.

````python
class Vehicle:
    pass


class LandVehicle(Vehicle):
    pass


class TrackedVehicle(LandVehicle):
    pass


my_vehicle = Vehicle()
my_land_vehicle = LandVehicle()
my_tracked_vehicle = TrackedVehicle()

for obj in [my_vehicle, my_land_vehicle, my_tracked_vehicle]:
    for cls in [Vehicle, LandVehicle, TrackedVehicle]:
        print(isinstance(obj, cls), end="\t")
    print()
    
````

Sì, la tabella conferma perfettamente le aspettative e riflette la struttura gerarchica che abbiamo impostato.

Se associamo visivamente l'output del codice alle combinazioni dei cicli annidati, otteniamo questa matrice di leggibilità:

| Oggetto esaminato | ...è istanza di `Vehicle`? | ...è istanza di `LandVehicle`? | ...è istanza di `TrackedVehicle`? |
| --- | --- | --- | --- |
| **`my_vehicle`** (da `Vehicle`) | **`True`** | `False` | `False` |
| **`my_land_vehicle`** (da `LandVehicle`) | **`True`** | **`True`** | `False` |
| **`my_tracked_vehicle`** (da `TrackedVehicle`) | **`True`** | **`True`** | **`True`** |

### Perché questo comportamento è logico e coerente?

1. **La riga di `my_vehicle` (Prima riga: `True, False, False`)**:
Questo oggetto è stato creato dalla ricetta base (`Vehicle`). Conosce le caratteristiche generali di un veicolo, ma non ha la minima idea di cosa sia un veicolo terrestre o cingolato. Pertanto, è istanza solo della sua classe d'origine.
2. **La riga di `my_land_vehicle` (Seconda riga: `True, True, False`)**:
Questo oggetto è un veicolo terrestre. È un'istanza di `LandVehicle` (la sua classe) ma è **anche** un'istanza di `Vehicle` (la superclasse). Questo accade perché un veicolo terrestre *è a tutti gli effetti* un veicolo. Non è invece un cingolato, quindi l'ultimo controllo è `False`.
3. **La riga di `my_tracked_vehicle` (Terza riga: `True, True, True`)**:
Il cingolato si trova in fondo alla catena. Essendo la classe più specializzata, eredita l'equipaggiamento di tutte quelle sopra di lei. Di conseguenza, l'oggetto risponde `True` a tutti i test: è contemporaneamente un'istanza di `TrackedVehicle`, di `LandVehicle` e di `Vehicle`.

In sintesi, `isinstance()` ragiona secondo la logica del legame **"IS-A" (È un)**: un cingolato *è un* veicolo terrestre, e un veicolo terrestre *è un* veicolo.

## L'operatore `is`

Esiste un altro operatore in Python che merita di essere menzionato, poiché lavora direttamente con gli oggetti: l'operatore **`is`**.

```python
oggetto_uno is oggetto_due

```

L'operatore `is` verifica se due variabili (in questo caso `oggetto_uno` e `oggetto_due`) **fanno riferimento allo stesso identico oggetto** in memoria.

Per capire come funziona, è fondamentale ricordare un concetto chiave di Python: le variabili non contengono fisicamente gli oggetti, ma ospitano soltanto dei **riferimenti** (o puntatori) che indirizzano alla posizione dell'oggetto nella memoria interna di Python.

Di conseguenza, quando assegni il valore di una variabile-oggetto a un'altra variabile (es. `var_b = var_a`), **non stai copiando l'oggetto**, ma stai semplicemente duplicando il suo riferimento. Entrambe le variabili ora punteranno alla stessa identica area di memoria. Proprio per questo motivo, un operatore come `is` si rivela estremamente utile in circostanze particolari.

````python
class SampleClass:
    def __init__(self, val):
        self.val = val


object_1 = SampleClass(0)
object_2 = SampleClass(2)
object_3 = object_1
object_3.val += 1

print(object_1 is object_2)
print(object_2 is object_3)
print(object_3 is object_1)
print(object_1.val, object_2.val, object_3.val)

string_1 = "Mary had a little "
string_2 = "Mary had a little lamb"
string_1 += "lamb"

print(string_1 == string_2, string_1 is string_2)
    
````

Analizziamolo in dettaglio:

È presente una classe molto semplice dotata di un costruttore essenziale, che crea una sola proprietà. La classe viene utilizzata per istanziare due oggetti. Il primo dei due viene poi assegnato a un'altra variabile, e la sua proprietà `val` viene incrementata di uno.

Successivamente, l'operatore `is` viene applicato tre volte per controllare tutte le possibili coppie di oggetti; vengono inoltre stampati tutti i valori della proprietà `val`.

L'ultima parte del codice esegue un altro esperimento. Dopo tre assegnazioni, entrambe le stringhe contengono lo stesso identico testo, ma questi testi sono memorizzati in oggetti diversi.

I risultati dimostrano che `object_1` e `object_3` sono a tutti gli effetti lo stesso oggetto, mentre `string_1` e `string_2` non lo sono, nonostante il loro contenuto sia identico.

## Come Python trova proprietà e metodi

Adesso esamineremo da vicino il modo in cui Python gestisce l'ereditarietà dei metodi.

Dai un'occhiata all'esempio nell'editor per vedere come l'interprete si muove all'interno della gerarchia delle classi quando deve individuare ed eseguire un metodo.

````python
class Super:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return "My name is " + self.name + "."


class Sub(Super):
    def __init__(self, name):
        Super.__init__(self, name)


obj = Sub("Andy")

print(obj)
````

Analizziamolo in dettaglio:

È presente una classe chiamata `Super`, che definisce un proprio costruttore utilizzato per assegnare la proprietà dell'oggetto chiamata `name`.

La classe definisce anche il metodo `__str__()`, che permette alla classe stessa di presentare la propria identità sotto forma di testo chiaro.
Successivamente, la classe viene utilizzata come base per creare una sottoclasse chiamata `Sub`. La classe `Sub` definisce un proprio costruttore, che invoca quello della superclasse. Nota come abbiamo fatto: `Super.__init__(self, name)`.

Abbiamo nominato esplicitamente la superclasse e indicato il metodo da invocare, `__init__()`, fornendo tutti gli argomenti necessari.
Abbiamo istanziato un oggetto della classe `Sub` e lo abbiamo stampato.

> **Nota:** Poiché non c'è alcun metodo `__str__()` all'interno della classe `Sub`, la stringa stampata deve essere prodotta all'interno della classe `Super`. Questo significa che il metodo `__str__()` è stato ereditato dalla classe `Sub`.


Dai un'occhiata al codice nell'editor. Lo abbiamo modificato per mostrarti un altro modo per accedere a qualsiasi entità definita all'interno della superclasse.

Nell'ultimo esempio abbiamo nominato esplicitamente la superclasse. In questo esempio, invece, facciamo uso della funzione **`super()`**, che accede alla superclasse senza la necessità di conoscerne il nome:

````python
class Super:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return "My name is " + self.name + "."


class Sub(Super):
    def __init__(self, name):
        super().__init__(name)


obj = Sub("Andy")

print(obj)
    
````

La funzione `super()` crea un contesto nel quale non devi (anzi, non devi assolutamente) passare l'argomento `self` al metodo che stai invocando; ecco perché è possibile attivare il costruttore della superclasse utilizzando un solo argomento.

> **Nota:** Puoi usare questo meccanismo non solo per invocare il costruttore della superclasse, ma anche per ottenere l'accesso a una qualsiasi delle risorse disponibili all'interno della superclasse stessa.

Proviamo ora a fare qualcosa di simile, ma con le proprietà (più precisamente: con le variabili di classe).

Dai un'occhiata all'esempio nell'editor per vedere come funziona.

````python
# Testing properties: class variables.
class Super:
    supVar = 1


class Sub(Super):
    subVar = 2


obj = Sub()

print(obj.subVar)
print(obj.supVar)
````

Come puoi vedere, la classe `Super` definisce una variabile di classe chiamata `supVar`, mentre la classe `Sub` definisce una variabile chiamata `subVar`.

Lo stesso effetto può essere osservato anche con le variabili di istanza (proprietà specifiche dell'oggetto): dai un'occhiata al secondo esempio nell'editor per vedere come vengono ereditate all'atto dell'inizializzazione.

````python
# Testing properties: instance variables.
class Super:
    def __init__(self):
        self.supVar = 11


class Sub(Super):
    def __init__(self):
        super().__init__()
        self.subVar = 12


obj = Sub()

print(obj.subVar)
print(obj.supVar)
````

Il costruttore della classe `Sub` crea una variabile di istanza chiamata `subVar`, mentre il costruttore di `Super` fa lo stesso con una variabile chiamata `supVar`. Come in precedenza, entrambe le variabili sono accessibili dall'interno dell'oggetto della classe `Sub`.

> **Nota:** l'esistenza della variabile `supVar` è ovviamente condizionata dall'invocazione del costruttore della classe `Super`. Omettere questa chiamata comporterebbe l'assenza della variabile nell'oggetto creato (prova tu stesso).

A questo punto è possibile formulare una regola generale che descrive il comportamento di Python.
Quando tenti di accedere a una qualsiasi entità di un oggetto, Python proverà a cercarla seguendo questo ordine preciso:

1. All'interno dell'oggetto stesso;
2. In tutte le classi coinvolte nella linea di ereditarietà dell'oggetto, muovendosi dal basso verso l'alto;
3. Se entrambi i tentativi falliscono, viene sollevata un'eccezione (`AttributeError`).

La prima condizione merita un'attenzione particolare. Come sai, tutti gli oggetti che derivano da una specifica classe possono avere insiemi di attributi differenti, e alcuni di questi attributi possono essere aggiunti all'oggetto molto tempo dopo la sua creazione.

L'esempio nell'editor riassume tutto questo in una linea di ereditarietà a tre livelli. Analizzalo attentamente.

````python
class Level1:
    variable_1 = 100
    def __init__(self):
        self.var_1 = 101

    def fun_1(self):
        return 102


class Level2(Level1):
    variable_2 = 200
    def __init__(self):
        super().__init__()
        self.var_2 = 201
    
    def fun_2(self):
        return 202


class Level3(Level2):
    variable_3 = 300
    def __init__(self):
        super().__init__()
        self.var_3 = 301

    def fun_3(self):
        return 302


obj = Level3()

print(obj.variable_1, obj.var_1, obj.fun_1())
print(obj.variable_2, obj.var_2, obj.fun_2())
print(obj.variable_3, obj.var_3, obj.fun_3())
````

Tutti i commenti che abbiamo fatto finora sono relativi all'ereditarietà singola, ovvero quando una sottoclasse ha esattamente una superclasse. Questa è la situazione più comune (nonché quella raccomandata).

Python, tuttavia, offre molto di più in questo ambito. Nelle prossime lezioni ti mostreremo alcuni esempi di ereditarietà multipla.

L'ereditarietà multipla si verifica quando una classe ha più di una superclasse. Dal punto di vista sintattico, tale ereditarietà viene presentata come un elenco di superclasi separate da virgole, inserite all'interno delle parentesi dopo il nome della nuova classe — proprio come qui:

```python
class SuperA:
    var_a = 10
    def fun_a(self):
        return 11
 
 
class SuperB:
    var_b = 20
    def fun_b(self):
        return 21
 
 
class Sub(SuperA, SuperB):
    pass
 
obj = Sub()
 
print(obj.var_a, obj.fun_a())
print(obj.var_b, obj.fun_b())

```

La classe `Sub` ha due superclassi: `SuperA` e `SuperB`. Questo significa che la classe `Sub` eredita tutte le risorse offerte sia da `SuperA` che da `SuperB`.

Il codice stampa:

```text
10 11
20 21

```

Ora è giunto il momento di introdurre un termine completamente nuovo: **overriding** (sovrascrittura).
Cosa pensi che accadrà se più di una superclasse definisce un'entità con lo stesso identico nome?

Analizziamo l'esempio nell'editor.

````python
class Level1:
    var = 100
    def fun(self):
        return 101


class Level2(Level1):
    var = 200
    def fun(self):
        return 201


class Level3(Level2):
    pass


obj = Level3()

print(obj.var, obj.fun())
````

Entrambe le classi, `Level1` e `Level2`, definiscono un metodo chiamato `fun()` e una proprietà chiamata `var`. Questo significa forse che l'oggetto della classe `Level3` sarà in grado di accedere a due copie di ciascuna entità? Niente affatto.

L'entità definita più tardi (nel senso della linea di ereditarietà) effettua l'**overriding** (sovrascrive) la stessa entità definita in precedenza.

Come puoi notare, la variabile di classe `var` e il metodo `fun()` della classe `Level2` sovrascrivono le entità con lo stesso nome derivate dalla classe `Level1`.

Questa caratteristica può essere utilizzata intenzionalmente per modificare i comportamenti predefiniti (o definiti in precedenza) di una classe, qualora una delle sue sottoclassi abbia la necessità di agire in modo diverso rispetto al proprio antenato.

Possiamo anche dire che Python cerca un'entità muovendosi dal basso verso l'alto, e si ritiene pienamente soddisfatto non appena incontra la prima entità che possiede il nome desiderato.

Ma come funziona quando una classe ha due antenati che offrono la stessa identica entità e si trovano sullo stesso livello? In altre parole, cosa dobbiamo aspettarci quando una classe nasce sfruttando l'ereditarietà multipla? Scopriamolo.

Dai un'occhiata all'esempio nell'editor.

````python
class Left:
    var = "L"
    var_left = "LL"
    def fun(self):
        return "Left"


class Right:
    var = "R"
    var_right = "RR"
    def fun(self):
        return "Right"


class Sub(Left, Right):
    pass


obj = Sub()

print(obj.var, obj.var_left, obj.var_right, obj.fun())
    
````

La classe `Sub` eredita le proprie risorse da due superclassi, `Left` e `Right` (nomi scelti appositamente per essere significativi).

Non c'è alcun dubbio sul fatto che la variabile di classe `var_right` provenga dalla classe `Right` e che `var_left` provenga rispettivamente da `Left`.

Questo è chiaro. Ma da dove arriva invece `var`? È possibile indovinarlo? Lo stesso problema si presenta con il metodo `fun()`: verrà invocato da `Left` o da `Right`? Proviamo a eseguire il programma; il suo output è:

```text
L LL RR Left

```

Questo dimostra che entrambi i casi ambigui trovano una soluzione all'interno della classe `Left`. È una premessa sufficiente per formulare una regola generale? Certamente.

Possiamo affermare che Python cerca i componenti di un oggetto nel seguente ordine:

1. All'interno dell'oggetto stesso;
2. Nelle sue superclassi, muovendosi dal basso verso l'alto;
3. Se ci sono più classi sullo stesso livello in un percorso di ereditarietà, Python le scansiona **da sinistra a destra**.

Serve sapere altro? Ti basta fare una piccola modifica al codice — sostituisci `class Sub(Left, Right):` con `class Sub(Right, Left):`, poi esegui nuovamente il programma e guarda cosa succede.

Cosa vedi ora? L'output diventa:

```text
R LL RR Right

```

Vedi lo stesso risultato o qualcosa di diverso?

## Come strutturare una gerarchia di classi

Costruire una gerarchia di classi non è un semplice esercizio di stile o "arte per l'arte".

Quando dividi un problema strutturandolo in varie classi e decidi quali di esse debbano posizionarsi in cima e quali in fondo alla gerarchia, ti trovi a dover analizzare il problema con estrema attenzione. Prima di mostrarti come procedere (e quali errori evitare), vogliamo però mettere in evidenza un effetto interessante.

Non si tratta di nulla di straordinario — è semplicemente una diretta conseguenza delle regole generali che abbiamo esaminato finora — ma tenerlo a mente può essere la chiave di volta per comprendere a fondo il funzionamento di determinati codici e per capire come sfruttare questo stesso effetto per costruire un insieme di classi flessibile.

Dai un'occhiata al codice nell'editor per scoprire di cosa si tratta.

````python
class One:
    def do_it(self):
        print("do_it from One")

    def doanything(self):
        self.do_it()


class Two(One):
    def do_it(self):
        print("do_it from Two")


one = One()
two = Two()

one.doanything()
two.doanything()
````

Analizziamolo in dettaglio:

Sono presenti due classi, chiamate `One` e `Two`, in cui `Two` deriva da `One`. Niente di speciale. Tuttavia, c'è un dettaglio che salta all'occhio: il metodo `do_it()`.

Il metodo `do_it()` è definito due volte: originariamente all'interno di `One` e successivamente all'interno di `Two`. L'essenza di questo esempio risiede nel fatto che viene invocato una sola volta, all'interno di `One` (tramite il metodo `doanything()`).

La domanda è: quale dei due metodi verrà invocato dalle ultime due righe del codice?

La prima invocazione sembra semplice, e in effetti lo è: invocare `doanything()` dall'oggetto chiamato `one` attiverà ovviamente il primo dei metodi.

La seconda invocazione richiede invece un po' di attenzione. È semplice anche questa, a patto di tenere a mente il modo in cui Python individua i componenti delle classi. La seconda invocazione lancerà il metodo `do_it()` nella forma esistente all'interno della classe `Two`, a prescindere dal fatto che l'invocazione fisica avvenga dentro la classe `One`.

Di fatto, il codice produce il seguente output:

```text
do_it from One
do_it from Two

```

> **Nota:** La situazione in cui una sottoclasse è in grado di modificare il comportamento della sua superclasse (proprio come in questo esempio) prende il nome di **polimorfismo**. La parola viene dal greco (*polys*: "molto, tanto" e *morphe*: "forma"), il che significa che una medesima classe può assumere varie forme a seconda delle ridefinizioni effettuate dalle sue sottoclassi.

Il metodo che viene ridefinito in una qualsiasi delle sottoclassi, cambiando così il comportamento della superclasse, viene definito **virtuale**.

In altre parole, nessuna classe è scolpita nella pietra una volta per tutte. Il comportamento di ciascuna classe può essere modificato in qualsiasi momento da una delle sue sottoclassi.

Adesso ti mostreremo come utilizzare il polimorfismo per estendere la flessibilità delle classi.

Dai un'occhiata all'esempio nell'editor per vedere questa dinamica in azione.

````python
import time

class TrackedVehicle:
    def control_track(left, stop):
        pass

    def turn(left):
        control_track(left, True)
        time.sleep(0.25)
        control_track(left, False)


class WheeledVehicle:
    def turn_front_wheels(left, on):
        pass

    def turn(left):
        turn_front_wheels(left, True)
        time.sleep(0.25)
        turn_front_wheels(left, False)
    
````

Ti ricorda qualcosa? Sì, certamente. Fa riferimento all'esempio mostrato all'inizio del modulo, quando abbiamo parlato dei concetti generali della programmazione a oggetti.

Può sembrare strano, ma non abbiamo usato l'ereditarietà in alcun modo — volevamo solo dimostrarti che questo non ci limita e che siamo riusciti comunque a raggiungere il nostro obiettivo.

Abbiamo definito due classi separate in grado di produrre due tipi diversi di veicoli terrestri. La differenza principale tra loro sta nel modo in cui curvano. Un veicolo a ruote si limita a girare le ruote anteriori (generalmente). Un veicolo cingolato, invece, deve bloccare uno dei due cingoli.

Riesci a seguire il codice?

* Un veicolo cingolato effettua una curva bloccando e azionando uno dei suoi cingoli (questo viene fatto dal metodo `control_track()`, che verrà implementato più avanti);
* Un veicolo a ruote curva quando girano le sue ruote anteriori (questo viene fatto dal metodo `turn_front_wheels()`);
* Il metodo `turn()` utilizza il metodo più adatto per ciascuno specifico veicolo.

Riesci a vedere cosa c'è che non va in questo codice?

I metodi `turn()` si assomigliano decisamente troppo per lasciarli in questa forma.

Ristrutturiamo il codice: introdurremo una superclasse per raccogliere tutti gli aspetti simili dei veicoli in movimento, spostando poi le caratteristiche specifiche all'interno delle sottoclassi.

Dai un'occhiata di nuovo al codice nell'editor.

````python
import time

class Vehicle:
    def change_direction(left, on):
        pass

    def turn(left):
        change_direction(left, True)
        time.sleep(0.25)
        change_direction(left, False)


class TrackedVehicle(Vehicle):
    def control_track(left, stop):
        pass

    def change_direction(left, on):
        control_track(left, on)


class WheeledVehicle(Vehicle):
    def turn_front_wheels(left, on):
        pass

    def change_direction(left, on):
        turn_front_wheels(left, on)
````

Ecco cosa abbiamo fatto:

* Abbiamo definito una superclasse chiamata `Vehicle`, che utilizza il metodo `turn()` per implementare uno schema generale di svolta, mentre la svolta vera e propria viene eseguita da un metodo chiamato `change_direction()`. Nota: questo secondo metodo è vuoto, poiché inseriremo tutti i dettagli operativi all'interno delle sottoclassi (un metodo del genere viene spesso definito **metodo astratto**, in quanto dimostra solo una possibilità che verrà concretizzata in seguito);
* Abbiamo definito una sottoclasse chiamata `TrackedVehicle` (nota: deriva dalla classe `Vehicle`) che concretizza il metodo `change_direction()` utilizzando il metodo specifico (concreto) chiamato `control_track()`;
* Allo stesso modo, la sottoclasse chiamata `WheeledVehicle` esegue lo stesso stratagemma, ma utilizza il metodo `turn_front_wheels()` per costringere il veicolo a curvare.

Il vantaggio più importante (tralasciando i fattori di leggibilità) è che questa struttura del codice ti permette di implementare un algoritmo di svolta completamente nuovo modificando semplicemente il metodo `turn()`; questa operazione può essere fatta **in un unico punto**, poiché tutti i veicoli vi obbediranno.

Questo è il modo in cui il polimorfismo aiuta lo sviluppatore a mantenere il codice pulito e coerente.

### Composizione vs Ereditarietà

L'ereditarietà non è l'unico modo per costruire classi adattabili. Puoi raggiungere gli stessi obiettivi (non sempre, ma molto spesso) utilizzando una tecnica chiamata **composizione**.

La composizione è il processo di strutturazione di un oggetto che prevede l'utilizzo di altri oggetti distinti. Gli oggetti impiegati nella composizione forniscono un insieme di caratteristiche desiderate (proprietà e/o metodi), quindi possiamo dire che si comportano come blocchi costruttivi usati per edificare una struttura più complessa.

Possiamo riassumere la differenza in questo modo:

* **L'ereditarietà** estende le capacità di una classe aggiungendo nuovi componenti e modificando quelli esistenti; in altre parole, la ricetta completa è contenuta all'interno della classe stessa e di tutti i suoi antenati; l'oggetto fa propri tutti gli elementi della classe e li utilizza;
* **La composizione** progetta una classe come un contenitore in grado di memorizzare e utilizzare altri oggetti (derivati da altre classi), dove ciascuno di questi oggetti implementa una parte del comportamento desiderato della classe principale.

Illustriamo questa differenza riprendendo i veicoli definiti in precedenza. L'approccio precedente ci ha guidato verso una gerarchia di classi in cui la classe più in alto conosceva le regole generali per far curvare il veicolo, ma non sapeva come controllare i componenti specifici (ruote o cingoli).

Le sottoclassi implementavano questa capacità introducendo meccanismi specializzati. Facciamo adesso la stessa cosa (o quasi), ma usando la composizione. La classe principale — come nell'esempio precedente — è consapevole di come si fa a curvare il veicolo, ma la svolta effettiva viene eseguita da un oggetto specializzato memorizzato in una proprietà chiamata `controller`. Il controller è in grado di governare il veicolo manipolando le relative parti meccaniche.

Dai un'occhiata all'editor — ecco come potrebbe apparire questo scenario.

````python
import time

class Tracks:
    def change_direction(self, left, on):
        print("tracks: ", left, on)


class Wheels:
    def change_direction(self, left, on):
        print("wheels: ", left, on)


class Vehicle:
    def __init__(self, controller):
        self.controller = controller

    def turn(self, left):
        self.controller.change_direction(left, True)
        time.sleep(0.25)
        self.controller.change_direction(left, False)


wheeled = Vehicle(Wheels())
tracked = Vehicle(Tracks())

wheeled.turn(True)
tracked.turn(False)
````

Sono presenti due classi chiamate `Tracks` e `Wheels`, le quali sanno come controllare la direzione del veicolo. C'è anche una classe chiamata `Vehicle` che può utilizzare uno qualsiasi dei controller disponibili (i due già definiti, o altri che verranno definiti in futuro); il controller stesso viene passato alla classe durante la fase di inizializzazione.

In questo modo, la capacità del veicolo di curvare viene composta utilizzando un oggetto esterno, anziché essere implementata all'interno della classe `Vehicle`.

In altre parole, abbiamo a disposizione un veicolo universale sul quale possiamo installare, a nostra scelta, dei cingoli oppure delle ruote.

## Ereditarietà singola vs Ereditarietà multipla

Come già sai, in Python non ci sono ostacoli all'uso dell'ereditarietà multipla. Puoi far derivare qualsiasi nuova classe da più di una classe definita in precedenza.

C'è solo un "ma". Il fatto che tu possa farlo non significa che tu debba farlo.

Non dimenticare che:

* Una classe a **ereditarietà singola è sempre più semplice, sicura e facile** da comprendere e mantenere;
* L'ereditarietà multipla è **sempre rischiosa**, poiché hai molte più opportunità di commettere un errore nell'identificare quali parti delle superclassi influenzeranno effettivamente la nuova classe;
* L'ereditarietà multipla può rendere l'**overriding estremamente insidioso**; inoltre, l'uso della funzione `super()` diventa ambiguo;
* L'ereditarietà multipla **viola il principio di singola responsabilità** (*Single Responsibility Principle*), in quanto unisce in una nuova classe due (o più) classi che non sanno nulla l'una dell'altra;
* Consigliamo caldamente l'ereditarietà multipla come **l'ultima delle soluzioni possibili** — se hai davvero bisogno di molte funzionalità differenti offerte da classi diverse, la composizione può essere un'alternativa migliore.


## Cos'è il Method Resolution Order (MRO) e perché non tutte le linee di ereditarietà hanno senso?

Il MRO, in generale, è il meccanismo (puoi pensarlo come a una strategia) con cui un determinato linguaggio di programmazione scansiona la parte superiore della gerarchia di una classe per trovare il metodo di cui ha bisogno in quel momento. Vale la pena sottolineare che lingue diverse usano MRO leggermente (o anche completamente) differenti. Python è una creatura unica sotto questo aspetto e le sue consuetudini sono piuttosto specifiche.

Ti mostreremo come funziona il MRO di Python in due casi particolari, che sono chiari esempi dei problemi che possono verificarsi quando si tenta di usare l'ereditarietà multipla in modo troppo spericolato. Iniziamo con un frammento di codice che all'inizio potrebbe sembrare semplice. Guarda cosa abbiamo preparato per te nell'editor.

````python
class Top:
    def m_top(self):
        print("top")


class Middle(Top):
    def m_middle(self):
        print("middle")


class Bottom(Middle):
    def m_bottom(self):
        print("bottom")


object = Bottom()
object.m_bottom()
object.m_middle()
object.m_top()
````

Siamo certi che se analizzi questo frammento di codice da solo, non noterai alcuna anomalia. Sì, hai perfettamente ragione — sembra chiaro, semplice e non desta alcuna preoccupazione.

Nessuna sorpresa finora. Applichiamo adesso una piccolissima modifica a questo codice. Dai un'occhiata:

````python
class Top:
    def m_top(self):
        print("top")


class Middle(Top):
    def m_middle(self):
        print("middle")


class Bottom(Middle, Top):
    def m_bottom(self):
        print("bottom")


object = Bottom()
object.m_bottom()
object.m_middle()
object.m_top()
````

Riesci a vedere la differenza? Si nasconde in questa riga:

```python
class Bottom(Middle, Top):

```

In questo modo bizzarro abbiamo trasformato un codice estremamente semplice, caratterizzato da un chiaro percorso di ereditarietà singola, in un misterioso enigma di ereditarietà multipla. *"È valido?"*, potresti chiederti. Sì, lo è. *"Com'è possibile?"*, dovresti domandarti ora, e speriamo che tu senticcatamente il bisogno di porre questa domanda.

Come puoi notare, l'ordine in cui le due superclassi sono state elencate tra parentesi è coerente con la struttura del codice: la classe `Middle` precede la classe `Top`, proprio come nel percorso di ereditarietà reale.

Nonostante la sua stranezza, l'esempio è corretto e funziona come previsto, ma va precisato che questa notazione non apporta alcuna nuova funzionalità o significato aggiuntivo.

Modifichiamo il codice ancora una volta: adesso invertiamo i nomi delle due superclassi nella definizione della classe `Bottom`. Ecco come si presenta il frammento di codice ora:

```python
class Top:
    def m_top(self):
        print("top")


class Middle(Top):
    def m_middle(self):
        print("middle")


class Bottom(Top, Middle):
    def m_bottom(self):
        print("bottom")


object = Bottom()
object.m_bottom()
object.m_middle()
object.m_top()

```

Riesci a vedere il problema? Se esegui questo codice, otterrai un errore di tipo `TypeError` che dice: "Cannot create a consistent method resolution order (MRO) for bases Top, Middle".

### Il problema del diamante (The diamond problem)

Il secondo esempio dello spettro di problemi che possono scaturire dall'ereditarietà multipla è illustrato da un classico enigma della programmazione a oggetti, chiamato **problema del diamante** (o della losanga). Questo nome riflette la forma geometrica che assume il diagramma delle classi:

* In cima troviamo la superclasse radice, chiamata `A`;
* Subito sotto ci sono due sottoclassi che derivano direttamente da `A`: parliamo delle classi `B` e `C`;
* Infine, in fondo alla struttura, è presente un'ultima sottoclasse chiamata `D`, la quale eredita contemporaneamente da `B` e da `C` (o da `C` e `B`, poiché in Python l'ordine di queste due varianti cambia il comportamento del codice).

Riesci a visualizzare il rombo (o diamante) che si viene a creare nell'architettura?

Dai un'occhiata al codice nell'editor. Troverai questa identica struttura espressa nella sintassi di Python.

````python
class A:
    pass


class B(A):
    pass


class C(A):
    pass


class D(B, C):
    pass


d = D()
    
````

Alcuni linguaggi di programmazione non consentono affatto l'ereditarietà multipla e, di conseguenza, ti impediscono a monte di costruire una struttura a diamante — questa è la strada che Java e C# hanno scelto di seguire fin dalle loro origini.

Python, tuttavia, ha imboccato una via diversa: permette l'ereditarietà multipla e non si fa problemi se scrivi ed esegui un codice come quello nell'editor. Ricorda però che il MRO (Method Resolution Order) entra sempre in gioco ed è l'unico a governare le regole del gioco.

Ristrutturiamo il nostro esempio della pagina precedente per renderlo a tutti gli effetti un diagramma a diamante, proprio come vedi qui sotto:

````python
class Top:
    def m_top(self):
        print("top")


class Middle_Left(Top):
    def m_middle(self):
        print("middle_left")


class Middle_Right(Top):
    def m_middle(self):
        print("middle_right")


class Bottom(Middle_Left, Middle_Right):
	def m_bottom(self):
		print("bottom")


object = Bottom()
object.m_bottom()
object.m_middle()
object.m_top()
````

> **Nota:** entrambe le classi `Middle` definiscono un metodo con lo stesso nome: `m_middle()`.

Questo introduce una piccola incertezza nel nostro esempio, anche se siamo assolutamente certi che saprai rispondere a questa domanda fondamentale: quale dei due metodi `m_middle()` verrà effettivamente invocato quando viene eseguita la riga seguente?

```python
Object.m_middle()

```

In altre parole, cosa vedrai apparire sullo schermo: `middle_left` o `middle_right`?

Non c'è bisogno di correre — pensaci bene e tieni a mente il funzionamento del MRO di Python!

Sei pronto?

Sì, hai ragione. L'invocazione attiverà il metodo `m_middle()` che proviene dalla classe `Middle_Left`. La spiegazione è semplice: questa classe è elencata prima di `Middle_Right` nella lista di ereditarietà della classe `Bottom`. Se vuoi sincerarti che non ci siano dubbi al riguardo, prova a invertire queste due classi nell'elenco e verifica i risultati.

Se vuoi sperimentare dinamiche ancora più profonde legate all'ereditarietà multipla e alle "pietre preziose", prova a modificare il nostro frammento di codice inserendo un'ulteriore versione del metodo `m_middle()` all'interno della classe `Upper`, e analizza attentamente il suo comportamento.

Come puoi notare, i diamanti possono portare qualche problema nella tua vita — sia quelli veri, sia quelli offerti da Python.

# OOP: Eccezzioni e gestione degli errori

## Qualcosa in più sulle eccezioni

Discutere della programmazione a oggetti offre un'ottima opportunità per tornare a parlare delle eccezioni. La natura orientata agli oggetti delle eccezioni in Python le rende uno strumento estremamente flessibile, capace di adattarsi a esigenze specifiche, persino a quelle che ancora non conosci.

Prima di tuffarci nel lato prettamente orientato agli oggetti delle eccezioni, vogliamo mostrarti alcuni aspetti sintattici e semantici di come Python gestisce il blocco `try-except`, poiché offre qualcosa in più rispetto a quanto ti abbiamo presentato finora.

La prima caratteristica che vogliamo discutere qui è un ramo aggiuntivo che può essere inserito all'interno (o meglio, subito dopo) del blocco `try-except`: si tratta della porzione di codice che inizia con `else`, proprio come nell'esempio nell'editor.

````python
def reciprocal(n):
    try:
        n = 1 / n
    except ZeroDivisionError:
        print("Division failed")
        return None
    else:
        print("Everything went fine")
        return n


print(reciprocal(2))
print(reciprocal(0))
````

Una porzione di codice contrassegnata in questo modo viene eseguita quando (e solo quando) non è stata sollevata alcuna eccezione all'interno del blocco `try:`. Possiamo dire che dopo il `try:` viene eseguito esattamente un solo ramo: o quello che inizia con `except` (ricordando che possono esserci più rami di questo tipo) o quello che inizia con `else`.

> **Nota:** il ramo `else:` deve essere posizionato obbligatoriamente dopo l'ultimo ramo `except`.

Il blocco `try-except` può essere esteso in un ulteriore modo: aggiungendo una sezione introdotta dalla parola chiave `finally` (che deve essere tassativamente l'ultimo ramo del blocco dedicato alla gestione delle eccezioni).

> **Nota:** queste due varianti (`else` e `finally`) non dipendono in alcun modo l'una dall'altra: possono coesistere o presentarsi in modo del tutto indipendente.

Il blocco `finally` viene **sempre eseguito** (conclude l'esecuzione del blocco `try-except`, da cui il nome), a prescindere da ciò che è accaduto in precedenza — anche nel caso in cui venga sollevata un'eccezione, e sia che questa venga gestita o meno.


Dai un'occhiata al codice nell'editor per vedere come si posizionano i vari blocchi.

````python
def reciprocal(n):
    try:
        n = 1 / n
    except ZeroDivisionError:
        print("Division failed")
        n = None
    else:
        print("Everything went fine")
    finally:
        print("It's time to say goodbye")
        return n


print(reciprocal(2))
print(reciprocal(0))
````

## Le eccezioni sono classi

Tutti gli esempi precedenti si limitavano a rilevare un tipo specifico di eccezione e a rispondere in modo appropriato. Ora andremo più a fondo, guardando dentro l'eccezione stessa.

Probabilmente non ti sorprenderà sapere che **le eccezioni sono classi**. Inoltre, quando viene sollevata un'eccezione, viene istanziato un oggetto di quella classe; questo oggetto attraversa tutti i livelli di esecuzione del programma, alla ricerca di un ramo `except` preparato a gestirlo.

Un simile oggetto porta con sé informazioni utili che possono aiutarti a identificare con precisione tutti gli aspetti della situazione in corso. Per raggiungere questo obiettivo, Python offre una variante speciale della clausola di eccezione — puoi vederla nell'editor.

````python
try:
    i = int("Hello!")
except Exception as e:
    print(e)
    print(e.__str__())
    
````

Come puoi notare, l'istruzione `except` è stata estesa e contiene una dicitura aggiuntiva introdotta dalla parola chiave `as`, seguita da un identificatore. Questo identificatore ha lo scopo di catturare l'oggetto eccezione, permettendoti così di analizzarne la natura e trarre conclusioni utili.

> **Nota:** lo scope (l'ambito di visibilità) dell'identificatore copre esclusivamente il proprio ramo `except` e non si estende oltre.

L'esempio mostra un modo molto semplice di utilizzare l'oggetto ricevuto: stamparlo direttamente a schermo. Come puoi notare, l'output viene generato dal metodo `__str__()` dell'oggetto e contiene un breve messaggio che ne descrive la causa.

Lo stesso identico messaggio verrebbe stampato nel caso in cui non ci fosse un blocco `except` adeguato nel codice e Python fosse costretto a gestire l'eccezione da solo.


Tutte le eccezioni integrate in Python formano una gerarchia di classi. Non c'è alcun ostacolo all'estensione di questa gerarchia qualora lo ritenessi opportuno per i tuoi progetti.

Dai un'occhiata al codice nell'editor per vedere come iniziare.

````python
def print_exception_tree(thisclass, nest = 0):
    if nest > 1:
        print("   |" * (nest - 1), end="")
    if nest > 0:
        print("   +---", end="")

    print(thisclass.__name__)

    for subclass in thisclass.__subclasses__():
        print_exception_tree(subclass, nest + 1)


print_exception_tree(BaseException)
````

Questo programma mostra tutte le classi di eccezione predefinite sotto forma di una stampa ad albero.

Poiché l'albero è l'esempio perfetto di una struttura dati ricorsiva, la ricorsione sembra essere lo strumento migliore per esplorarlo. La funzione `print_exception_tree()` accetta due argomenti:

* Un punto all'interno dell'albero da cui iniziare l'esplorazione;
* Un livello di annidamento (che useremo per tracciare un disegno semplificato dei rami dell'albero).

Inizieremo dalla radice dell'albero: la radice di tutte le classi di eccezione in Python è la classe `BaseException` (la superclasse di qualsiasi altra eccezione).

Per ciascuna delle classi incontrate, viene eseguita la stessa serie di operazioni:

1. Viene stampato il suo nome, recuperato tramite la proprietà `__name__`;
2. Viene scorsa la lista delle sottoclassi restituita dal metodo `__subclasses__()`, invocando ricorsivamente la funzione `print_exception_tree()` e incrementando di conseguenza il livello di annidamento.

Nota come sono stati disegnati i rami e le ramificazioni. La stampa non è ordinata in alcun modo — se cerchi una sfida, puoi provare a ordinarla tu stesso. Inoltre, ci sono alcune lievi imprecisioni nel modo in cui vengono presentati alcuni rami. Se lo desideri, puoi correggere anche quelle.

## Anatomia dettagliata delle eccezioni

Diamo un'occhiata più da vicino all'oggetto eccezione, poiché contiene alcuni elementi davvero interessanti (torneremo presto su questo argomento quando esamineremo le tecniche di base dell'input/output in Python, dato che il loro sottosistema di eccezioni estende leggermente questi oggetti).

La classe `BaseException` introduce una proprietà chiamata `args`. Si tratta di una tupla progettata per raccogliere tutti gli argomenti passati al costruttore della classe. Risulta vuota se il costruttore è stato invocato senza argomenti, contiene un solo elemento quando al costruttore viene passato un singolo argomento (senza contare l'argomento `self`), e così via.

Abbiamo preparato una semplice funzione per stampare la proprietà `args` in modo elegante. Puoi vedere la funzione all'interno dell'editor.

````python
def print_args(args):
    lng = len(args)
    if lng == 0:
        print("")
    elif lng == 1:
        print(args[0])
    else:
        print(str(args))


try:
    raise Exception
except Exception as e:
    print(e, e.__str__(), sep=' : ' ,end=' : ')
    print_args(e.args)

try:
    raise Exception("my exception")
except Exception as e:
    print(e, e.__str__(), sep=' : ', end=' : ')
    print_args(e.args)

try:
    raise Exception("my", "exception")
except Exception as e:
    print(e, e.__str__(), sep=' : ', end=' : ')
    print_args(e.args)
````

Abbiamo utilizzato questa funzione per stampare il contenuto della proprietà `args` in tre casi differenti, nei quali un'eccezione della classe `Exception` viene sollevata in tre modi diversi. Per rendere il tutto più evidente, abbiamo stampato anche l'oggetto stesso, insieme al risultato dell'invocazione del metodo `__str__()`.

Il primo caso rientra nella normale routine: c'è semplicemente il nome `Exception` dopo la parola chiave `raise`. Ciò significa che l'oggetto di questa classe è stato creato nel modo più comune possibile.

Il secondo e il terzo caso potrebbero sembrare un po' insoliti a prima vista, ma non c'è nulla di strano: si tratta semplicemente di invocazioni esplicite del costruttore. Nella seconda istruzione `raise`, il costruttore viene invocato con un solo argomento, mentre nella terza con due.

Come puoi notare, l'output del programma riflette questa struttura, mostrando il contenuto corretto della proprietà `args`.

## Come creare le proprie eccezioni personalizzate

La gerarchia delle eccezioni non è chiusa né definitiva: puoi sempre estenderla se desideri o hai la necessità di creare un ecosistema popolato dalle tue eccezioni personali.

Questa pratica si rivela molto utile quando sviluppi un modulo complesso che deve rilevare errori specifici e sollevare eccezioni, e vuoi che tali eccezioni siano facilmente distinguibili da quelle standard fornite da Python.

Per farlo, basta definire le nuove eccezioni come sottoclassi derivate da quelle predefinite.

> **Nota:** se vuoi creare un'eccezione che sia una specializzazione di un'eccezione integrata esistente, falla derivare direttamente da quella. Se invece vuoi costruire una tua gerarchia indipendente, senza che sia strettamente legata a un errore specifico di Python, falla derivare da una delle classi di eccezione di livello superiore, come `Exception`.

Immagina di aver creato un'aritmetica completamente nuova, governata dalle tue leggi e dai tuoi teoremi. È evidente che anche la divisione è stata ridefinita e deve comportarsi in modo diverso rispetto a quella classica. Di conseguenza, questa nuova divisione dovrebbe sollevare un'eccezione personalizzata, distinta dal comune `ZeroDivisionError` integrato; tuttavia, è ragionevole ipotizzare che in alcune circostanze tu (o l'utente della tua aritmetica) voglia gestire tutte le divisioni per zero allo stesso modo.

Esigenze di questo tipo possono essere soddisfatte nel modo mostrato nell'editor.

````python
class MyZeroDivisionError(ZeroDivisionError):	
    pass


def do_the_division(mine):
    if mine:
        raise MyZeroDivisionError("some worse news")
    else:		
        raise ZeroDivisionError("some bad news")


for mode in [False, True]:
    try:
        do_the_division(mode)
    except ZeroDivisionError:
        print('Division by zero')

for mode in [False, True]:
    try:
        do_the_division(mode)
    except MyZeroDivisionError:
        print('My division by zero')
    except ZeroDivisionError:
        print('Original division by zero')
````

Guarda il codice e analizziamolo insieme:

Abbiamo definito la nostra eccezione personalizzata, chiamata `MyZeroDivisionError`, derivandola direttamente dalla classe integrata `ZeroDivisionError`. Come puoi notare, abbiamo scelto di non aggiungere alcun nuovo componente all'interno della classe.

Di conseguenza, un'eccezione di questo tipo può essere trattata — a seconda del punto di vista desiderato — come un comune `ZeroDivisionError`, oppure può essere considerata in modo separato.

La funzione `do_the_division()` solleva un'eccezione `MyZeroDivisionError` o `ZeroDivisionError`, in base al valore passato come argomento.

La funzione viene invocata in totale quattro volte: le prime due invocazioni vengono gestite utilizzando un solo ramo `except` (quello più generale), mentre le ultime due vengono intercettate da due rami diversi, in grado di distinguere le eccezioni (non dimenticare: l'ordine dei rami fa una differenza fondamentale!).


Quando ti appresti a costruire un universo completamente nuovo, popolato da creature inedite che non hanno nulla in comune con le cose a noi familiari, potresti voler strutturare una tua gerarchia di eccezioni dedicata.

Ad esempio, se stai lavorando a un grande sistema di simulazione progettato per modellare le attività di una pizzeria, può essere una buona idea creare una gerarchia di eccezioni separata.

Puoi iniziare a costruirla definendo un'eccezione generale da usare come nuova classe base per qualsiasi altra eccezione più specifica. Noi l'abbiamo impostata in questo modo:

```python
class PizzaError(Exception):
    def __init__(self, pizza, message):
        Exception.__init__(self, message)
        self.pizza = pizza

```

> **Nota:** in questo blocco vogliamo raccogliere informazioni più dettagliate rispetto a una normale `Exception`. Per questo motivo, il nostro costruttore accetta due argomenti: uno che specifica la pizza oggetto del processo e uno che contiene una descrizione più o meno precisa del problema.

Come puoi vedere, passiamo il secondo parametro al costruttore della superclasse e salviamo il primo all'interno di una nostra proprietà.

Un problema ancora più specifico (come un eccesso di formaggio) può richiedere un'eccezione altrettanto specifica. È possibile far derivare la nuova classe da quella appena definita, ovvero `PizzaError`, proprio come abbiamo fatto qui:

```python
class TooMuchCheeseError(PizzaError):
    def __init__(self, pizza, cheese, message):
        PizzaError.__init__(self, pizza, message)
        self.cheese = cheese

```

L'eccezione `TooMuchCheeseError` necessita di una quantità maggiore di informazioni rispetto alla generica `PizzaError`, quindi le aggiungiamo all'interno del costruttore; il parametro `cheese` viene poi memorizzato per essere elaborato successivamente.


Dai un'occhiata al codice nell'editor. Abbiamo unito le due eccezioni definite in precedenza e le abbiamo messe all'opera all'interno di un piccolo frammento di codice dimostrativo.

````python
class PizzaError(Exception):
    def __init__(self, pizza, message):
        Exception.__init__(self, message)
        self.pizza = pizza


class TooMuchCheeseError(PizzaError):
    def __init__(self, pizza, cheese, message):
        PizzaError.__init__(self, pizza, message)
        self.cheese = cheese


def make_pizza(pizza, cheese):
    if pizza not in ['margherita', 'capricciosa', 'calzone']:
        raise PizzaError(pizza, "no such pizza on the menu")
    if cheese > 100:
        raise TooMuchCheeseError(pizza, cheese, "too much cheese")
    print("Pizza ready!")

for (pz, ch) in [('calzone', 0), ('margherita', 110), ('mafia', 20)]:
    try:
        make_pizza(pz, ch)
    except TooMuchCheeseError as tmce:
        print(tmce, ':', tmce.cheese)
    except PizzaError as pe:
        print(pe, ':', pe.pizza)
        
````

Una di queste eccezioni viene sollevata all'interno della funzione `make_pizza()` quando viene rilevata una delle due seguenti situazioni di errore: una richiesta di pizza errata oppure una richiesta con troppo formaggio.

> **Nota:**
> * Rimuovere il ramo che inizia con `except TooMuchCheeseError` farà sì che tutte le eccezioni sollevate vengano classificate e gestite genericamente come `PizzaError`;
> * Rimuovere il ramo che inizia con `except PizzaError` farà sì che le eccezioni `TooMuchCheeseError` rimangano non gestite, provocando l'interruzione del programma.
> 
> 

La soluzione precedente, sebbene elegante ed efficiente, presenta un punto debole importante. A causa del modo un po' troppo rigido in cui sono stati dichiarati i costruttori, le nuove eccezioni non possono essere utilizzate così come sono senza fornire l'elenco completo degli argomenti richiesti.

Risolviamo questo problema impostando dei valori predefiniti (*default values*) per tutti i parametri del costruttore. Dai un'occhiata:

```python
class PizzaError(Exception):
    def __init__(self, pizza='unknown', message=''):
        Exception.__init__(self, message)
        self.pizza = pizza
 
 
class TooMuchCheeseError(PizzaError):

        PizzaError.__init__(self, pizza, message)
```
