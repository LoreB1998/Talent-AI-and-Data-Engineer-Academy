# Il metodo `capitalize()`

Iniziamo ad analizzare alcuni metodi standard per le stringhe in Python. Li passeremo in rassegna in ordine alfabetico.

Il metodo `capitalize()` fa esattamente quello che suggerisce il nome: crea una **nuova stringa** a partire da quella originale, modificando i caratteri nel seguente modo:

* Se il primo carattere della stringa è una lettera (ricorda: il primo carattere è l'elemento con **indice 0**, non semplicemente il primo carattere visibile), questo viene convertito in **maiuscolo**.
* Tutte le restanti lettere della stringa vengono convertite in **minuscolo**.

Ci sono due regole fondamentali da non dimenticare:

> 1. **La stringa originale non viene modificata in alcun modo** – l'immutabilità delle stringhe in Python va rispettata senza eccezioni.
> 2. **Il metodo restituisce la stringa modificata come risultato** – se non la utilizzi (assegnandola a una variabile o passandola a una funzione), svanirà senza lasciare traccia.
> 
> 

**Nota:** I metodi non devono essere invocati necessariamente solo tramite variabili. Possono essere applicati direttamente anche alle stringhe letterali (es. `"test".capitalize()`). Useremo spesso questa convenzione per rendere gli esempi più chiari e lineari, senza aggiungere assegnazioni superflue.

Dai un'occhiata all'esempio nell'editor ed eseguilo.

````python
print("Alpha".capitalize())
print('ALPHA'.capitalize())
print(' Alpha'.capitalize())
print('123'.capitalize())
print("αβγδ".capitalize())
````

# Il metodo `center()`

La variante a un singolo parametro del metodo `center()` crea una copia della stringa originale cercando di centrarla all'interno di uno spazio della larghezza specificata.

La centratura viene eseguita in modo molto semplice, aggiungendo degli spazi vuoti (caratteri di spaziatura) prima e dopo la stringa originale. Non aspettarti dinamiche complesse da questo metodo: il suo funzionamento è elementare.

L'esempio nell'editor utilizza delle parentesi quadre proprio per mostrarti con chiarezza dove inizia e dove finisce la stringa centrata.

````python
# Demonstrating the center() method:
print('[' + 'alpha'.center(9) + ']')
print('[' + 'Beta'.center(2) + ']')
print('[' + 'Beta'.center(4) + ']')
print('[' + 'Beta'.center(6) + ']')
````

La variante a due parametri di `center()` utilizza il carattere passato come secondo argomento al posto del classico spazio vuoto.

Analizza l'esempio seguente:

```python
print('[' + 'alpha'.center(10, '*') + ']')

```

In questo caso, il metodo riempirà lo spazio rimanente a destra e a sinistra con il carattere asterisco `*`, fino a raggiungere la lunghezza totale di 10 caratteri.

````python
print('[' + 'alpha'.center(10, '*') + ']')
````

# Il metodo `endswith()`

Il metodo `endswith()` controlla se una stringa termina con la sottostringa specificata come argomento. Restituisce `True` o `False` a seconda del risultato della verifica.

> **Nota:** La sottostringa deve combaciare esattamente con la parte finale della stringa principale, toccando l'ultimo carattere — non basta che si trovi genericamente "nei pressi" della fine.

Dai un'occhiata all'esempio nell'editor, analizzalo ed eseguilo.

# Il metodo `find()`

Il metodo `find()` è simile a `index()`, che già conosci: cerca una sottostringa e restituisce l'indice della sua prima occorrenza. Tuttavia, presenta due differenze fondamentali:

* **È più sicuro:** non genera un errore se la sottostringa cercata non esiste, ma restituisce semplicemente `-1`.
* **Funziona solo con le stringhe:** non provare ad applicarlo ad altre sequenze (come le liste).

Dai un'occhiata al codice nell'editor per vedere come utilizzarlo.

````python
# Demonstrating the find() method:
print("Eta".find("ta"))
print("Eta".find("mma"))
````

> **Nota:** Non usare `find()` se il tuo unico obiettivo è verificare se un singolo carattere è presente in una stringa — l'operatore `in` è decisamente più veloce e leggibile.

Ecco un altro esempio:

````python
t = 'theta'
print(t.find('eta'))
print(t.find('et'))
print(t.find('the'))
print(t.find('ha'))
````

Riesci a prevedere l'output? Esegui il codice per verificare le tue ipotesi.

Se desideri effettuare la ricerca non dall'inizio della stringa, ma a partire da una posizione specifica, puoi utilizzare la variante a due parametri del metodo `find()`.

Guarda questo esempio:

```python
print('kappa'.find('a', 2))

```

Il secondo argomento specifica l'indice dal quale avviare la ricerca (non è obbligatorio che rientri necessariamente nei limiti della lunghezza della stringa).

In questo caso specifico, tra le due lettere 'a' presenti in `'kappa'`, verrà trovata solo la seconda (quella all'indice 4), poiché la ricerca parte dall'indice 2 (la prima 'p'). Prova a eseguire lo snippet per verificare.

Inoltre, puoi sfruttare il metodo `find()` all'interno di un ciclo per scovare tutte le occorrenze di una sottostringa, come in questo esempio:

````python
the_text = """A variation of the ordinary lorem ipsum
text has been used in typesetting since the 1960s 
or earlier, when it was popularized by advertisements 
for Letraset transfer sheets. It was introduced to 
the Information Age in the mid-1980s by the Aldus Corporation, 
which employed it in graphics and word-processing templates
for its desktop publishing program PageMaker (from Wikipedia)"""

fnd = the_text.find('the')
while fnd != -1:
    print(fnd)
    fnd = the_text.find('the', fnd + 1)
    
````

Esiste anche una variante a tre parametri del metodo find(): il terzo argomento indica il primo indice che non verrà preso in considerazione durante la ricerca (rappresenta, di fatto, il limite superiore, che è escluso dal controllo).

Guarda il nostro esempio qui sotto:

````python
print('kappa'.find('a', 1, 4))
print('kappa'.find('a', 2, 4))
````

# Il metodo `isalnum()`

Il metodo senza parametri `isalnum()` verifica se la stringa contiene **esclusivamente cifre numeriche o caratteri alfabetici** (lettere), e restituisce `True` o `False` a seconda del risultato.

Se nella stringa sono presenti spazi, segni di punteggiatura o caratteri speciali, il metodo restituirà `False`.

Dai un'occhiata all'esempio nell'editor ed eseguilo.

````python
# Demonstrating the isalnum() method:
print('lambda30'.isalnum())
print('lambda'.isalnum())
print('30'.isalnum())
print('@'.isalnum())
print('lambda_30'.isalnum())
print(''.isalnum())
````

# Il metodo `isalpha()`

Il metodo `isalpha()` è più specifico rispetto a `isalnum()`: è interessato **esclusivamente alle lettere**.

Se la stringa contiene cifre numeriche, spazi, segni di punteggiatura o caratteri speciali, il metodo restituirà `False`. Restituisce `True` solo se ogni singolo carattere della stringa è una lettera dell'alfabeto.

Dai un'occhiata all'esempio nell'editor:

````python
# Example 1: Demonstrating the isapha() method:
print("Moooo".isalpha())
print('Mu40'.isalpha())
````

````python
# Example 2: Demonstrating the isdigit() method:
print('2018'.isdigit())
print("Year2019".isdigit())
````

### Il metodo `islower()`

Il metodo `islower()` è una variante più selettiva di `isalpha()`: accetta **esclusivamente lettere minuscole**.

Se la stringa contiene almeno una lettera maiuscola, il metodo restituirà `False`.

> **Nota importante:** I caratteri non alfabetici (come cifre numeriche, spazi o segni di punteggiatura) vengono semplicemente ignorati dal controllo. Questo significa che se la stringa contiene sia lettere minuscole sia numeri o spazi (ma nessuna lettera maiuscola), il metodo restituirà comunque `True`. Se invece non contiene alcuna lettera (es. solo numeri), restituirà `False`.

Dai un'occhiata all'esempio nell'editor:

````python
# Example: Demonstrating the islower() method:
print("Moooo".islower())
print('moooo'.islower())
    
````

# Il metodo `isspace()`

Il metodo `isspace()` identifica **esclusivamente gli spazi bianchi** (i cosiddetti *whitespaces*). Qualsiasi altro tipo di carattere presente all'interno della stringa farà sì che il metodo restituisca `False`.

Oltre al classico spazio vuoto che inserisci con la barra spaziatrice, questo metodo riconosce come spazi bianchi anche altri caratteri di formattazione e controllo, tra cui:

* Il carattere di tabulazione (`\t`)
* Il carattere di nuova riga o a capo (`\n`)
* Il ritorno a capo (`\r`)

Se la stringa è vuota (ossia non contiene alcun carattere, nemmeno uno spazio: `""`), il metodo restituirà comunque `False`.

Dai un'occhiata all'esempio nell'editor:

````python
# Example: Demonstrating the isspace() method:
print(' \n '.isspace())
print(" ".isspace())
print("mooo mooo mooo".isspace())
````

# Il metodo `isupper()`

Il metodo `isupper()` è la controparte speculare di `islower()` e si concentra **esclusivamente sulle lettere maiuscole**.

Analogamente a quanto visto per il minuscolo, il metodo restituisce `True` solo se la stringa contiene almeno un carattere alfabetico e tutte le lettere presenti sono maiuscole. I caratteri non alfabetici (cifre numeriche, spazi, simboli) vengono ignorati dal controllo. Se la stringa non contiene alcuna lettera (ad esempio, è composta solo da numeri), il risultato sarà `False`.

Dai un'occhiata al codice nell'editor:

````python
# Example: Demonstrating the isupper() method:
print("Moooo".isupper())
print('moooo'.isupper())
print('MOOOO'.isupper())
````

### Il metodo `join()`

Il metodo `join()` è leggermente più articolato rispetto a quelli visti finora, quindi analizziamolo insieme passo dopo passo per capire esattamente come funziona:

* Come suggerisce il nome, il metodo esegue un'unione (*join*). Si aspetta un singolo argomento, che deve essere un'**iterable** (tipicamente una **lista**).
* È fondamentale che **tutti gli elementi all'interno della lista siano stringhe**. Se anche un solo elemento è di tipo diverso (come un intero o un float), il metodo solleverà un'eccezione di tipo `TypeError`.
* Tutti gli elementi della lista vengono uniti in un'unica grande stringa, ma...
* ...la stringa *da cui il metodo viene invocato* agisce come **separatore**, venendo inserita tra un elemento e l'altro.
* Infine, la nuova stringa così creata viene restituita come risultato.

Per comprendere meglio la logica, puoi pensare alla sintassi in questo modo:

$$\texttt{separatore}.\texttt{join}(\texttt{lista\_di\_stringhe})$$

Dai un'occhiata all'esempio nell'editor per vedere l'effetto pratico:

````python
# Demonstrating the join() method:
print(",".join(["omicron", "pi", "rho"]))
````

Analizziamo in dettaglio cosa succede quando esegui questo comando:

* **Il separatore:** Il metodo `join()` viene invocato direttamente dalla stringa `","`, che contiene una virgola. Questa stringa di attivazione funge da collante. Può essere lunga a piacere (ad esempio `" - "`) oppure completamente vuota (`""`), a seconda di come desideri unire gli elementi.
* **L'argomento:** L'argomento passato a `join()` è una lista che contiene tre stringhe distinte: `"omicron"`, `"pi"` e `"rho"`.
* **Il risultato:** Il metodo prende i tre elementi della lista e li unisce in sequenza, inserendo la virgola tra ognuno di essi.

# Il metodo `lower()`

Il metodo `lower()` crea una copia della stringa di origine, sostituisce tutte le lettere maiuscole con le rispettive minuscole e restituisce la nuova stringa così ottenuta.

È importante ricordare che, poiché le stringhe in Python sono immutabili, **la stringa originale rimane del tutto inalterata**.

Se la stringa di partenza non contiene alcuna lettera maiuscola (o è composta solo da numeri, spazi e punteggiatura), il metodo restituirà semplicemente una copia identica alla stringa originale, senza sollevare alcun errore.

> **Nota:** Il metodo `lower()` non accetta alcun parametro.

Dai un'occhiata all'esempio nell'editor per vedere come si comporta:

````python
# Demonstrating the lower() method:
print("SiGmA=60".lower())
````

# Il metodo `lstrip()`

La versione senza parametri del metodo `lstrip()` (la "l" sta per *left*, sinistra) restituisce una nuova stringa creata a partire da quella originale, rimuovendo tutti gli spazi bianchi iniziali (*leading whitespaces*), ovvero quelli posizionati all'estrema sinistra della stringa.

Anche in questo caso, la stringa di partenza non viene modificata, ma viene generata una copia pulita. I caratteri di spazio bianco rimossi includono non solo lo spazio classico, ma anche le tabulazioni (`\t`) e i caratteri di nuova riga (`\n`) che si trovano all'inizio del testo. Gli spazi posizionati alla fine della stringa o all'interno delle parole non vengono toccati.

Analizza il codice di esempio nell'editor per vedere come funziona.

````python
# Demonstrating the lstrip() method:
print("[" + " tau ".lstrip() + "]")
````

La versione a un parametro del metodo `lstrip()` si comporta in modo simile a quella senza parametri, ma invece di rimuovere gli spazi bianchi, **rimuove tutti i caratteri specificati all'interno della stringa passata come argomento**.

La ricerca e la rimozione avvengono a partire dall'estrema sinistra della stringa originale e si interrompono non appena viene incontrato un carattere che **non** è presente nell'argomento.

Guarda questo esempio per capire la logica:

````python
print("www.cisco.com".lstrip("w."))
````

### Come funziona nel dettaglio:

L'argomento contiene l'insieme dei caratteri da eliminare (in questo caso, la lettera `w` e il punto `.`). L'ordine con cui inserisci i caratteri nell'argomento non ha importanza: `"w."` o `".w"` produce lo stesso identico risultato.

Python esamina la stringa `"www.cisco.com"` da sinistra verso destra:

1. Incontra la prima `w` $\rightarrow$ è nell'argomento? Sì, viene rimossa.
2. Incontra la seconda `w` $\rightarrow$ è nell'argomento? Sì, viene rimossa.
3. Incontra la terza `w` $\rightarrow$ è nell'argomento? Sì, viene rimossa.
4. Incontra il punto `.` $\rightarrow$ è nell'argomento? Sì, viene rimosso.
5. Incontra la lettera `c` $\rightarrow$ è nell'argomento? **No**.

A questo punto il processo si blocca immediatamente. Anche se alla fine della stringa sono presenti altri punti (come in `.com`), questi non verranno toccati perché la rimozione avviene esclusivamente sul lato sinistro (*left strip*) e si è già arrestata.

L'output di questo snippet sarà:

```text
cisco.com

```

# Il metodo `replace()`

Il metodo `replace()` a due parametri restituisce una copia della stringa originale in cui **tutte le occorrenze** della sottostringa specificata nel primo argomento vengono sostituite con la stringa specificata nel secondo argomento.

Come per gli altri metodi delle stringhe, l'originale rimane intatto e viene generata una nuova stringa. Se la sottostringa da cercare non viene trovata, il metodo restituisce semplicemente una copia della stringa iniziale, senza modifiche e senza lanciare errori.

Dai un'occhiata al codice di esempio nell'editor ed eseguilo:

```python
print("www.netacad.com".replace("netacad", "cisco"))

```

* **Primo argomento (`"netacad"`):** È l'elemento che Python cercherà all'interno del testo.
* **Secondo argomento (`"cisco"`):** È il testo che prenderà il posto di ogni occorrenza trovata.

L'output generato sarà:

```text
www.cisco.com

```

> **Nota:** Se il secondo argomento è una stringa vuota (`""`), l'effetto pratico sarà la completa **rimozione** di tutte le occorrenze della prima stringa.

````python
# Demonstrating the replace() method:
print("www.netacad.com".replace("netacad.com", "pythoninstitute.org"))
print("This is it!".replace("is", "are"))
print("Apple juice".replace("juice", ""))
````

Se il secondo argomento è una stringa vuota, la sostituzione si traduce di fatto nella rimozione della stringa del primo argomento. Che tipo di "magia" avviene, invece, se è il primo argomento a essere una stringa vuota?

La variante a tre parametri del metodo `replace()` utilizza invece il terzo argomento (un numero intero) per **limitare il numero massimo di sostituzioni** da effettuare, procedendo da sinistra verso destra.

Guarda il codice di esempio modificato qui sotto:

````python
print("This is it!".replace("is", "are", 1))
print("This is it!".replace("is", "are", 2))
````

# Il metodo `rfind()`

Le versioni a uno, due e tre parametri del metodo `rfind()` svolgono quasi le stesse identiche funzioni delle loro controparti prive del prefisso "r", ma **avviano la ricerca a partire dalla fine della stringa** anziché dall'inizio (da qui il prefisso "r", che sta per *right*, destra).

Un dettaglio fondamentale da ricordare è che, sebbene la ricerca proceda da destra verso sinistra, l'indice restituito in caso di successo è sempre calcolato **dall'inizio della stringa** (indice basato su zero), esattamente come nel metodo `find()` standard. Se l'elemento cercato non viene trovato, il metodo restituisce comunque `-1`.

Dai un'occhiata al codice di esempio nell'editor e prova a indovinare il suo output. Esegui poi il codice per verificare se avevi ragione.

````python
# Demonstrating the rfind() method:
print("tau tau tau".rfind("ta"))
print("tau tau tau".rfind("ta", 9))
print("tau tau tau".rfind("ta", 3, 9))
````

# Il metodo `rstrip()`

Le due varianti del metodo `rstrip()` (dove la "r" sta per *right*, destra) svolgono quasi la stessa identica funzione dei metodi `lstrip()`, ma agiscono sul **lato opposto della stringa**, ovvero la fine (destra).

* **La versione senza parametri:** Rimuove tutti gli spazi bianchi finali (*trailing whitespaces*), come spazi vuoti, tabulazioni (`\t`) o caratteri di nuova riga (`\n`) posizionati all'estrema destra del testo.
* **La versione a un parametro:** Rimuove dall'estrema destra tutti i caratteri specificati nella stringa passata come argomento, procedendo verso sinistra e interrompendosi non appena incontra un carattere non presente nell'elenco.

Dai un'occhiata all'esempio di codice nell'editor. Riesci a indovinare il suo output? Esegui il codice per verificare le tue ipotesi.

````python
# Demonstrating the rstrip() method:
print("[" + " upsilon ".rstrip() + "]")
print("cisco.com".rstrip(".com"))
````

# Il metodo `split()`

Il metodo `split()` fa esattamente quello che dice il nome: divide (*split*) la stringa e costruisce una lista di tutte le sottostringhe individuate.

Nella sua versione senza parametri, il metodo assume che le sottostringhe siano **delimitate da spazi bianchi** (spazi, tabulazioni `\t` o caratteri di nuova riga `\n`). Questi spazi bianchi fungono da separatori: non prendono parte attiva all'operazione e non vengono copiati all'interno della lista risultante. Inoltre, sequenze di più spazi bianchi consecutivi vengono trattate come un unico separatore.

Se la stringa è completamente vuota, anche la lista risultante sarà vuota.

Dai un'occhiata al codice nell'editor:

````python
# Demonstrating the split() method:
print("phi       chi\npsi".split())
````

# Il metodo `strip()`

Come per gli altri metodi di *stripping*, gli spazi o i caratteri bianchi situati all'interno della stringa (tra le parole) non vengono minimamente toccati. Esiste anche per questo metodo la variante a un parametro, che permette di specificare una stringa contenente l'insieme di caratteri personalizzati da eliminare contemporaneamente da entrambe le estremità.

Dai un'occhiata al secondo esempio nell'editor:

````python
# Demonstrating the strip() method:
print("[" + "   aleph   ".strip() + "]")
````
