# Come i computer comprendono i singoli caratteri

Hai scritto alcuni programmi interessanti da quando hai iniziato questo corso, ma tutti hanno elaborato un solo tipo di dati: i numeri. Come sai (puoi vederlo ovunque intorno a te), moltissimi dati informatici non sono numeri: nomi, cognomi, indirizzi, titoli, poesie, articoli scientifici, email, sentenze giudiziarie, dichiarazioni d'amore e molto, molto altro ancora.

Tutti questi dati devono essere memorizzati, inseriti, visualizzati, cercati e trasformati dai computer contemporanei proprio come qualsiasi altro dato, non importa se si tratta di singoli caratteri o di enciclopedie in più volumi.

**Com'è possibile?**

Come si può fare in Python? Questo è ciò di cui discuteremo ora. Cominciamo da come i computer comprendono i singoli caratteri.

I computer memorizzano i caratteri sotto forma di **numeri**. Ogni carattere utilizzato da un computer corrisponde a un numero unico, e viceversa. Questa associazione deve includere più caratteri di quanti ci si potrebbe aspettare. Molti di essi sono invisibili agli esseri umani, ma essenziali per i computer.

Alcuni di questi caratteri sono chiamati *whitespaces* (spazi vuoti), mentre altri sono denominati *control characters* (caratteri di controllo), perché il loro scopo è controllare i dispositivi di input/output.

Un esempio di spazio vuoto completamente invisibile a occhio nudo è un codice speciale, o una coppia di codici (diversi sistemi operativi possono gestire la questione in modo differente), utilizzati per contrassegnare la fine delle linee all'interno dei file di testo.

Le persone non vedono questo segno (o questi segni), ma sono in grado di osservare l'effetto della loro applicazione nel punto in cui le linee si interrompono.

Possiamo creare virtualmente un numero infinito di associazioni carattere-numero, ma la vita in un mondo in cui ogni tipo di computer utilizzasse una codifica dei caratteri diversa non sarebbe molto comoda. Questo sistema ha portato alla necessità di introdurre uno standard universale e ampiamente accettato, implementato da (quasi) tutti i computer e sistemi operativi del mondo.

Quello denominato **ASCII** (acronimo di *American Standard Code for Information Interchange*) è il più ampiamente utilizzato, e si può presumere che quasi tutti i dispositivi moderni (come computer, stampanti, telefoni cellulari, tablet, ecc.) utilizzino tale codice.

Il codice offre spazio per **256 caratteri diversi**, ma a noi interessano solo i primi 128. Se vuoi vedere come è strutturato il codice, guarda la tabella qui sotto. Clicca sulla tabella per ingrandirla. Guardala attentamente: ci sono alcuni fatti interessanti. Guarda il codice del carattere più comune – lo *spazio*. Questo è **32**.

Ora controlla il codice della lettera minuscola *a*. Questo è **97**. E ora trova la *A* maiuscola. Il suo codice è **65**. Ora calcola la differenza tra il codice di *a* e *A*. È uguale a **32**. È proprio il codice dello spazio. Interessante, vero?

Nota anche che le lettere sono disposte nello stesso ordine dell'alfabeto latino.

# I18N

Ovviamente, l'alfabeto latino non è sufficiente per l'intera umanità. Gli utenti di questo alfabeto sono in minoranza. Era necessario ideare qualcosa di più flessibile e capiente dell'ASCII, qualcosa in grado di rendere tutto il software del mondo adatto all'**internazionalizzazione**, perché lingue diverse utilizzano alfabeti completamente differenti, e a volte questi alfabeti non sono semplici come quello latino.

La parola *internationalization* viene comunemente abbreviata in **I18N**.

Perché? Guarda attentamente: c'è una **I** all'inizio della parola, poi ci sono **18** lettere diverse e una **N** alla fine.

Nonostante l'origine leggermente umoristica, il termine è utilizzato ufficialmente in molti documenti e standard.

L'I18N del software è uno standard nei tempi attuali. Ogni programma deve essere scritto in modo da poter essere utilizzato in tutto il mondo, tra culture, lingue e alfabeti differenti.

Una forma classica del codice ASCII utilizza **otto bit** per ogni segno. Otto bit significano **256 caratteri diversi**. I primi 128 sono utilizzati per l'alfabeto latino standard (sia caratteri maiuscoli che minuscoli). È possibile inserire tutti gli altri caratteri nazionali utilizzati nel mondo nelle restanti 128 posizioni?

**No. Non lo è.**

# Code points e code pages

Ora abbiamo bisogno di un nuovo termine: un **code point** (punto di codice).

Un *code point* è un numero che dà forma a un carattere. Per esempio, **32** è il *code point* che produce uno spazio nella codifica ASCII. Possiamo dire che il codice ASCII standard è composto da **128 code points**.

Poiché l'ASCII standard occupa 128 dei 256 *code points* possibili, si può usufruire solo dei restanti 128.

Non è abbastanza per tutte le lingue possibili, ma può essere sufficiente per una sola lingua o per un piccolo gruppo di lingue simili.

È possibile impostare la metà superiore dei *code points* in modo diverso per lingue differenti? Sì, è possibile. Un concetto del genere si chiama **code page** (pagina di codice).

Una *code page* è uno standard per l'utilizzo dei 128 *code points* superiori al fine di memorizzare specifici caratteri nazionali. Per esempio, esistono diverse *code pages* per l'Europa occidentale e l'Europa orientale, per gli alfabeti cirillico e greco, per le lingue araba ed ebraica, e così via.

Ciò significa che lo stesso identico *code point* può dare forma a caratteri diversi quando viene utilizzato in *code pages* differenti.

Per esempio, il *code point* **200** produce la **Č** (una lettera usata da alcune lingue slave) quando viene utilizzato dalla *code page* **ISO/IEC 8859-2**, e produce la **Ш** (una lettera cirillica) quando viene utilizzato dalla *code page* **ISO/IEC 8859-5**.

Di conseguenza, per determinare il significato di uno specifico *code point*, è necessario conoscere la *code page* di destinazione.

In altre parole, i *code points* derivati dal concetto di *code page* sono **ambigui**.

## Unicode

Le *code pages* hanno aiutato l'industria informatica a risolvere i problemi di I18N per un certo periodo, ma ben presto è emerso che non sarebbero state una soluzione permanente.

Il concetto che ha risolto il problema a lungo termine è stato **Unicode**.

Unicode assegna caratteri univoci (non ambigui) come lettere, trattini, ideogrammi, ecc. a **più di un milione** di *code points*.

I primi 128 *code points* di Unicode sono identici all'ASCII, e i primi 256 *code points* di Unicode sono identici alla *code page* **ISO/IEC 8859-1** (una *code page* progettata per le lingue dell'Europa occidentale).


## UCS-4

Lo standard Unicode non dice nulla su come codificare e memorizzare i caratteri nella memoria e nei file. Si limita a nominare tutti os caratteri disponibili e ad assegnarli a dei *planes* (piani: un gruppo di caratteri di origine, applicazione o natura simile).

Esiste più di uno standard che descrive le tecniche utilizzate per implementare Unicode nei computer reali e nei sistemi di archiviazione informatica. Il più generale di questi è **UCS-4**.

Il nome deriva da *Universal Character Set*.

UCS-4 utilizza **32 bit** (quattro byte) per memorizzare ogni carattere, e il codice corrisponde semplicemente al numero univoco del *code point* di Unicode. Un file contenente testo codificato in UCS-4 può iniziare con un **BOM** (*byte order mark*), una combinazione di bit non stampabile che annuncia la natura del contenuto del file. Alcune utility potrebbero richiederlo.

Come si può notare, l'UCS-4 è uno standard piuttosto dispendioso: quadruplica le dimensioni di un testo rispetto all'ASCII standard. Fortunatamente, esistono forme più intelligenti per codificare i testi in Unicode.

## UTF-8

Uno dei più comunemente utilizzati è **UTF-8**.

Il nome deriva da *Unicode Transformation Format*.

Il concetto è molto intelligente. UTF-8 utilizza per ciascuno dei *code points* solo il numero di bit strettamente necessario per rappresentarli.

Per esempio:

* Tutti i caratteri latini (e tutti i caratteri ASCII standard) occupano **otto bit**;
* I caratteri non latini occupano **16 bit**;
* Gli ideogrammi **CJK** (Cina-Giappone-Corea) occupano **24 bit**.

A causa delle caratteristiche del metodo utilizzato da UTF-8 per memorizzare i *code points*, non vi è alcuna necessità di utilizzare il BOM, ma alcuni strumenti lo cercano durante la lettura del file e molti editor lo impostano durante il salvataggio.

**Python 3** supporta pienamente Unicode e UTF-8:

* È possibile utilizzare caratteri codificati in Unicode/UTF-8 per nominare variabili e altre entità;
* È possibile utilizzarli durante tutte le operazioni di input e output.

Ciò significa che Python 3 è completamente predisposto per l'I18N.
