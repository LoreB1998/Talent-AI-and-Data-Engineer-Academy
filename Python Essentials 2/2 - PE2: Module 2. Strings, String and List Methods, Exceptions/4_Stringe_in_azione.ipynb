# Confrontare le stringhe (*Comparing strings*)

Le stringhe in Python possono essere confrontate utilizzando lo stesso insieme di operatori che si usa per i numeri. Dai un'occhiata a questi operatori: sono tutti validi anche per il confronto tra stringhe.

```text
==   !=   >   >=   <   <=

```

C'è un però: i risultati di questi confronti a volte possono lasciarti un po' sorpreso. Non dimenticare che Python non comprende le sottigliezze linguistiche o il significato delle parole, ma si limita a confrontare i valori dei punti di codice (*code points*) dei caratteri, uno alla volta.

I risultati che si ottengono da un'operazione del genere a volte lasciano sbalorditi. Partiamo dai casi più semplici.

Due stringhe sono uguali quando sono composte dagli stessi identici caratteri nel medesimo ordine. Di conseguenza, due stringhe non sono uguali quando non presentano gli stessi caratteri nello stesso ordine. Entrambi i confronti seguenti restituiscono `True`:

```python
'alpha' == 'alpha'
'alpha' != 'Alpha'

```

La relazione finale tra due stringhe viene determinata confrontando il **primo carattere differente** in cui si imbatte il controllo (tenendo sempre a mente i punti di codice ASCII/UNICODE).

Quando confronti due stringhe di lunghezza diversa e la più corta è identica alla parte iniziale della stringa più lunga, **la stringa più lunga viene considerata maggiore**. Esattamente come in questo caso:

```python
'alpha' < 'alphabet'

```

Questa espressione è `True`.

Il confronto tra stringhe è **sempre sensibile alle maiuscole** (*case-sensitive*). All'interno delle tabelle dei caratteri, le lettere maiuscole hanno un valore numerico inferiore rispetto alle lettere minuscole. Per questo motivo, la seguente espressione è `True`:

```python
'beta' > 'Beta'

```

Anche se una stringa contiene esclusivamente cifre numeriche, non è comunque un numero. Viene interpretata esattamente per quello che è, come qualsiasi altra stringa ordinaria, e il suo (potenziale) aspetto numerico non viene preso in considerazione in alcun modo.

Guarda gli esempi:

````python
print('10' == '010')
print('10' > '010')
print('10' > '8')
print('20' < '8')
print('20' < '80')
````

Confrontare le stringhe con i numeri è in genere una pessima idea.

Gli unici confronti che puoi eseguire in totale sicurezza sono quelli indicati dagli operatori `==` e `!=`. Il primo restituirà sempre `False`, mentre il secondo restituirà sempre `True`, poiché una stringa e un numero appartengono a tipi di dato differenti e non potranno mai essere uguali.

L'utilizzo di uno qualsiasi degli altri operatori di confronto rimanenti (`>`, `>=`, `<`, `<=`) solleverà invece un'eccezione di tipo **`TypeError`**.

Verifichiamo questo comportamento:

````python
print('10' == 10)
print('10' != 10)
print('10' == 1)
print('10' != 1)
# print('10' > 10) # TypeError: '>' not supported between instances of 'str' and 'int'
````

# Ordinamento (*Sorting*)

Il concetto di confronto è strettamente legato a quello di ordinamento (o meglio, l'ordinamento è a tutti gli effetti un'applicazione molto sofisticata del confronto).

Questa è l'occasione perfetta per mostrarti due modi possibili per ordinare le liste che contengono stringhe. Si tratta di un'operazione estremamente comune nel mondo reale: ogni volta che ti trovi davanti a un elenco di nomi, prodotti, titoli o città, ti aspetti che siano in ordine.

Ipotizziamo di voler ordinare la seguente lista:

```python
greek = ['omega', 'alpha', 'pi', 'gamma']

```

In generale, Python offre due strade diverse per ordinare le liste.

La prima è implementata sotto forma di una funzione integrata chiamata `sorted()`. La funzione accetta un argomento (la lista) e **restituisce una nuova lista**, all'interno della quale gli elementi dell'argomento originale sono disposti in ordine. (Nota: questa descrizione è leggermente semplificata rispetto all'implementazione reale, ma approfondiremo i dettagli più avanti).

La lista di partenza rimane del tutto intatta.

Dai un'occhiata al codice nell'editor ed eseguilo.

````python
# Demonstrating the sorted() function:
first_greek = ['omega', 'alpha', 'pi', 'gamma']
first_greek_2 = sorted(first_greek)

print(first_greek)
print(first_greek_2)
````

Il secondo metodo agisce direttamente sulla lista stessa: **non viene creata alcuna nuova lista**. L'ordinamento viene eseguito *in situ* (sul posto) attraverso il metodo che si chiama `sort()`.

A differenza della funzione `sorted()`, il metodo `sort()` modifica in modo permanente la struttura della lista originale e restituisce `None`.

Dai un'occhiata all'esempio per vederlo in azione:

````python
# Demonstrating the sort() method:
second_greek = ['omega', 'alpha', 'pi', 'gamma']
print(second_greek)

second_greek.sort()
print(second_greek)
````

# Stringhe e numeri (*Strings vs. numbers*)

Ci sono due questioni aggiuntive che meritano di essere discusse a questo punto: come convertire un numero (un intero o un numero a virgola mobile) in una stringa, e viceversa. Può capitare spesso che sia necessario effettuare una simile trasformazione; inoltre, si tratta di una procedura di routine per elaborare i dati di input e output.

La conversione da numero a stringa è semplicissima, poiché è un'operazione sempre possibile. Viene eseguita attraverso una funzione integrata chiamata `str()`.

Esattamente come in questo caso:

````python
itg = 13
flt = 1.3
si = str(itg)
sf = str(flt)

print(si + ' ' + sf)
````

La trasformazione inversa (da stringa a numero) è possibile quando, e solo quando, la stringa rappresenta un numero valido. Se questa condizione non viene soddisfatta, l'operazione solleverà un'eccezione di tipo **`ValueError`**.

Si utilizza la funzione `int()` se si desidera ottenere un numero intero, e la funzione `float()` se invece serve un valore a virgola mobile.

Esattamente come in questo caso:

````python
si = '13'
sf = '1.3'
itg = int(si)
flt = float(sf)

print(itg + flt)
````
