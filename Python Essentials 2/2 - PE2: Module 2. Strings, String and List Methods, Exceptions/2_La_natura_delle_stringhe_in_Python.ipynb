# Stringhe - una breve panoramica

Facciamo una breve rassegna sulla natura delle stringhe in Python.

Prima di tutto, le stringhe di Python (o semplicemente stringhe, poiché non discuteremo delle stringhe di altri linguaggi) sono **sequenze immutabili**.

È molto importante notare questo aspetto, perché significa che dovresti aspettarti da esse alcuni comportamenti già familiari.

Analizziamo il codice nell'editor per capire di cosa stiamo parlando:

* Dai un'occhiata all'**Esempio 1**. La funzione `len()` utilizzata per le stringhe restituisce il numero di caratteri contenuti dagli argomenti. Lo snippet restituisce **2**.

# Stringhe multilinea

Questo è un ottimo momento per mostrarti un altro modo di specificare le stringhe all'interno del codice sorgente di Python. Nota che la sintassi che già conosci non ti permette di utilizzare una stringa che occupa più di una riga di testo.

Per questo motivo, il seguente codice è errato:

```python
multiline = 'Line #1
Line #2'

print(len(multiline))

```

Fortunatamente, per questo tipo di stringhe, Python offre una sintassi separata, comoda e semplice.

Guarda il codice nell'editor. Ecco come si presenta.

````python
multiline = '''Line #1
Line #2'''

print(len(multiline))
````

Come puoi vedere, la stringa inizia con tre apostrofi (apici singoli), non con uno. Lo stesso triplo apostrofo viene utilizzato per terminarla.

Il numero di righe di testo inserite all'interno di una stringa del genere è arbitrario.

Lo snippet restituisce **15**.

Conta i caratteri attentamente. Questo risultato è corretto o no? A prima vista sembra a posto, ma quando si contano i caratteri, non lo è più.

`Line #1` contiene sette caratteri. Due righe di questo tipo comprendono 14 caratteri. Abbiamo perso un carattere? Dove? Come?

No, non lo abbiamo perso.

Il carattere mancante è semplicemente invisibile: si tratta di uno spazio vuoto (*whitespace*). Si trova tra le due righe di testo.

Viene denotato come: `\n`.

Ricordi? È un carattere speciale (di controllo) utilizzato per forzare un avanzamento di riga (da cui il suo nome: *line feed*, o LF). Non puoi vederlo, ma conta.

Le stringhe multilinea possono anche essere delimitate da virgolette triple (apici doppi), proprio come qui:

````python
multiline = """Line #1
Line #2"""
 
print(len(multiline))
````

# Operazioni sulle stringhe

Come altri tipi di dati, le stringhe hanno un proprio insieme di operazioni consentite, sebbene siano piuttosto limitate rispetto ai numeri.

In generale, le stringhe possono essere:

* **concatenate** (unite)
* **replicate**.

La prima operazione viene eseguita dall'operatore `+` (nota: non si tratta di un'addizione), mentre la seconda dall'operatore `*` (nota ancora: non si tratta di una moltiplicazione).

La capacità di utilizzare lo stesso operatore con tipi di dati completamente diversi (come numeri e stringhe) è chiamata **overloading** (sovraccarico), poiché tale operatore viene sovraccaricato di compiti differenti.

Analizza l'esempio:

````python
str1 = 'a'
str2 = 'b'

print(str1 + str2)
print(str2 + str1)
print(5 * 'a')
print('b' * 4)
````

- L'operatore `+` utilizzato con due o più stringhe produce una nuova stringa contenente tutti i caratteri dei suoi argomenti (nota: l'ordine conta – questo `+` in versione sovraccaricata, a differenza della sua versione numerica, **non è commutativo**).

- L'operatore `*` richiede una stringa e un numero come argomenti; in questo caso l'ordine non importa – puoi mettere il numero prima della stringa o viceversa, il risultato sarà lo stesso: una nuova stringa creata dalla *n*-esima replica della stringa originale.

# `ord()`

Se vuoi conoscere il valore del *code point* ASCII/Unicode di uno specifico carattere, puoi utilizzare una funzione chiamata `ord()` (da *ordinal*).

La funzione richiede come argomento una **stringa di un solo carattere** – violare questo requisito causa un'eccezione di tipo `TypeError` – e restituisce un numero che rappresenta il *code point* dell'argomento.

Guarda il codice nell'editor ed eseguilo.

````python
# Demonstrating the ord() function.

char_1 = 'a'
char_2 = ' '  # space

print(ord(char_1))
print(ord(char_2))
````

# `chr()`

Se conosci il *code point* (numero) e vuoi ottenere il carattere corrispondente, puoi utilizzare una funzione chiamata `chr()`.

La funzione accetta un *code point* e restituisce il relativo carattere.

Invocarla con un argomento non valido (per esempio, un *code point* negativo o non valido) causa un'eccezione di tipo `ValueError` o `TypeError`.

Esegui il codice nell'editor.

````python
# Demonstrating the chr() function.

print(chr(97))
print(chr(945))
````

# Le stringhe come sequenze

## Indicizzazione

Ti abbiamo detto in precedenza che le stringhe in Python sono sequenze. È giunto il momento di mostrarti cosa significa concretamente.

Le stringhe non sono liste, ma puoi trattarle come liste in molti casi specifici.

Per esempio, se desideri accedere a uno qualsiasi dei caratteri di una stringa, puoi farlo utilizzando l'**indicizzazione**, proprio come nell'esempio seguente. Esegui il programma:

````python
# Indexing strings.

the_string = 'silly walks'

for ix in range(len(the_string)):
    print(the_string[ix], end=' ')

print()
````

## Iterazione

Anche l'iterazione attraverso le stringhe funziona. Guarda l'esempio seguente:

````python
# Iterating through a string.

the_string = 'silly walks'

for character in the_string:
    print(character, end=' ')

print()
````

# Slices (Affettamento)

Inoltre, tutto ciò che sai sulle *slices* (fette) è ancora utilizzabile.

Abbiamo raccolto alcuni esempi che mostrano come funzionano le *slices* nel mondo delle stringhe. Guarda il codice nell'editor, analizzalo ed eseguilo.

````python
# Slices

alpha = "abdefg"

print(alpha[1:3])
print(alpha[3:])
print(alpha[:3])
print(alpha[3:-2])
print(alpha[-3:4])
print(alpha[::2])
print(alpha[1::2])
````

# Gli operatori `in` e `not in`

## L'operatore `in`

L'operatore `in` non dovrebbe sorprenderti quando viene applicato alle stringhe: verifica semplicemente se il suo argomento di sinistra (una stringa) può essere trovato in un qualsiasi punto all'interno dell'argomento di destra (un'altra stringa).

Il risultato del controllo è semplicemente `True` o `False`.

Guarda il programma di esempio qui sotto. Ecco come funziona l'operatore `in`:

````python
alphabet = "abcdefghijklmnopqrstuvwxyz"

print("f" in alphabet)
print("F" in alphabet)
print("1" in alphabet)
print("ghi" in alphabet)
print("Xyz" in alphabet)
````

## L'operatore `not in`

Come probabilmente sospetti, anche l'operatore `not in` è applicabile in questo contesto.

Ecco come funziona:

````python
alphabet = "abcdefghijklmnopqrstuvwxyz"

print("f" not in alphabet)
print("F" not in alphabet)
print("1" not in alphabet)
print("ghi" not in alphabet)
print("Xyz" not in alphabet)
````

# Le stringhe in Python sono immutabili

Le stringhe di Python sono **immutabili**. Questa è una caratteristica molto importante. Cosa significa?

Significa principalmente che la somiglianza tra stringhe e liste è limitata. Non tutto ciò che puoi fare con una lista può essere fatto con una stringa.

La prima differenza importante è che non è consentito utilizzare l'istruzione `del` per rimuovere qualcosa da una stringa.

Il seguente esempio non funzionerà:

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
del alphabet[0]

```

L'unica cosa che puoi fare con `del` e una stringa è rimuovere la stringa nel suo complesso. Prova a farlo.

Le stringhe di Python non hanno il metodo `append()` – non puoi espanderle in alcun modo.

L'esempio seguente è errato:

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
alphabet.append("A")

```

Insieme all'assenza del metodo `append()`, anche il metodo `insert()` è illecito:

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
alphabet.insert(0, "A")

```

Non pensare che l'immutabilità di una stringa limiti la tua capacità di operare con le stringhe.

L'unica conseguenza è che devi ricordartene e implementare il tuo codice in un modo leggermente diverso – guarda il codice di esempio nell'editor.

````python
alphabet = "bcdefghijklmnopqrstuvwxy"

alphabet = "a" + alphabet
alphabet = alphabet + "z"

print(alphabet)
````

# Operazioni sulle stringhe: continua

## `min()`

Ora che hai compreso che le stringhe sono sequenze, possiamo mostrarti alcune funzionalità delle sequenze meno ovvie. Le presenteremo usando le stringhe, ma non dimenticare che anche le liste possono adottare gli stessi trucchi.

Cominciamo con una funzione chiamata `min()`.

La funzione trova l'elemento minimo della sequenza passata come argomento. C'è una condizione: la sequenza (stringa, lista, non importa) non può essere vuota, altrimenti si otterrà un'eccezione di tipo `ValueError`.

````python
# Demonstrating min() - Example 1:
print(min("aAbByYzZ"))


# Demonstrating min() - Examples 2 & 3:
t = 'The Knights Who Say "Ni!"'
print('[' + min(t) + ']')

t = [0, 1, 2]
print(min(t))
````

## `max()`

Allo stesso modo, una funzione chiamata `max()` trova l'elemento massimo della sequenza.

Anche in questo caso vale la stessa regola: la sequenza non deve essere vuota, altrimenti Python solleverà un'eccezione di tipo `ValueError`. Nel caso delle stringhe, determina il carattere con il *code point* ASCII/Unicode più alto.

````python
# Demonstrating max() - Example 1:
print(max("aAbByYzZ"))


# Demonstrating max() - Examples 2 & 3:
t = 'The Knights Who Say "Ni!"'
print('[' + max(t) + ']')

t = [0, 1, 2]
print(max(t))
````

## Il metodo `index()`

Il metodo `index()` (si tratta di un metodo, non di una funzione) cerca nella sequenza partendo dall'inizio, al fine di trovare il primo elemento che corrisponde al valore specificato nel suo argomento.

> **Nota:** l'elemento cercato deve essere presente nella sequenza – la sua assenza causerà un'eccezione di tipo `ValueError`.

Il metodo restituisce l'indice della prima occorrenza dell'argomento (il che significa che il risultato minimo possibile è 0, mentre il massimo è la lunghezza della sequenza decrementata di 1).

````python
# Demonstrating the index() method:
print("aAbByYzZaA".index("b"))
print("aAbByYzZaA".index("Z"))
print("aAbByYzZaA".index("A"))
````

## La funzione `list()`

La funzione `list()` prende il suo argomento (una stringa) e crea una nuova lista contenente tutti i caratteri della stringa, uno per ogni elemento della lista.

> **Nota:** non si tratta di una funzione strettamente legata alle stringhe – `list()` è in grado di creare una nuova lista a partire da molte altre entità (per esempio, da tuple e dizionari).

Dai un'occhiata all'esempio di codice nell'editor.

````python
# Demonstrating the list() function:
print(list("abcabc"))
````

### Il metodo `count()`

Il metodo `count()` conta tutte le occorrenze dell'elemento all'interno della sequenza. L'assenza di tale elemento non causa alcun problema e restituisce semplicemente `0`.

Dai un'occhiata al secondo esempio nell'editor. Riesci a indovinarne l'output?

````python
# Demonstrating the count() method:
print("abcabc".count("b"))
print('abcabc'.count("d"))
````
