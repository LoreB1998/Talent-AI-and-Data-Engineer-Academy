# Ottenere la data locale corrente e creare oggetti `date`

Una delle classi principali fornite dal modulo `datetime` è la classe chiamata **`date`**. Gli oggetti di questa classe rappresentano una data composta da tre elementi: anno, mese e giorno.

Guarda il codice qui sotto per vedere come funziona in pratica e come ottenere la data locale corrente utilizzando il metodo `today()`.



````python
from datetime import date

# Ottiene la data locale corrente
oggi = date.today()

print("Data corrente:", oggi)
print("Anno:", oggi.year)
print("Mese:", oggi.month)
print("Giorno:", oggi.day)
````

Il metodo `today` restituisce un oggetto `date` che rappresenta la data locale corrente. Nota che l'oggetto `date` ha tre attributi: `year` (anno), `month` (mese) e `day` (giorno).

Fai attenzione, perché questi attributi sono in sola lettura. Per creare un oggetto `date`, devi passare i parametri anno, mese e giorno come segue:

````python
from datetime import date

my_date = date(2019, 11, 4)
print(my_date)
````

# Creazione di un oggetto date da un timestamp

La classe `date` ci offre la possibilità di creare un oggetto `date` a partire da un timestamp.

Nei sistemi Unix, il **timestamp** esprime il numero di secondi trascorsi dal 1° gennaio 1970, 00:00:00 (UTC). Questa data viene chiamata **epoca Unix** (Unix epoch), poiché è il momento in cui ha avuto inizio il conteggio del tempo sui sistemi Unix.

Il timestamp è di fatto la differenza tra una data specifica (incluso l'orario) e il 1° gennaio 1970, 00:00:00 (UTC), espressa in secondi.

Per creare un oggetto `date` da un timestamp, dobbiamo passare un timestamp Unix al metodo `fromtimestamp()`.

A questo scopo possiamo utilizzare il modulo `time`, che fornisce funzioni correlate al tempo. Una di queste è la funzione chiamata `time()`, che restituisce il numero di secondi trascorsi dal 1° gennaio 1970 fino al momento corrente sotto forma di numero a virgola mobile (float). Dai un'occhiata amichevole all'esempio nell'editor:


````python
from datetime import date
import time

timestamp = time.time()
print("Timestamp:", timestamp)

d = date.fromtimestamp(timestamp)
print("Date:", d)
    
````

Esegui il codice per vedere l'output.

Se esegui il codice di esempio diverse volte, potrai notare come il timestamp si incrementi da solo. Vale la pena aggiungere che il risultato della funzione `time` dipende dalla piattaforma, poiché nei sistemi Unix e Windows i secondi intercalari (*leap seconds*) non vengono conteggiati.

> **Nota:** In questa parte del corso parleremo anche del modulo `time`.

## Creazione di un oggetto date utilizzando il formato ISO

Il modulo `datetime` fornisce diversi metodi per creare un oggetto `date`. Uno di questi è il metodo `fromisoformat`, che accetta una data nel formato `AAAA-MM-GG` conforme allo standard **ISO 8601**.

Lo standard ISO 8601 definisce il modo in cui vengono rappresentate la data e l'ora. Viene utilizzato molto spesso, quindi vale la pena dedicare un momento a familiarizzare con esso. Dai un'occhiata allo schema che descrive i valori richiesti dal formato:

Ora guarda il codice nell'editor ed eseguilo.
Nel nostro esempio, `AAAA` è 2019, `MM` è 11 (novembre) e `GG` è 04 (quarto giorno di novembre).

```python
from datetime import date

d = date.fromisoformat('2019-11-04')
print(d)

```

Quando inserisci la data, assicurati di aggiungere lo `0` prima del mese o del giorno se sono espressi da un numero inferiore a 10.

> **Nota:** Il metodo `fromisoformat` è disponibile in Python a partire dalla versione 3.7.

## Il metodo `replace()`

A volte potresti avere la necessità di sostituire l'anno, il mese o il giorno con un valore diverso. Non puoi farlo modificando direttamente gli attributi `year`, `month` e `day` perché, come abbiamo visto, sono in sola lettura. In questo caso, puoi utilizzare il metodo chiamato `replace`.

Esegui il codice nell'editor per vedere come funziona:

```python
from datetime import date

d = date(2019, 11, 4)
print(d)

d = d.replace(year=2026, month=5)
print(d)

```

Il metodo `replace` restituisce un nuovo oggetto data con i valori modificati, lasciando inalterato l'oggetto originale.

# Che giorno della settimana è?
Uno dei metodi più utili che rende più semplice il lavoro con le date è il metodo chiamato `weekday`. Questo restituisce il giorno della settimana sotto forma di numero intero, dove 0 corrisponde a lunedì e 6 a domenica. Esegui il codice nell'editor.

````python
from datetime import date

d = date(2019, 11, 4)
print(d.weekday())
````

La classe `date` ha un metodo simile chiamato `isoweekday`, che restituisce anch'esso il giorno della settimana sotto forma di numero intero, ma in questo caso 1 corrisponde a lunedì e 7 a domenica:

```python
from datetime import date
 
d = date(2019, 11, 4)
print(d.isoweekday())

```

**Risultato:**

```text
1

```

Come puoi notare, per la stessa identica data otteniamo un intero diverso, che esprime però lo stesso giorno della settimana. L'intero restituito dal metodo `isoweekday` segue la specifica ISO 8601.

## Il modulo `time`

In aggiunta alla classe `time`, la libreria standard di Python offre un modulo chiamato `time`, il quale fornisce funzioni correlate al tempo. Hai già avuto l'opportunità di conoscere la funzione chiamata `time()` durante la spiegazione della classe `date`. Ora vedremo un'altra funzione utile disponibile in questo modulo.

## La funzione `ctime()`

Il modulo `time` fornisce una funzione chiamata `ctime`, che converte in una stringa il tempo espresso in secondi a partire dal 1° gennaio 1970 (epoca Unix).
Ricordi il risultato della funzione `time`? Questo è proprio ciò che devi passare a `ctime`. Dai un'occhiata l'esempio nell'editor.

```python

```

**Risultato:**

```text

```

La funzione `ctime` restituisce una stringa corrispondente al timestamp passato. Nel nostro esempio, il timestamp esprime il 4 novembre 2019 alle 14:53:00.

È anche possibile chiamare la funzione `ctime` senza specificare il tempo in secondi. In questo caso, verrà restituito il tempo corrente:

## Le funzioni `gmtime()` e `localtime()`

Alcune delle funzioni disponibili nel modulo `time` richiedono la conoscenza della classe `struct_time`, ma prima di conoscerle, vediamo come si presenta la classe:

```python
time.struct_time:
    tm_year   # Specifica l'anno.
    tm_mon    # Specifica il mese (valore da 1 a 12)
    tm_mday   # Specifica il giorno del mese (valore da 1 a 31)
    tm_hour   # Specifica l'ora (valore da 0 to 23)
    tm_min    # Specifica il minuto (valore da 0 to 59)
    tm_sec    # Specifica il secondo (valore da 0 to 61)
    tm_wday   # Specifica il giorno della settimana (valore da 0 to 6)
    tm_yday   # Specifica il giorno dell'anno (valore da 1 to 366)
    tm_isdst  # Specifica se si applica l'ora legale (1 – sì, 0 – no, -1 – non è noto)
    tm_zone   # Specifica il nome del fuso orario (valore in forma abbreviata)
    tm_gmtoff # Specifica l'offset a est di UTC (valore in secondi)

```

La classe `struct_time` consente inoltre l'accesso ai valori tramite indici. L'indice 0 restituisce il valore in `tm_year`, mentre l'8 restituisce il valore in `tm_isdst`.

Le eccezioni sono `tm_zone` e `tm_gmtoff`, alle quali non è possibile accedere utilizzando gli indici. Vediamo come utilizzare la classe `struct_time` in pratica. Esegui il codice nell'editor.

```python
import time

timestamp = 1572879180
print(time.gmtime(timestamp))
print(time.localtime(timestamp))

```

**Risultato:**

```text
time.struct_time(tm_year=2019, tm_mon=11, tm_mday=4, tm_hour=14, tm_min=53, tm_sec=0, tm_wday=0, tm_yday=308, tm_isdst=0)
time.struct_time(tm_year=2019, tm_mon=11, tm_mday=4, tm_hour=14, tm_min=53, tm_sec=0, tm_wday=0, tm_yday=308, tm_isdst=0)

```

L'esempio mostra due funzioni che convertono il tempo trascorso dall'epoca Unix in un oggetto `struct_time`. La differenza tra di esse è che la funzione `gmtime` restituisce l'oggetto `struct_time` in formato UTC, mentre la funzione `localtime` restituisce il tempo locale. Per la funzione `gmtime`, l'attributo `tm_isdst` è sempre 0.

## Le funzioni `asctime()` e `mktime()`

Il modulo `time` include funzioni che si aspettano un oggetto `struct_time` o una tupla che memorizzi i valori in base agli indici presentati durante la spiegazione della classe `struct_time`. Esegui l'esempio nell'editor.

```python
import time

timestamp = 1572879180
st = time.gmtime(timestamp)

print(time.asctime(st))
print(time.mktime((2019, 11, 4, 14, 53, 0, 0, 308, 0)))

```

**Risultato:**

```text
Mon Nov 4 14:53:00 2019
1572879180.0

```

La prima delle funzioni, chiamata `asctime`, converte un oggetto `struct_time` o una tupla in una stringa. Nota che viene utilizzata la già nota funzione `gmtime` per ottenere l'oggetto `struct_time`. Se non fornisci un argomento alla funzione `asctime`, verrà utilizzato il tempo restituito dalla funzione `localtime`.

La seconda funzione, chiamata `mktime`, converte un oggetto `struct_time` o una tupla che esprime il tempo locale nel numero di secondi a partire dall'epoca Unix. Nel nostro esempio, le abbiamo passato una tupla composta dai seguenti valori:

* `2019` => `tm_year`
* `11` => `tm_mon`
* `4` => `tm_mday`
* `14` => `tm_hour`
* `53` => `tm_min`
* `0` => `tm_sec`
* `0` => `tm_wday`
* `308` => `tm_yday`
* `0` => `tm_isdst`

## Metodi che restituiscono la data e il tempo correnti

La classe `datetime` ha diversi metodi che restituiscono la data e il tempo correnti. Questi metodi sono:

* `today()` — restituisce la data e il tempo locali correnti con l'attributo `tzinfo` impostato su `None`;
* `now()` — restituisce la data e il tempo locali correnti allo stesso modo del metodo `today`, a meno che non si passi l'argomento opzionale `tz`. L'argomento di questo metodo deve essere un oggetto di una sottoclasse di `tzinfo`;
* `utcnow()` — restituisce la data e il tempo UTC correnti con l'attributo `tzinfo` impostato su `None`.

Esegui il codice nell'editor per vederli tutti in pratica. Cosa puoi dire riguardo all'output?

```python
from datetime import datetime

print("today:", datetime.today())
print("now:", datetime.now())
print("utcnow:", datetime.utcnow())

```

Come puoi vedere, il risultato di tutti e tre i metodi è lo stesso. Le piccole differenze sono causate dal tempo trascorso tra le chiamate successive.
Nota: puoi leggere di più sugli oggetti `tzinfo` nella documentazione.

## Formattazione di data e ora

Tutte le classi del modulo `datetime` presentate finora hanno un metodo chiamato `strftime`. Questo è un metodo molto importante, perché ci permette di restituire la data e il tempo nel formato da noi specificato.

Il metodo `strftime` accetta un solo argomento sotto forma di stringa che specifica un formato, il quale può essere composto da direttive.
Una direttiva è una stringa costituita dal carattere `%` (percento) e da una lettera minuscola o maiuscola. Ad esempio, la direttiva `%Y` indica l'anno con il secolo come numero decimale. Vediamolo in un esempio. Esegui il codice nell'editor.

```python
from datetime import date

d = date(2020, 1, 4)
print(d.strftime('%Y/%m/%d'))

```

**Risultato:**

```text
2020/01/04

```

Nell'esempio abbiamo passato al metodo `strftime` un formato composto da tre direttive separate da `/` (barra). Naturalmente, il carattere separatore può essere sostituito da un altro carattere o persino da una stringa.

Puoi inserire qualsiasi carattere nel formato, ma solo le direttive riconoscibili saranno sostituite con i valori appropriati. Nel nostro formato abbiamo utilizzato le seguenti direttive:

* `%Y` – restituisce l'anno con il secolo come numero decimale. Nel nostro esempio è 2020.
* `%m` – restituisce il mese come numero decimale riempito con zeri. Nel nostro esempio è 01.
* `%d` – restituisce il giorno come numero decimale riempito con zeri. Nel nostro esempio è 04.

Nota: puoi trovare tutte le direttive disponibili qui.

La formattazione del tempo funziona nello stesso modo della formattazione della data, ma richiede l'uso delle direttive appropriate. Diamo un'occhiata più da vicino a una manciata di esse nell'editor.

```python
from datetime import time
from datetime import datetime

t = time(14, 53)
print(t.strftime("%H:%M:%S"))

dt = datetime(2020, 11, 4, 14, 53)
print(dt.strftime("%y/%B/%d %H:%M:%S"))

```

**Risultato:**

```text
14:53:00
20/November/04 14:53:00

```

Il primo dei formati utilizzati riguarda solo il tempo. Come puoi indovinare, `%H` restituisce l'ora come numero decimale riempito con zeri, `%M` restituisce il minuto come numero decimale riempito con zeri, mentre `%S` restituisce il secondo come numero decimale riempito con zeri. Nel nostro esempio, `%H` viene sostituito da 14, `%M` da 53 e `%S` da 00.

Il secondo formato utilizzato combina le direttive di data e tempo. Sono presenti due nuove direttive, `%y` e `%B`. La direttiva `%y` restituisce l'anno senza il secolo come numero decimale riempito con zeri (nel nostro esempio è 20). La direttiva `%B` restituisce il mese come nome completo nella lingua del sistema locale (nel nostro esempio è November).

In generale hai molta libertà nella creazione dei formati, ma devi ricordare di usare le direttive in modo corretto. Come esercizio, puoi verificare cosa succede se, per esempio, provi a usare la direttiva `%Y` nel formato passato al metodo `strftime` di un oggetto tempo. Prova a scoprire da solo il motivo per cui hai ottenuto tale risultato. Buona fortuna!


## La funzione `strftime()` nel modulo `time`

Probabilmente non ti sorprenderà sapere che la funzione `strftime` è disponibile anche nel modulo `time`. Differisce leggermente dai metodi `strftime` presenti nelle classi fornite dal modulo `datetime` perché, oltre all'argomento del formato, può anche accettare (opzionalmente) una tupla o un oggetto `struct_time`.

Se non passi una tupla o un oggetto `struct_time`, la formattazione verrà eseguita utilizzando il tempo locale corrente. Dai un'occhiata all'esempio nell'editor.

```python
import time

timestamp = 1572879180
st = time.gmtime(timestamp)

print(time.strftime("%Y/%m/%d %H:%M:%S", st))

```

Il nostro risultato si presenta come segue:

**Risultato:**

```text

2020/10/12 12:19:40

```

La creazione di un formato avviene nello stesso modo visto per i metodi `strftime` nel modulo `datetime`. Nel nostro esempio utilizziamo le direttive `%Y`, `%m`, `%d`, `%H`, `%M` e `%S` che già conosci.

Nella prima chiamata della funzione formattiamo l'oggetto `struct_time`, mentre nella seconda chiamata (senza l'argomento opzionale) formattiamo il tempo locale. Puoi trovare tutte le direttive disponibili nel modulo `time` qui.

## Il metodo `strptime()`

Sapere come creare un formato può essere d'aiuto quando si utilizza un metodo chiamato `strptime` nella classe `datetime`. A differenza del metodo `strftime`, esso crea un oggetto `datetime` a partire da una stringa che rappresenta una data e un tempo.

Il metodo `strptime` richiede di specificare il formato in cui sono stati salvati la data e il tempo. Vediamolo in un esempio. Dai un'occhiata al codice nell'editor.

```python
from datetime import datetime
print(datetime.strptime("2019/11/04 14:53:00", "%Y/%m/%d %H:%M:%S"))

```

**Risultato:**

```text
2019-11-04 14:53:00

```

Nell'esempio abbiamo specificato due argomenti obbligatori. Il primo è una data e un tempo sotto forma di stringa: `"2019/11/04 14:53:00"`, mentre il secondo è il formato che facilita l'analisi (*parsing*) per ottenere l'oggetto `datetime`. Fai attenzione, perché se il formato specificato non corrisponde alla data e al tempo presenti nella stringa, verrà sollevata un'eccezione `ValueError`.

> **Nota:** Nel modulo `time` puoi trovare una funzione chiamata `strptime`, che analizza una stringa che rappresenta un tempo per convertirla in un oggetto `struct_time`. Il suo utilizzo è analogo al metodo `strptime` nella classe `datetime`:

```python
import time
print(time.strptime("2019/11/04 14:53:00", "%Y/%m/%d %H:%M:%S"))

```

Il suo risultato sarà il seguente:

**Risultato:**

```text
time.struct_time(tm_year=2019, tm_mon=11, tm_mday=4, tm_hour=14, tm_min=53, tm_sec=0, tm_wday=0, tm_yday=308, tm_isdst=-1)

```

## Operazioni con date e ore

Prima o poi dovrai eseguire dei calcoli su date e ore. Fortunatamente, nel modulo `datetime` è presente una classe chiamata `timedelta` che è stata creata proprio a questo scopo.

Per creare un oggetto `timedelta`, basta eseguire una sottrazione su oggetti `date` o `datetime`, proprio come abbiamo fatto nell'esempio nell'editor. Eseguilo.

```python
from datetime import date
from datetime import datetime

d1 = date(2020, 11, 4)
d2 = date(2019, 11, 4)

print(d1 - d2)

dt1 = datetime(2020, 11, 4, 0, 0, 0)
dt2 = datetime(2019, 11, 4, 14, 53, 0)

print(dt1 - dt2)

```

**Risultato:**

```text
366 days, 0:00:00
365 days, 9:07:00

```

L'esempio mostra la sottrazione sia per gli oggetti `date` che per quelli `datetime`. Nel primo caso otteniamo la differenza in giorni, che è di 366 giorni. Nota che viene visualizzata anche la differenza in ore, minuti e secondi. Nel secondo caso otteniamo un risultato diverso, perché abbiamo specificato l'orario che è stato incluso nei calcoli. Come risultato otteniamo 365 giorni, 9 ore e 7 minuti.

Tra un momento imparerai di più sulla creazione di oggetti `timedelta` e sulle operazioni che puoi effettuare con essi.


## Creazione di oggetti `timedelta`

Hai già imparato che un oggetto `timedelta` può essere restituito come risultato della sottrazione di due oggetti `date` o `datetime`.
Naturalmente puoi anche creare un oggetto autonomamente. A questo scopo, conosciamo gli argomenti accettati dal costruttore della classe, che sono: `days` (giorni), `seconds` (secondi), `microseconds` (microsecondi), `milliseconds` (millisecondi), `minutes` (minuti), `hours` (ore) e `weeks` (settimane). Ciascuno di essi è opzionale e ha come valore predefinito 0.

```python
from datetime import timedelta

delta = timedelta(weeks=2, days=2, hours=3)
print(delta)

```

**Risultato:**

```text
16 days, 3:00:00

```

Il risultato di 16 giorni si ottiene convertendo l'argomento `weeks` in giorni (2 settimane = 14 giorni) e aggiungendo l'argomento `days` (2 giorni). Questo è il comportamento normale, perché l'oggetto `timedelta` memorizza internamente solo giorni, secondi e microsecondi. Allo stesso modo, l'argomento delle ore viene convertito. Dai un'occhiata all'esempio seguente:

```python
from datetime import timedelta
 
delta = timedelta(weeks=2, days=2, hours=3)
print("Days:", delta.days)
print("Seconds:", delta.seconds)
print("Microseconds:", delta.microseconds)

```

**Risultato:**

```text
Days: 16
Seconds: 10800
Microseconds: 0

```

Il risultato di 10800 si ottiene convertendo 3 ore in secondi. In questo modo l'oggetto `timedelta` memorizza gli argomenti passati durante la sua creazione. Le settimane vengono convertite in giorni, le ore e i minuti in secondi, e i millisecondi in microsecondi.

Sai già come l'oggetto `timedelta` memorizza internamente gli argomenti passati. Vediamo come può essere utilizzato in pratica.
Guarda alcune operazioni supportate dalle classi del modulo `datetime`. Esegui il codice che abbiamo fornito nell'editor.

```python
from datetime import timedelta
from datetime import date
from datetime import datetime

delta = timedelta(weeks=2, days=2, hours=2)
print(delta)

delta2 = delta * 2
print(delta2)

d = date(2019, 10, 4) + delta2
print(d)

dt = datetime(2019, 10, 4, 14, 53) + delta2
print(dt)

```

**Risultato:**

```text
16 days, 2:00:00
32 days, 4:00:00
2019-11-05
2019-11-05 18:53:00

```

L'oggetto `timedelta` può essere moltiplicato per un numero intero. Nel nostro esempio moltiplichiamo per 2 l'oggetto che rappresenta 16 giorni e 2 hours. Di conseguenza otteniamo un oggetto `timedelta` che rappresenta 32 giorni e 4 ore.

Nota che sia i giorni che le ore sono stati moltiplicati per 2. Un'altra operazione interessante che utilizza l'oggetto `timedelta` è l'addizione. Nell'esempio abbiamo aggiunto l'oggetto `timedelta` agli oggetti `date` e `datetime`.

Come risultato di queste operazioni, otteniamo oggetti `date` e `datetime` incrementati dei giorni e delle ore memorizzati nell'oggetto `timedelta`.
L'operazione di moltiplicazione presentata consente di aumentare rapidamente il valore dell'oggetto `timedelta`, mentre l'addizione può aiutarti a ottenere una data nel futuro.

Naturalmente le classi `timedelta`, `date` e `datetime` supportano molte altre operazioni. Ti invitiamo a familiarizzare con esse consultando la documentazione.
