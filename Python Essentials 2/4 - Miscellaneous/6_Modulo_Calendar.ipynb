# Introduzione al modulo `calendar`

Insieme ai moduli `datetime` e `time`, la libreria standard di Python mette a disposizione un modulo chiamato `calendar` che, come suggerisce il nome, offre funzionalità legate al calendario.

Una di queste è, ovviamente, la visualizzazione del calendario. È importante notare che i giorni della settimana vengono visualizzati da lunedì a domenica e ogni giorno della settimana ha una sua rappresentazione sotto forma di numero intero:

| Giorno della settimana | Valore intero | Costante |
| --- | --- | --- |
| Lunedì | 0 | `calendar.MONDAY` |
| Martedì | 1 | `calendar.TUESDAY` |
| Mercoledì | 2 | `calendar.WEDNESDAY` |
| Giovedì | 3 | `calendar.THURSDAY` |
| Venerdì | 4 | `calendar.FRIDAY` |
| Sabato | 5 | `calendar.SATURDAY` |
| Domenica | 6 | `calendar.SUNDAY` |


# Il tuo primo calendario

Inizierai la tua avventura con il modulo `calendar` usando una semplice funzione chiamata `calendar`, che consente di visualizzare il calendario di un intero anno. Vediamo come usarla per mostrare il calendario del 2020. Esegui il codice nell'editor e osserva cosa succede.

```python
import calendar
print(calendar.calendar(2020))

```

Il risultato visualizzato è simile a quello del comando `cal` disponibile nei sistemi Unix. Se desideri modificare la formattazione predefinita del calendario, puoi utilizzare i seguenti parametri:

* **`w`** – larghezza della colonna della data (predefinito: 2)
* **`l`** – numero di righe per ogni settimana (predefinito: 1)
* **`c`** – numero di spazi tra le colonne dei mesi (predefinito: 6)
* **`m`** – numero di colonne dei mesi (predefinito: 3)

La funzione `calendar` richiede obbligatoriamente di specificare l'anno, mentre gli altri parametri responsabili della formattazione sono opzionali. Ti invitiamo a sperimentare questi parametri autonomamente.

Un'ottima alternativa alla funzione appena descritta è la funzione `prcal`, che accetta gli stessi parametri di `calendar`, ma non richiede l'uso della funzione `print()` per mostrare il risultato a schermo. Il suo utilizzo si presenta così:

```python
import calendar
calendar.prcal(2020)

```


# Calendario per un mese specifico

Il modulo `calendar` include una funzione chiamata `month`, che consente di visualizzare il calendario di un mese specifico. Il suo utilizzo è davvero semplice, basta specificare l'anno e il mese – controlla il codice nell'editor:

```python
import calendar
print(calendar.month(2020, 11))

```

L'esempio mostra il calendario di novembre 2020. Come per la funzione `calendar`, puoi modificare la formattazione predefinita usando i seguenti parametri:

* **`w`** – larghezza della colonna della data (predefinito: 2)
* **`l`** – numero di righe per ogni settimana (predefinito: 1)

> **Nota:** Puoi anche utilizzare la funzione `prmonth`, che ha gli stessi parametri della funzione `month`, ma non richiede l'uso di `print()` per visualizzare il calendario.


# La funzione `setfirstweekday()`

Come già sai, nel modulo `calendar` il primo giorno della settimana predefinito è il lunedì. Tuttavia, puoi modificare questo comportamento utilizzando una funzione chiamata `setfirstweekday`.

Ricordi la tabella che mostra i giorni della settimana e la loro rappresentazione numerica? È il momento di usarla, perché il metodo `setfirstweekday` richiede un parametro che esprima il giorno della settimana sotto forma di valore intero. Dai un'occhiata all'esempio nell'editor.

```python
import calendar

calendar.setfirstweekday(calendar.SUNDAY)
calendar.prmonth(2020, 12)

```

L'esempio utilizza la costante `calendar.SUNDAY`, che contiene il valore 6. Ovviamente potresti passare questo valore direttamente alla funzione `setfirstweekday`, ma la versione con la costante è decisamente più elegante.

Come risultato, otterremo un calendario del mese di dicembre 2020 nel quale il primo giorno di tutte le settimane è la domenica.


# La funzione `weekday()`

Un'altra utile funzione fornita dal modulo `calendar` è `weekday`, che restituisce il giorno della settimana come valore intero per un determinato anno, mese e giorno. Vediamola in pratica.

Esegui il codice nell'editor per verificare quale giorno della settimana cadeva il 24 dicembre 2020:

```python
import calendar
print(calendar.weekday(2020, 12, 24))

```


# La funzione `weekheader()`

Avrai probabilmente notato che il calendario contiene le intestazioni dei giorni in forma abbreviata. Se necessario, puoi ottenere i nomi brevi dei giorni della settimana usando il metodo `weekheader`.

Il metodo `weekheader` richiede di specificare la larghezza in caratteri per un singolo giorno della settimana. Se la larghezza fornita è superiore a 3, otterrai comunque i nomi abbreviati composti da tre caratteri.

Vediamo come ottenere un'intestazione più compatta. Esegui il codice nell'editor.

```python
import calendar
print(calendar.weekheader(2))

```

**Risultato:**

```text
Mo Tu We Th Fr Sa Su

```

> **Nota:** Se modifichi il primo giorno della settimana, ad esempio utilizzando la funzione `setfirstweekday`, questo influirà direttamente sul risultato della funzione `weekheader`.


# Come verificare se un anno è bisestile?

Il modulo `calendar` fornisce due funzioni molto utili per verificare se gli anni sono bisestili.

La prima, chiamata `isleap`, restituisce `True` se l'anno passato è bisestile, altrimenti restituisce `False`. La seconda, chiamata `leapdays`, restituisce il numero di anni bisestili all'interno di un determinato intervallo di anni.

Esegui il codice nell'editor:

```python
import calendar

print(calendar.isleap(2020))
print(calendar.leapdays(2010, 2021))  # Fino al 2021 escluso.

```

**Risultato:**

```text
True
3

```

Nell'esempio otteniamo come risultato `3`, perché nel periodo dal 2010 al 2020 ci sono esattamente tre anni bisestiles (nota: il 2021 non è incluso nel conteggio). Si tratta degli anni 2012, 2016 e 2020.


# Classi per la creazione di calendari

Le funzioni che ti abbiamo mostrato finora non rappresentano tutto ciò che il modulo `calendar` ha da offrire. Oltre a esse, possiamo utilizzare le seguenti classi:

* **`calendar.Calendar`** – fornisce metodi per preparare i dati del calendario per la formattazione;
* **`calendar.TextCalendar`** – viene utilizzata per creare calendari di testo standard;
* **`calendar.HTMLCalendar`** – viene utilizzata per creare calendari in formato HTML;
* **`calendar.LocalTextCalendar`** – è una sottoclasse di `calendar.TextCalendar`. Il costruttore di questa classe accetta il parametro `locale`, usato per restituire i nomi dei mesi e dei giorni nella lingua locale appropriata;
* **`calendar.LocalHTMLCalendar`** – è una sottoclasse di `calendar.HTMLCalendar`. Il costruttore accetta il parametro `locale` per adattare i nomi dei mesi e dei giorni alla lingua locale.

Durante questo corso hai già avuto modo di creare calendari testuali quando abbiamo esaminato le funzioni del modulo. È il momento di provare qualcosa di nuovo: diamo un'occhiata più da vicino ai metodi della classe `Calendar`.


# Creazione di un oggetto `Calendar`

Il costruttore della classe `Calendar` accetta un parametro opzionale chiamato `firstweekday`, che di default è uguale a 0 (lunedì).

Il parametro `firstweekday` deve essere un numero intero compreso tra 0 e 6. A questo scopo possiamo utilizzare le costanti che già conosciamo – guarda il codice nell'editor:

```python
import calendar  

c = calendar.Calendar(calendar.SUNDAY)

for weekday in c.iterweekdays():
    print(weekday, end=" ")

```

Il programma restituirà il seguente output:

```text
6 0 1 2 3 4 5

```

L'esempio di codice utilizza il metodo della classe `Calendar` chiamato `iterweekdays`, che restituisce un iteratore per i numeri dei giorni della settimana.

Il primo valore restituito è sempre uguale al valore della proprietà `firstweekday`. Poiché nel nostro esempio il primo valore restituito è 6, significa che la settimana inizia di domenica.


# Il metodo `itermonthdates()`

La classe `Calendar` ha diversi metodi che restituiscono un iteratore. Uno di questi è il metodo `itermonthdates`, che richiede di specificare l'anno e il mese.

Come risultato, vengono restituiti tutti i giorni del mese e dell'anno specificati, oltre a tutti i giorni precedenti l'inizio del mese o successivi alla fine del mese che sono necessari per completare le settimane intere.

Ogni giorno è rappresentato da un oggetto `datetime.date`. Dai un'occhiata all'esempio nell'editor.

```python
import calendar  

c = calendar.Calendar()

for date in c.itermonthdates(2019, 11):
    print(date, end=" ")

```

Il codice visualizza tutti i giorni in novembre 2019. Poiché il primo giorno di novembre 2019 era un venerdì, vengono restituiti anche i giorni precedenti per completare la settimana: `2019-10-28` (lunedì), `2019-10-29` (martedì), `2019-10-30` (mercoledì), `2019-10-31` (giovedì).

L'ultimo giorno di novembre 2019 era un sabato, quindi per completare l'ultima settimana viene restituito un giorno in più: `2019-12-01` (domenica).


# Altri metodi che restituiscono iteratori

Un altro metodo utile nella classe `Calendar` è `itermonthdays`, che accetta come parametri l'anno e il mese, e poi restituisce un iteratore per i giorni del mese rappresentati da numeri.

Dai un'occhiata all'esempio nell'editor.

```python
import calendar  

c = calendar.Calendar()

for day in c.itermonthdays(2019, 11):
    print(day, end=" ")

```

Avrai sicuramente notato l'elevato numero di `0` restituiti come risultato dal codice di esempio. Questi rappresentano i giorni al di fuori dell'intervallo del mese specificato, che vengono aggiunti per completare la settimana.

I primi quattro zeri rappresentano il `28/10/2019` (lunedì), `29/10/2019` (martedì), `30/10/2019` (mercoledì) e `31/10/2019` (giovedì). I numeri successivi sono i giorni effettivi del mese, ad eccezione dell'ultimo valore `0`, che sostituisce la data del `01/12/2019` (domenica).

Esistono altri tre metodi simili nella classe `Calendar` che si differenziano per la tipologia di dati restituiti:

* **`itermonthdays2`** – restituisce i giorni sotto forma di tuple composte dal numero del giorno del mese e dal numero del giorno della settimana;
* **`itermonthdays3`** – restituisce i giorni sotto forma di tuple composte dai numeri di anno, mese e giorno del mese. Questo metodo è disponibile a partire dalla versione 3.7 di Python;
* **`itermonthdays4`** – restituisce i giorni sotto forma di tuple composte dai numeri di anno, mese, giorno del mese e giorno della settimana. Questo metodo è disponibile a partire dalla versione 3.7 di Python.

# Il metodo `monthdays2calendar()`

La classe `Calendar` possiede diversi altri metodi utili che puoi approfondire consultando la documentazione ufficiale.

Uno di questi è il metodo `monthdays2calendar`, che accetta l'anno e il mese e restituisce una lista di settimane per quel mese specifico. Ogni settimana è a sua volta una lista di tuple composte dal numero del giorno del mese e dal numero del giorno della settimana. Guarda il codice nell'editor:

```python
import calendar  

c = calendar.Calendar()

for data in c.monthdays2calendar(2020, 12):
    print(data)

```

Nota che i numeri dei giorni al di fuori del mese corrente sono rappresentati da `0`, mentre i giorni della settimana sono espressi con un numero da 0 a 6, dove 0 è lunedì e 6 è domenica.
