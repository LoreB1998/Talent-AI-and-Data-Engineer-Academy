# Ottenere informazioni sul sistema operativo

Prima di creare la tua prima struttura di directory, vediamo come estrarre le informazioni sul sistema operativo corrente. ﺣ un'operazione immediata grazie alla funzione `uname`, la quale restituisce un oggetto dotato dei seguenti attributi:

* **`sysname`** ﻗ memorizza il nome del sistema operativo;
* **`nodename`** ﻗ memorizza il nome della macchina all'interno della rete;
* **`release`** ﻗ memorizza il rilascio (release) del sistema operativo;
* **`version`** ﻗ memorizza la versione specifica del sistema operativo;
* **`machine`** ﻗ memorizza l'identificativo dell'hardware (ad esempio, `x86_64`).

Vediamo come funziona in pratica:

```python
import os
print(os.uname())

```

**Risultato:**

```text
posix.uname_result(sysname='Linux', nodename='192d19f04766', release='4.4.0-164-generic', version='#192-Ubuntu SMP Fri Sep 13 12:02:50 UTC 2019', machine='x86_64')

```

Come puoi notare, la funzione `uname` restituisce un oggetto che racchiude i dettagli del sistema operativo. Il codice qui sopra ﺣ۷ stato eseguito su Ubuntu 16.04.6 LTS; non sorprenderti se sulla tua macchina otterrai un risultato differente, poichﺣ۸ l'output dipende strettamente dal sistema in uso.

> **Attenzione:** La funzione `os.uname()` ﺣ۷ disponibile esclusivamente su sistemi Unix/Linux. Se utilizzi Windows, questa funzione genererﺣ  un errore; puoi ottenere un risultato analogo utilizzando la funzione `uname()` presente all'interno del modulo **`platform`**.

Per distinguere rapidamente il sistema operativo corrente in modo cross-platform, il modulo `os` offre l'attributo `name`, che assume uno dei seguenti valori testuali:

* **`posix`** ﻗ se stai utilizzando un sistema operativo Unix/Linux/macOS;
* **`nt`** ﻗ se stai utilizzando Windows;
* **`java`** ﻗ se il tuo codice ﺣ۷ in esecuzione all'interno dell'ambiente Jython.

Su Ubuntu, l'attributo `name` restituirﺣ  la stringa `posix`:

```python
import os
print(os.name)

```

**Risultato:**

```text
posix

```

> **NOTA:** Sui sistemi Unix esiste il comando da terminale `uname` che restituisce le medesime informazioni (se eseguito con l'opzione `uname -a`) fornite dalla funzione `os.uname()`.


# Creare directory in Python

Il modulo `os` fornisce la funzione `mkdir` che, analogamente al comando `mkdir` presente in Unix e Windows, permette di creare una nuova cartella. La funzione richiede come argomento un percorso (path), che puﺣﺎ essere sia relativo che assoluto. Ricordiamo brevemente come si presentano i due tipi di percorso:

* **`my_first_directory`** ﻗ ﺣ۷ un percorso relativo: creerﺣ  la cartella `my_first_directory` all'interno della directory di lavoro corrente;
* **`./my_first_directory`** ﻗ ﺣ۷ un percorso relativo che punta esplicitamente alla cartella corrente. Produce lo stesso identico risultato del percorso precedente;
* **`../my_first_directory`** ﻗ ﺣ۷ un percorso relativo che creerﺣ  la cartella nella directory di livello superiore (cartella padre) rispetto a quella corrente;
* **`/python/my_first_directory`** ﻗ ﺣ۷ un percorso assoluto (su sistemi Unix) che creerﺣ  la cartella `my_first_directory` all'interno della cartella `python`, situata a sua volta nella radice del sistema (*root*).

Guarda il codice nell'editor. Mostra un esempio di creazione della cartella `my_first_directory` tramite un percorso relativo minimale, passando semplicemente il nome della cartella.

```python
import os

os.mkdir("my_first_directory")
print(os.listdir())

```

La funzione `mkdir` crea la directory nel percorso specificato. Tieni presente che se provi a eseguire questo programma due volte di seguito, Python solleverﺣ  un'eccezione di tipo **`FileExistsError`**. Questo accade perchﺣ۸ non ﺣ۷ possibile creare una cartella se questa ﺣ۷ giﺣ  esistente sul disco.

Oltre al percorso, la funzione `mkdir` puﺣﺎ accettare come secondo argomento opzionale il parametro `mode`, utile per impostare i permessi di accesso della directory. Tuttavia, su alcuni sistemi operativi (tra cui Windows), questo argomento viene ignorato. Per modificare i permessi di una cartella ti consigliamo di utilizzare la funzione `chmod`, il cui comportamento ricalca l'omonimo comando `chmod` dei sistemi Unix.

Nel codice d'esempio ﺣ۷ stata inserita anche un'altra funzione fondamentale di questo modulo: `listdir`. La funzione `os.listdir()` restituisce una lista contenente i nomi dei file e delle sottocartelle presenti nel percorso passato come argomento. Se non viene specificato alcun percorso, la funzione analizzerﺣ  la directory di lavoro corrente. ﺣ importante notare che dall'elenco restituito da `listdir` rimangono escluse le diciture speciali `'.'` e `'..'`, che vengono invece mostrate solitamente nei sistemi Unix quando si esegue il comando `ls -a`.

> **NOTA:** Sia su Windows che su Unix esiste il comando di sistema `mkdir` che richiede il percorso di una directory. L'equivalente da terminale del codice Python appena visto ﺣ۷ il comando `mkdir my_first_directory`.

# Creazione ricorsiva di directory

La funzione `mkdir` ﺣ۷ molto utile, ma cosa succede se hai bisogno di creare una cartella posizionata all'interno di un'altra cartella che ancora non esiste sul disco? Naturalmente potresti creare prima la cartella padre e successivamente la sottocartella, ma il modulo `os` ti offre una soluzione piﺣﺗ efficiente: la funzione `makedirs`.

La funzione `makedirs` consente la **creazione ricorsiva** delle directory, il che significa che verranno create automaticamente tutte le cartelle intermedie mancanti lungo il percorso specificato. Guarda il codice nell'editor per vedere come funziona in pratica:

```python
import os

os.makedirs("my_first_directory/my_second_directory")
os.chdir("my_first_directory")
print(os.listdir())

```

**Risultato del codice:**

```text
['my_second_directory']

```

Questo frammento di codice genera due directory: la prima (`my_first_directory`) viene creata all'interno della cartella di lavoro corrente, mentre la seconda (`my_second_directory`) nasce direttamente dentro la prima. Non ﺣ۷ stato necessario spostarsi manualmente nella prima cartella per poter generare la seconda, poichﺣ۸ ci ha pensato in autonomia la funzione `makedirs`.

Nel codice abbiamo utilizzato la funzione `chdir` per spostarci all'interno di `my_first_directory` e dimostrare, tramite `listdir()`, che la sottocartella `my_second_directory` ﺣ۷ stata effettivamente creata al suo interno. La funzione `os.chdir()` riceve come argomento un percorso relativo o assoluto e cambia la directory di lavoro corrente impostandola sul percorso specificato.

> **NOTA:** L'equivalente della funzione `makedirs` sui sistemi Unix ﺣ۷ il comando `mkdir` accompagnato dal flag `-p` (`mkdir -p percorso`), mentre su Windows il comando standard `mkdir` supporta nativamente la creazione dell'intero percorso (`mkdir percorso`).

# Dove mi trovo adesso?

Ora sai come creare le directory e come spostarti da una all'altra. Tuttavia, quando lavori con strutture di cartelle molto ampie e ramificate, puﺣﺎ capitare di perdere traccia della posizione esatta in cui il programma sta operando.

Il modulo `os` risolve questo problema fornendo la funzione `getcwd` (*get current working directory*), che restituisce il percorso della cartella di lavoro corrente. Guarda il codice nell'editor per osservarne l'applicazione pratica:

```python
import os

os.makedirs("my_first_directory/my_second_directory")
os.chdir("my_first_directory")
print(os.getcwd())
os.chdir("my_second_directory")
print(os.getcwd())

```

**Risultato indicativo:**

```text
.../my_first_directory
.../my_first_directory/my_second_directory

```

In questo esempio creiamo la struttura di cartelle annidate. Subito dopo cambiamo la directory corrente spostandoci in `my_first_directory` e ne stampiamo il percorso assoluto tramite `getcwd()` (prima riga dell'output). Successivamente entriamo in `my_second_directory` e interroghiamo nuovamente il sistema con `getcwd()` (seconda riga dell'output). Come puoi notare, la funzione restituisce sempre il percorso assoluto completo.

> **NOTA:** Sui sistemi Unix-like l'equivalente della funzione `getcwd()` ﺣ۷ il comando `pwd` (*print working directory*), che stampa a schermo il nome della cartella di lavoro corrente.

# Eliminare directory in Python

Il modulo `os` consente anche di rimuovere le directory dal disco, offrendo la possibilitﺣ  di eliminare una singola cartella oppure un intero percorso comprensivo di sottocartelle. Per rimuovere una singola directory si utilizza la funzione `rmdir`, passando il percorso come argomento. Guarda il codice nell'editor:

```python
import os

os.mkdir("my_first_directory")
print(os.listdir())
os.rmdir("my_first_directory")
print(os.listdir())

```

L'esempio ﺣ۷ lineare: per prima cosa viene creata la cartella `my_first_directory`, che viene subito dopo rimossa tramite la funzione `rmdir`. La funzione `listdir()` viene invocata nei passaggi intermedi per confermare visivamente l'avvenuta rimozione (alla fine restituisce infatti una lista vuota). Quando utilizzi `rmdir`, assicurati che la directory esista e sia **completamente vuota**, altrimenti il sistema solleverﺣ  un'eccezione.

Per rimuovere una cartella insieme alle sue sottocartelle puoi affidarti alla funzione `removedirs`, inserendo come argomento il percorso completo che desideri cancellare:

```python
import os
 
os.makedirs("my_first_directory/my_second_directory")
os.removedirs("my_first_directory/my_second_directory")
print(os.listdir())

```

Esattamente come visto per `rmdir`, se una delle directory specificate nel percorso non esiste o non ﺣ۷ vuota, il processo si interromperﺣ  sollevando un'eccezione.

> **NOTA:** Sia su Windows che su Unix esiste il comando `rmdir` che rimuove le cartelle vuote. Per eliminare una cartella e tutto il suo contenuto (inclusi i file), nei sistemi Unix si adopera il comando `rm` con il flag `-r` (`rm -r`), mentre su Windows si utilizza il comando `rmdir` con lo switch `/s`.

# La funzione `system()`

Tutte le funzioni dedicate alla gestione del file system presentate in questa parte del corso possono essere sostituite dall'utilizzo di una funzione piﺣﺗ generica chiamata `system`, la quale esegue un comando di sistema passandoglielo sotto forma di stringa di testo.

La funzione `os.system()` ﺣ۷ disponibile sia su Windows che su Unix, ma il valore restituito varia a seconda della piattaforma:

* Su **Windows** restituisce il valore di ritorno della shell dopo aver eseguito il comando;
* Su **Unix** restituisce lo stato di uscita (*exit status*) del processo generato.

Diamo un'occhiata al codice nell'editor per vederne l'applicazione pratica:

```python
import os

returned_value = os.system("mkdir my_first_directory")
print(returned_value)

```

**Risultato:**

```text
0

```

Questo esempio funziona correttamente su entrambi i sistemi operativi. Nel nostro caso abbiamo ricevuto come output il valore `0`, che nei sistemi Unix e Windows indica che il comando ﺣ۷ stato completato con successo. Questo conferma che la cartella `my_first_directory` ﺣ۷ stata creata correttamente sul disco.
