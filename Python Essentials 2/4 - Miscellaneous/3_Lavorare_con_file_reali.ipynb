### Elaborazione di file di testo

In questa lezione prepareremo un semplice file di testo con un contenuto breve e lineare.
Ti mostreremo alcune tecniche di base che puoi utilizzare per leggere il contenuto del file al fine di elaborarlo.
L'elaborazione sarà molto semplice: copierai il contenuto del file sulla console e conterai tutti i caratteri che il programma ha letto.
Ma ricorda: la nostra definizione di file di testo è molto rigorosa. Per noi, si tratta di un file di testo normale (plain text) – può contenere solo testo, senza alcuna decorazione aggiuntiva (formattazione, font diversi, ecc.).
Ecco perché dovresti evitare di creare il file utilizzando un elaboratore di testi avanzato come MS Word, LibreOffice Writer o simili. Usa gli strumenti essenziali offerti dal tuo sistema operativo: Blocco note, vim, gedit, ecc.
Se i tuoi file di testo contengono caratteri nazionali non coperti dal set di caratteri ASCII standard, potresti aver bisogno di un passaggio aggiuntivo. L'invocazione della funzione `open()` potrebbe richiedere un argomento che indichi la codifica specifica del testo.
Ad esempio, se stai usando un sistema operativo Unix/Linux configurato per utilizzare UTF-8 come impostazione di sistema, la funzione `open()` potrebbe apparire come segue:

```python
stream = open('file.txt', 'rt', encoding='utf-8')

```

dove l'argomento `encoding` deve essere impostato su un valore che sia una stringa rappresentante la corretta codifica del testo (in questo caso, UTF-8).
Consulta la documentazione del tuo sistema operativo per trovare il nome della codifica adeguato al tuo ambiente.

La lettura del contenuto di un file di testo può essere eseguita utilizzando diversi metodi – nessuno di essi è migliore o peggiore degli altri. Dipende da quale preferisci.
Alcuni di essi saranno a volte più pratici e a volte più problematici. Sii flessibile. Non aver paura di cambiare le tue preferenze.
Il più basilare di questi metodi è quello offerto dalla funzione `read()`, che hai potuto vedere in azione nella lezione precedente.
Se applicata a un file di testo, la funzione è in grado di:

* leggere un numero desiderato di caratteri (compreso uno solo) dal file e restituirli sotto forma di stringa;
* leggere l'intero contenuto del file e restituirlo sotto forma di stringa;
* se non c'è più nulla da leggere (la testina di lettura virtuale raggiunge la fine del file), la funzione restituisce una stringa vuota.

Inizieremo con la variante più semplice e useremo un file chiamato `text.txt`. Il file ha il seguente contenuto:

```text
Output
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.

```

Ora guarda il codice nell'editor e analizziamolo.

````python
from os import strerror

try:
    counter = 0
    stream = open('text.txt', "rt")
    char = stream.read(1)
    while char != '':
        print(char, end='')
        counter += 1
        char = stream.read(1)
    stream.close()
    print("\n\nCharacters in file:", counter)
except IOError as e:
    print("I/O error occurred: ", strerror(e.errno))
````

La routine è piuttosto semplice:

* usi il meccanismo `try-except` e apri il file con il nome prestabilito (nel nostro caso `text.txt`);
* provi a leggere il primissimo carattere dal file (`char = stream.read(1)`);
* se l'operazione va a buon fine (cosa dimostrata dal risultato positivo del controllo della condizione del `while`), stampi il carattere (nota l'argomento `end=` – è importante! Non vuoi passare a una nuova riga dopo ogni singolo carattere!);
* aggiorni anche il contatore (`counter`);
* provi a leggere il carattere successivo, e il processo si ripete.

Se sei assolutamente sicuro che la dimensione del file sia sicura e che tu possa leggere l'intero file in memoria tutto in una volta, puoi farlo – la funzione `read()`, invocata senza alcun argomento o con un argomento che restituisce `None`, farà il lavoro per te.

Ricorda: leggere un file grande un terabyte usando questo metodo può mandare in blocco il tuo sistema operativo. Non aspettarti miracoli – la memoria del computer non è infinitamente elastica.

## `readline()`

Se desideri trattare il contenuto del file come un insieme di righe e non come un semplice mucchio di caratteri, il metodo `readline()` ti sarà di grande aiuto.

Questo metodo prova a leggere una riga completa di testo dal file e, in caso di successo, la restituisce sotto forma di stringa. Altrimenti, se non c'è più nulla da leggere, restituisce una stringa vuota.

Questo approccio apre nuove opportunità: ora puoi anche contare facilmente le righe, e non solo i caratteri.

Sfruttiamo questa funzionalità. Dai un'occhiata al codice nell'editor.

```python
from os import strerror

try:
    character_counter = line_counter = 0
    stream = open('text.txt', 'rt')
    line = stream.readline()
    while line != '':
        line_counter += 1
        for char in line:
            print(char, end='')
            character_counter += 1
        line = stream.readline()
    stream.close()
    print("\n\nCharacters in file:", character_counter)
    print("Lines in file:     ", line_counter)
except IOError as e:
    print("I/O error occurred:", strerror(e.errno))
```


## `readlines()`
Un altro metodo che tratta il file di testo come un insieme di righe, e non di caratteri, è `readlines()`.

Il metodo `readlines()`, quando viene invocato senza argomenti, prova a leggere l'intero contenuto del file e restituisce una lista di stringhe, in cui ogni elemento corrisponde a una riga del file.

Se non sei sicuro che la dimensione del file sia sufficientemente ridotta e non vuoi rischiare di sovraccaricare il sistema operativo, puoi impostare il metodo `readlines()` in modo che non legga più di un determinato numero di byte alla volta (il valore restituito rimane lo stesso: una lista di stringhe).

Siti libero di sperimentare con il seguente codice di esempio per capire come funziona il metodo `readlines()`:

```python
stream = open("text.txt")
print(stream.readlines(20))
print(stream.readlines(20))
print(stream.readlines(20))
print(stream.readlines(20))
stream.close()

```

La dimensione massima accettata per il buffer di input viene passata al metodo come argomento. Puoi aspettarti che `readlines()` sia in grado di elaborare il contenuto di un file in modo più efficace rispetto a `readline()`, poiché potrebbe richiedere un numero inferiore di invocazioni.

> **Nota:** quando non c'è più nulla da leggere dal file, il metodo restituisce una lista vuota. Puoi usare questo comportamento per rilevare la fine del file.
> Per quanto riguarda la dimensione del buffer, puoi ipotizzare che aumentarla possa migliorare le prestazioni di input, ma non esiste una regola d'oro: prova a trovare tu stesso i valori ottimali.


Guarda il codice nell'editor. Lo abbiamo modificato per mostrarti come utilizzare `readlines()`.

```python
from os import strerror

try:
    ccnt = lcnt = 0
    s = open('text.txt', 'rt')
    lines = s.readlines(20)
    while len(lines) != 0:
        for line in lines:
            lcnt += 1
            for ch in line:
                print(ch, end='')
                ccnt += 1
        lines = s.readlines(10)
    s.close()
    print("\n\nCharacters in file:", ccnt)
    print("Lines in file:     ", lcnt)
except IOError as e:
    print("I/O error occurred:", strerror(e.errno))

```

Abbiamo deciso di utilizzare un buffer lungo 15 byte. Non considerarla una raccomandazione. Abbiamo usato questo valore specifico per evitare che la prima invocazione di `readlines()` consumi l'intero file. Vogliamo costringere il metodo a lavorare di più e mostrare le sue capacità. Nel codice sono presenti due cicli annidati: quello esterno usa il risultato di `readlines()` per iterare, mentre quello interno stampa le linee carattere per carattere.

L'ultimo esempio che vogliamo presentare mostra una caratteristica molto interessante dell'oggetto restituito dalla funzione `open()` in modalità testo. Pensiamo che possa sorprenderti: l'oggetto è un'istanza della classe degli iterabili. Strano? Per nulla. Utile? Sì, assolutamente. Il protocollo di iterazione definito per l'oggetto file è molto semplice: il suo metodo `__next__` si limita a restituire la riga successiva letta dal file. Inoltre, puoi aspettarti che l'oggetto invochi automaticamente `close()` quando una qualsiasi delle letture del file raggiunge la fine del file stesso.

Guarda l'editor e osserva come il codice sia diventato semplice e chiaro.

```python
from os import strerror

try:
	ccnt = lcnt = 0
	for line in open('text.txt', 'rt'):
		lcnt += 1
		for ch in line:
			print(ch, end='')
			ccnt += 1
	print("\n\nCharacters in file:", ccnt)
	print("Lines in file:     ", lcnt)
except IOError as e:
	print("I/O error occurred: ", strerror(e.errno))

```

## Gestione dei file di testo: `write()`

Scrivere file di testo sembra essere più semplice, poiché di fatto esiste un solo metodo che può essere utilizzato per eseguire questo compito. Il metodo si chiama `write()` e si aspetta un solo argomento: una stringa che verrà trasferita in un file aperto (non dimenticare: la modalità di apertura deve riflettere il modo in cui i dati vengono trasferiti – scrivere su un file aperto in modalità lettura non funzionerà).

All'argomento di `write()` non viene aggiunto alcun carattere di nuova riga, quindi devi aggiungerlo tu stesso se vuoi che il file sia riempito con un certo numero di righe. L'esempio nell'editor mostra un codice molto semplice che crea un file chiamato `newtext.txt` (nota: la modalità di apertura `w` garantisce che il file venga creato da zero, anche se esiste già e contiene dati) e poi vi inserisce dieci righe.

```python
from os import strerror

try:
	file = open('newtext.txt', 'wt') # Viene creato un nuovo file (newtext.txt).
	for i in range(10):
		s = "line #" + str(i+1) + "\n"
		for char in s:
			file.write(char)
	file.close()
except IOError as e:
	print("I/O error occurred: ", strerror(e.errno))

```

La stringa da registrare è composta dalla parola *line*, seguita dal numero di riga. Abbiamo deciso di scrivere il contenuto della stringa carattere per carattere (operazione eseguita dal ciclo `for` interno), ma non sei obbligato a procedere in questo modo. Volevamo solo mostrarti che `write()` è in grado di operare sui singoli caratteri.

Il codice crea un file riempito con il seguente testo:

```text
Output
line #1
line #2
line #3
line #4
line #5
line #6
line #7
line #8
line #9
line #10

```

Sei in grado di stampare il contenuto del file sulla console? Ti incoraggiamo a testare il comportamento del metodo `write()` localmente sulla tua macchina.

Guarda l'esempio nell'editor. Abbiamo modificato il codice precedente per scrivere righe intere sul file di testo.

```python
from os import strerror

try:
    file = open('newtext.txt', 'wt')
    for i in range(10):
        file.write("line #" + str(i+1) + "\n")
    file.close()
except IOError as e:
    print("I/O error occurred: ", strerror(e.errno))

```

Il contenuto del file appena creato è lo stesso.
Nota: puoi usare lo stesso metodo per scrivere nello stream `stderr`, ma non cercare di aprirlo, poiché è sempre aperto implicitamente. Per esempio, se vuoi inviare una stringa di messaggio a `stderr` per distinguerla dal normale output del programma, il codice potrebbe apparire così:

```python
import sys
sys.stderr.write("Error message")

```

## Che cos'è un `bytearray`?

Prima di iniziare a parlare di file binari, dobbiamo parlarti di una delle classi specializzate che Python usa per memorizzare dati amorfi. I dati amorfi sono dati che non hanno una forma o una struttura specifica – sono semplicemente una serie di byte. Questo non significa che questi byte non possano avere un proprio significato o non possano rappresentare un oggetto utile, come ad esempio una grafica bitmap.

L'aspetto più importante è che nel punto in cui entriamo in contatto con i dati, non siamo in grado di sapere nulla al riguardo, o semplicemente non vogliamo farlo. I dati amorfi non possono essere memorizzati usando nessuno dei mezzi presentati in precedenza – non sono né stringhe né liste. Deve esserci un contenitore speciale in grado di gestire tali dati.

Python ha più di un contenitore di questo tipo – uno di questi è una classe specializzata chiamata `bytearray` – come suggerisce il nome, è un array che contiene byte (amorfi). Se vuoi avere un contenitore di questo tipo, ad esempio per leggere un'immagine bitmap ed elaborarla in qualche modo, devi crearlo esplicitamente utilizzando uno dei costruttori disponibili. Dai un'occhiata:

```python
data = bytearray(10)

```

Tale invocazione crea un oggetto `bytearray` in grado di memorizzare dieci byte. Nota: questo costruttore riempie l'intero array con zeri.

I `bytearray` assomigliano alle liste sotto molti aspetti. Per esempio, sono mutabili, sono soggetti alla funzione `len()` ed è possibile accedere a qualsiasi loro elemento usando l'indicizzazione convenzionale. C'è una limitazione importante: non devi impostare alcun elemento dell'array di byte con un valore che non sia un numero intero (la violazione di questa regola causerà un'eccezione `TypeError`) e non ti è consentito assegnare un valore che non rientri nell'intervallo da 0 a 255 inclusi (a meno che tu non voglia provocare un'eccezione `ValueError`).

Puoi trattare tutti gli elementi di un array di byte come valori interi – proprio come nell'esempio nell'editor:

```python
data = bytearray(10)

for i in range(len(data)):
    data[i] = 10 - i

for b in data:
    print(hex(b))

```

Nota: abbiamo usato due metodi per iterare gli array di byte e abbiamo fatto uso della funzione `hex()` per vedere gli elementi stampati come valori esadecimali. Ora ti mostreremo come scrivere un array di byte in un file binario – binario, in quanto non vogliamo salvare la sua rappresentazione leggibile – vogliamo scrivere una copia uno-a-uno del contenuto della memoria fisica, byte per byte.

Quindi, come si scrive un array di byte in un file binario? Guarda il codice nell'editor.

```python
from os import strerror

data = bytearray(10)

for i in range(len(data)):
    data[i] = 10 + i

try:
    bf = open('file.bin', 'wb')
    bf.write(data)
    bf.close()
except IOError as e:
    print("I/O error occurred:", strerror(e.errno))

# Il codice che legge i byte dallo stream dovrebbe andare qui.

```

Analizziamolo: prima inizializziamo il `bytearray` con valori successivi a partire da 10; se vuoi che il contenuto del file sia chiaramente leggibile, sostituisci 10 con qualcosa come `ord('a')` – questo produrrà byte contenenti valori corrispondenti alla parte alfabetica del codice ASCII (non pensare che questo renderà il file un file di testo – è comunque binario, poiché è stato creato con il flag `wb`);
poi creiamo il file usando la funzione `open()` – l'unica differenza rispetto alle varianti precedenti è la modalità di apertura contenente il flag `b`;
il metodo `write()` prende il suo argomento (`bytearray`) e lo invia (nella sua interezza) al file;
lo stream viene poi chiuso in modo convenzionale.

Il metodo `write()` restituisce il numero di byte scritti correttamente. Se i valori differiscono dalla lunghezza degli argomenti del metodo, ciò potrebbe segnalare alcuni errori di scrittura. In questo caso non abbiamo fatto uso del risultato – questo potrebbe non essere appropriato in ogni situazione. Prova a eseguire il codice e ad analizzare il contenuto del file di output appena creato. Lo utilizzerai nel passaggio successivo.

## Come leggere i byte da uno stream

La lettura da un file binario richiede l'uso di un metodo specializzato chiamato `readinto()`, poiché il metodo non crea un nuovo oggetto array di byte, ma ne riempie uno precedentemente creato con i valori presi dal file binario.

Nota:

* il metodo restituisce il numero di byte letti correttamente;
* il metodo tenta di riempire tutto lo spazio disponibile all'interno del suo argomento; se nel file ci sono più dati rispetto allo spazio nell'argomento, l'operazione di lettura si arresterà prima della fine del file; in caso contrario, il risultato del metodo potrebbe indicare che l'array di byte è stato riempito solo parzialmente (anche il risultato lo mostrerà, e la parte dell'array non utilizzata dal contenuto appena letto rimane intatta).

Guarda il codice completo qui sotto:

```python
from os import strerror

data = bytearray(10)

try:
    binary_file = open('file.bin', 'rb')
    binary_file.readinto(data)
    binary_file.close()

    for b in data:
        print(hex(b), end=' ')
except IOError as e:
    print("I/O error occurred:", strerror(e.errno))

```

Analizziamolo: dapprima apriamo il file (quello creato con il codice precedente) con la modalità descritta come `rb`; poi leggiamo il suo contenuto nell'array di byte chiamato `data`, della dimensione di dieci byte; infine, stampiamo il contenuto dell'array di byte – è lo stesso che ti aspettavi? Esegui il codice e controlla se funziona.

Un modo alternativo per leggere il contenuto di un file binario è offerto dal metodo chiamato `read()`. Invocato senza argomenti, tenta di leggere tutto il contenuto del file in memoria, rendendolo parte di un oggetto appena creato della classe `bytes`. Questa classe presenta alcune somiglianze con `bytearray`, a eccezione di una differenza significativa: è immutabile.

Fortunatamente non ci sono ostacoli alla creazione di un array di byte prendendo il suo valore iniziale direttamente dall'oggetto `bytes`, proprio come qui:

```python
from os import strerror

try:
    binary_file = open('file.bin', 'rb')
    data = bytearray(binary_file.read())
    binary_file.close()

    for b in data:
        print(hex(b), end=' ')

except IOError as e:
    print("I/O error occurred:", strerror(e.errno))

```

Fai attenzione: non usare questo tipo di lettura se non sei sicuro che il contenuto del file possa entrare nella memoria disponibile.

Se il metodo `read()` viene invocato con un argomento, questo specifica il numero massimo di byte da leggere. Il metodo tenta di leggere il numero desiderato di byte dal file e la lunghezza dell'oggetto restituito può essere utilizzata per determinare il numero di byte effettivamente letti. Puoi usare il metodo proprio come qui:

```python
try:
    binary_file = open('file.bin', 'rb')
    data = bytearray(binary_file.read(5))
    binary_file.close()

    for b in data:
        print(hex(b), end=' ')

except IOError as e:
    print("I/O error occurred:", strerror(e.errno))

```

## Copiare i file - uno strumento semplice e funzionale

Ora unirai tutte queste nuove conoscenze, vi aggiungerai alcuni elementi freschi e le userai per scrivere un codice reale in grado di copiare effettivamente il contenuto di un file. Naturalmente lo scopo non è quello di creare un sostituto migliore per comandi come `copy` (MS Windows) o `cp` (Unix/Linux), ma di vedere un modo possibile per creare uno strumento funzionante, anche se nessuno volesse usarlo.

Guarda il codice nell'editor.

```python
from os import strerror

srcname = input("Enter the source file name: ")
try:
    src = open(srcname, 'rb')
except IOError as e:
    print("Cannot open the source file: ", strerror(e.errno))
    exit(e.errno)	

dstname = input("Enter the destination file name: ")
try:
    dst = open(dstname, 'wb')
except Exception as e:
    print("Cannot create the destination file: ", strerror(e.errno))
    src.close()
    exit(e.errno)	

buffer = bytearray(65536)
total  = 0
try:
    readin = src.readinto(buffer)
    while readin > 0:
        written = dst.write(buffer[:readin])
        total += written
        readin = src.readinto(buffer)
except IOError as e:
    print("Cannot create the destination file: ", strerror(e.errno))
    exit(e.errno)	
    
print(total,'byte(s) succesfully written')
src.close()
dst.close()

```

Analizziamolo:

* **linee da 3 a 8:** chiedono all'utente il nome del file da copiare e tentano di aprirlo in lettura; interrompono l'esecuzione del programma se l'apertura fallisce; nota: usa la funzione `exit()` per arrestare l'esecuzione del programma e passare il codice di completamento al sistema operativo; qualsiasi codice di completamento diverso da 0 indica che il programma ha riscontrato dei problemi; usa il valore di `errno` per specificare la natura del problema;
* **linee da 10 a 16:** ripetono quasi la stessa azione, ma questa volta per il file di output;
* **linea 18:** preparano una porzione di memoria per il trasferimento dei dati dal file sorgente a quello di destinazione; tale area di trasferimento viene spesso chiamata buffer, da cui il nome della variabile; la dimensione del buffer è arbitraria – in questo caso abbiamo deciso di usare 64 kilobyte; tecnicamente, un buffer più grande è più veloce nel copiare gli elementi, poiché un buffer più grande significa meno operazioni di I/O; in realtà esiste sempre un limite, il cui superamento non porta ulteriori miglioramenti; testalo tu stesso se vuoi.
* **linea 19:** conta i byte copiati – questo è il contatore e il suo valore iniziale;
* **linea 21:** tenta di riempire il buffer per la primissima volta;
* **linea 22:** finché si ottiene un numero di byte diverso da zero, ripete le stesse azioni;
* **linea 23:** scrive il contenuto del buffer nel file di output (nota: abbiamo usato uno slice per limitare il numero di byte scritti, poiché `write()` preferisce sempre scrivere l'intero buffer);
* **linea 24:** aggiorna il contatore;
* **linea 25:** legge il blocco di file successivo;
* **linee da 30 a 32:** un po' di pulizia finale – il lavoro è fatto.
