# Tris Rule Based

1. **Mossa Vincente**: Se c'è una mossa che permette di vincere immediatamente, il PC la esegue.
2. **Mossa Bloccante**: Se c'è una mossa dell'avversario che potrebbe portare alla vittoria, il PC la blocca.
3. **Centro**: Se la casella centrale è disponibile, il PC la sceglie.
4. **Angoli**: Se un angolo è disponibile, il PC lo sceglie.
5. **Latari**: Se un lato è disponibile, il PC lo sceglie.



````python
import time
import random

class Tris:
    def __init__(self):
        self.board = [" " for _ in range(9)]
        self.current_winner = None 

    def stampa_griglia(self):
        for i in range(0, 9, 3):
            print(f" {self.board[i]} | {self.board[i+1]} | {self.board[i+2]} ")
            if i < 6:
                print("---+---+---")
        print("\n")

    def mosse_disponibili(self):
        return [i for i, spot in enumerate(self.board) if spot == " "]

    def esegui_mossa(self, posizione, segno):
        if self.board[posizione] == " ":
            self.board[posizione] = segno
            if self.controlla_vittoria(posizione, segno):
                self.current_winner = segno
            return True
        return False

    def controlla_vittoria(self, pos, segno):
        # Controllo riga
        r_idx = pos // 3
        if all([self.board[r_idx * 3 + i] == segno for i in range(3)]): return True
        # Controllo colonna
        c_idx = pos % 3
        if all([self.board[c_idx + i * 3] == segno for i in range(3)]): return True
        # Controllo diagonali
        if pos % 2 == 0:
            if all([self.board[i] == segno for i in [0,4,8]]): return True
            if all([self.board[i] == segno for i in [2,4,6]]): return True
        return False

    def mossa_strategica(self, segno):
        avversario = "O" if segno == "X" else "X"
        vuote = self.mosse_disponibili()

        # 1. ATTACCO: Se posso vincere ora, lo faccio.
        for mossa in vuote:
            self.board[mossa] = segno
            vittoria = self.controlla_vittoria(mossa, segno)
            
            # Reset immediato dello stato dopo il test
            self.board[mossa] = " "
            self.current_winner = None
            
            if vittoria:
                return mossa

        # 2. DIFESA: Se l'avversario sta per vincere, lo blocco.
        for mossa in vuote:
            self.board[mossa] = avversario
            pericolo = self.controlla_vittoria(mossa, avversario)
            
            # Reset immediato dello stato dopo il test
            self.board[mossa] = " "
            self.current_winner = None
            
            if pericolo:
                return mossa

        # 3. POSIZIONE FORTE: Centro
        if 4 in vuote:
            return 4

        # 4. OPPORTUNITÀ: Angoli
        angoli = [i for i in [0, 2, 6, 8] if i in vuote]
        if angoli:
            return random.choice(angoli)

        # 5. RIPIEGO: Caselle laterali
        return random.choice(vuote)
````

````python
def main():
    game = Tris()
    print("Seleziona modalità:\n1. Umano (X) vs Macchina (O)\n2. Macchina (X) vs Macchina (O)")
    scelta = input("Scelta: ")
    
    turno = "X" # X inizia sempre per primo
    while game.mosse_disponibili() and not game.current_winner: # Finché ci sono mosse disponibili e non c'è un vincitore
        game.stampa_griglia() # Stampa la griglia ad ogni turno
        
        if scelta == "1" and turno == "X": # Se è il turno dell'umano (X) e la modalità è 1, chiedi all'utente di inserire una mossa
            try:
                mossa = int(input(f"Turno di {turno}. Scegli (0-8): ")) # Chiede all'utente di inserire un numero da 0 a 8 per la mossa
            except: continue
        else:
            print(f"La macchina ({turno}) sta pensando...")
            mossa = game.mossa_strategica(turno) # Ottiene la mossa migliore per l'IA
            if scelta == "2": time.sleep(0.8) # Rallenta per vedere le mosse

        if game.esegui_mossa(mossa, turno): # Se la mossa è valida, eseguila e passa al turno successivo
            turno = "O" if turno == "X" else "X" # Cambia il turno

    game.stampa_griglia() # Stampa la griglia finale
    if game.current_winner: # Se c'è un vincitore, stampa chi ha vinto
        print(f"VITTORIA: {game.current_winner}") # Stampa il vincitore
    else:
        print("PAREGGIO!") # Se non c'è un vincitore e non ci sono mosse disponibili, è un pareggio

main()
````

In questo caso l'unica possibilità di vincere per l'utente è utilizzare una forchetta, ma il PC è in grado di bloccare entrambe le minacce contemporaneamente, garantendo così la vittoria.
