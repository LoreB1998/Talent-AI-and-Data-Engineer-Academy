## Esercizio 5

Creare un programma che permette all'utente di inserire una serie di numeri interrotta da 0, conti quante volte compare un numero, chiesto all'utente, nella lista

````python
ls = []
while True:
    n = int(input("Inserisci un valore > 0 (0 per interrompere): "))
    if n != 0:
        ls.append(n)
    else:
        print("Lista riempita")
        break

a = int(input("Inserisci il valore di cui vuoi sapere la frequenza: "))
count = 0
for num in ls:
    if a == num:
        count +=1

print(f"In {ls} ci sono {count} valori {a}")
````

````python
# Alternativa
ls = []
while True:
    n = int(input("Inserisci un valore > 0 (0 per interrompere): "))
    if n != 0:
        ls.append(n)
    else:
        print("Lista riempita!")
        break

a = int(input("Inserisci il valore di cui vuoi sapere la frequenza: "))
counts = ls.count(a)

print(f"In {ls} ci sono {counts} valori {a}")
````

# API

Le API (Application Programming Interface) sono un insieme di regole e protocolli che permettono a diverse applicazioni di comunicare tra loro. In pratica, un'API definisce come un software può interagire con un altro software, specificando le modalità di richiesta e risposta dei dati. Le API sono utilizzate per consentire l'integrazione tra diverse applicazioni, facilitando lo scambio di informazioni e funzionalità. Ad esempio, un'API può essere utilizzata per consentire a un'applicazione di accedere a dati o servizi offerti da un'altra applicazione, come ad esempio un servizio di pagamento o un database. Le API sono fondamentali per lo sviluppo di applicazioni moderne, poiché consentono di creare soluzioni più flessibili e scalabili, sfruttando le funzionalità offerte da altri software senza doverle sviluppare da zero.

Vediamo un esempi con random.org, un sito che offre un'API per generare numeri casuali. Per utilizzare l'API, è necessario inviare una richiesta HTTP al server di random.org, specificando i parametri desiderati, come ad esempio il numero di numeri casuali da generare e il range di valori. Il server risponderà con i numeri casuali generati, che possono essere utilizzati nella nostra applicazione.

Installiamo la libreria `requests` per inviare richieste HTTP:


````python
import requests
import json

RANDOMORG_KEY = "e83b97e6-75d6-46c2-a8a1-827f8d962938"

def getRandom(min_val, max_val, num=1, api_key=RANDOMORG_KEY):
    if not RANDOMORG_KEY:
        raise RuntimeError("Missing Key")
    
    response = requests.post(
        "https://api.random.org/json-rpc/4/invoke", # url dell'API
        json = {
            "jsonrpc": "2.0", # versione del protocollo JSON-RPC
            "method" : "generateIntegers", # metodo dell'API per generare numeri interi casuali
            "params" : { 
                "apiKey" : api_key, # chiave API per autenticarsi
                "n" : num, # numero di numeri casuali da generare
                "min" : min_val, # valore minimo del range
                "max" : max_val # valore massimo del range
            },
            "id": 1, # identificatore della richiesta (può essere qualsiasi valore univoco)
        }
    )
    response.raise_for_status() # verifica che la richiesta sia andata a buon fine (altrimenti solleva un'eccezione)
    print(response) # stampa la risposta grezza (per debug)
    payload = response.json() # converte la risposta in formato JSON in un dizionario Python

    # Estrazione del numero (l'API restituisce una lista in 'data')
    numero_casuale = payload['result']['random']['data'][0] # accede al primo numero generato (dato che n=1, c'è solo un numero nella lista)
    return numero_casuale # restituisce il numero casuale generato


try:
    risultato = getRandom(1, 100)
    print(f"Il numero estratto è: {risultato}")
except Exception as e:
    print(f"Errore: {e}")
````

## Esercizio 7

Creare una rubrica telefonica usando un dizionario. Il programma deve permettere di: 
- aggiungere un contatto 
- cercare un numero tramite nome 
- stampare tutta la rubrica

````python
import json
rubrica = {}

def inserisci(nome, numero):
    nome = nome.strip().title()
    numero = numero.strip()
    if nome in rubrica:
        print(f"Il contatto {nome} è già presente con numero {rubrica[nome]}.")
    else:
        rubrica[nome] = numero
        print(f"Contatto {nome} inserito con numero {numero}.")


def ricerca(nome):
    nome_cercato = nome.strip().lower()
    found = False
    
    # Usiamo .items() per avere nome e numero subito pronti
    for contatto, numero in rubrica.items():
        # Portiamo il nome in rubrica a minuscolo solo per il confronto
        if contatto.lower().startswith(nome_cercato):
            print(f" Il contatto {contatto} ha il numero {numero}.")
            found = True
            
    if not found:
        # Usiamo .title() solo qui per mostrare un output carino all'utente
        print(f"Contatto {nome.strip().title()} non trovato.")


def delete(nome):
    nome_cercato = nome.strip().lower()
    found = False
    
    for contatto in list(rubrica.keys()): # usiamo list() per evitare problemi di modifica del dizionario durante l'iterazione
        if contatto.lower().startswith(nome_cercato):
            del rubrica[contatto]
            print(f"Contatto {contatto} eliminato.")
            found = True
            
    if not found:
        print(f"Contatto {nome.strip().title()} non trovato per l'eliminazione.")


def stampa():
    for nome, numero in sorted(rubrica.items()):
        print(f"{nome}: {numero}")
    print()

while True:
    print("\n--- RUBRICA TELEFONICA ---")
    print("1. Aggiungi un contatto")
    print("2. Cerca un numero tramite nome")
    print("3. Stampa tutta la rubrica")
    print("4. Elimina un contatto")
    print("5. Esci")

    scelta = input("Scegli un'opzione (1-5): ")

    if scelta == '1':
        nome = input("Inserisci il nome del contatto: ")
        numero = input("Inserisci il numero di telefono: ")
        inserisci(nome, numero)
    elif scelta == '2':
        nome = input("Inserisci il nome del contatto da cercare: ")
        ricerca(nome)
    elif scelta == '3':
        stampa()
    elif scelta == '4':
        nome = input("Inserisci il nome del contatto da eliminare: ")
        delete(nome)
    elif scelta == '5':
        print("Uscita dalla rubrica. Arrivederci!")
        break
    else:
        print("Opzione non valida. Riprova.")
````
